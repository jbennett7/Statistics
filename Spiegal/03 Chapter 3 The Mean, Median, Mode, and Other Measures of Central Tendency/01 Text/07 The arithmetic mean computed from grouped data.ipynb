# The arithmetic mean computed from grouped data
When data are presented in a frequency distribution, all values falling within a given class interval are considered to be coincident with the class mark, or midpoint, of the interval. Formulas [(2)](04%20The%20arithmetic%20mean.ipynb#eq2) and [(6)](06%20Properties%20of%20the%20arithmetic%20mean.ipynb#eq6) are valid for such grouped data if we interpret $X_j$ as the class mark, $f_j$ as its corresponding class frequency, $A$ as any guessed or assumed class mark, and $d_j=X_j-A$ as the deviations of $X_j$ from $A$.

Computations using formulas [(2)](04%20The%20arithmetic%20mean.ipynb#eq2) and [(6)](06%20Properties%20of%20the%20arithmetic%20mean.ipynb#eq6) are sometimes called the _long_ and _short methods_, respectively (see Problems [3.15](../02%20Solved%20Problems/02%20The%20arithmetic%20mean.ipynb#Problem-3-15) and [3.20](../02%20Solved%20Problems/04%20The%20arithmetic%20mean%20computed%20from%20grouped%20data.ipynb#Problem-3-20)).

If class intervals all have equal size $c$, the deviations $d_j=X_j-A$ can all be expressed as $cu_j$, where $u_j$ can be positive or negative integers or zero (i.e., $0, \pm1, \pm2, \pm3,\ldots$), and formula [(6)](06%20Properties%20of%20the%20arithmetic%20mean.ipynb#eq6) becomes

<a id=eq7></a>
\begin{align}\label{eq7}\tag{7}
\bar{X}=A+\left(\frac{\sum^K_{j=1}{f_ju_j}}{N}\right)=A + \left(\frac{\sum{fu}}{N}\right)c
\end{align}

which is equivalent to the equation $\bar{X}=A+c\bar{u}$ (see Problem [3.21](../02%20Solved%20Problems/04%20The%20arithmetic%20mean%20computed%20from%20grouped%20data.ipynb#Problem-3-21)). This is called the _coding method_ for computing the mean. It is a very short method and should always be used for grouped data where the class-interval sizes are equal (see Problems [3.22](../02%20Solved%20Problems/04%20The%20arithmetic%20mean%20computed%20from%20grouped%20data.ipynb#Problem-3-22) and [3.23](../02%20Solved%20Problems/04%20The%20arithmetic%20mean%20computed%20from%20grouped%20data.ipynb#Problem-3-23). Note that in the coding method the values of the variable $X$ are _transformed_ into the values of the variable $u$ according to $X=A+cu$.
