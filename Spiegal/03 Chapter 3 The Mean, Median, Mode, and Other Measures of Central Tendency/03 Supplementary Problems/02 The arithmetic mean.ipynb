# The arithmetic mean

## Problem-3-53
A student received grades of $85$, $76$, $93$, $82$, and $96$ in five subjects. Determine the arithmetic mean of the grades.

### Solution-3-53
\begin{align}
\bar{X}=\frac{\sum{X}}{N}=\frac{85+76+93+82+96}{5} = 86.4
\end{align}

```r
p3.53 <- new.env()
p3.53$data <- c(85,76,93,82,96)
round(mean(p3.53$data))
rm(p3.53)
```

## Problem-3-55
A set of numbers consists of six $6$'s, seven $7$'s, eight $8$'s, nine $9$'s and ten $10$'s. What is the arithmetic mean of the numbers?

### Solution-3-55
\begin{align}
\bar{X}=\frac{\sum^K_{j=1}{f_jX_j}}{\sum^K_{j=1}{f_j}}=\frac{(6\cdot6)+(7\cdot7)+(8\cdot8)+(9\cdot9)+(10\cdot10)}{6+7+8+9+10} = 8.25
\end{align}

```r
p3.55 <- new.env()
p3.55$X <- c(6,7,8,9,10)
p3.55$frequency <- c(6,7,8,9,10)
round(weighted.mean(p3.55$X,p3.55$frequency),2)
rm(p3.55)
```

## Problem-3-58
The mean annual salary paid to all employees in a company is $\$36\,000$. The mean annual salaries paid to male and female employees of the company is $\$34\,000$ and $\$40\,000$ respectively. Determine the percentage of males and females employed by the company.

### Solution-3-58

```r
p3.58 <- new.env()
p3.58$mean <- 36000
p3.58$mean.male <- 34000
p3.58$mean.female <- 40000
p3.58$A <- matrix(c(p3.58$mean.male, 1, p3.58$mean.female, 1),2,2)
p3.58$b <- matrix(c(p3.58$mean,1),2,1)
p3.58$pct.male.female <- round(solve(p3.58$A,p3.58$b)*100,2)

cat("Percent Male: ",p3.58$pct.male.female[1],"\n")
cat("Percent Female: ",p3.58$pct.male.female[2], "\n")
rm(p3.58)
```

## Problem-3-59
Table [3.8](#Table-3-8) shows the distribution of the maximum loads in short tons ($1$ short ton = $2\,000$lb) supported by certain cables produced by a company. Determine the mean maximum loading, using (a) the long method and (b) the coding method.

```r
load("../../Saves/Table_3.8.Rdata")
```

#### Table-3-8

```r
Table.3.8
```

### Solution-3-59
The long method is defined in equation [(2)](../01%20Text/04%20The%20arithmetic%20mean.ipynb#eq2):
\begin{align}
\bar{X}=\frac{f_1X_1+f_2X_2+\ldots+f_KX_K}{f_1+f_2+\ldots+f_K}=\frac{\sum^K_{j=1}{f_jX_j}}{\sum^K_{j=1}{f_j}}=\frac{\sum{fX}}{\sum{f}}=\frac{\sum{fX}}{N}
\end{align}

The coding method is defined in equation [(7)](../01%20Text/07%20The%20arithmetic%20mean%20computed%20from%20grouped%20data.ipynb#eq7):
\begin{align}\label{eq7}\tag{7}
\bar{X}=A+\left(\frac{\sum^K_{j=1}{f_ju_j}}{N}\right)=A + \left(\frac{\sum{fu}}{N}\right)c
\end{align}

```r
p3.59a <- new.env()
p3.59a$X <- c(9.3+(9.7-9.3)/2,9.8+(10.2-9.8)/2,10.3+(10.7-10.3)/2,10.8+(11.2-10.8)/2,
             11.3+(11.7-11.3)/2,11.8+(12.2-11.8)/2,12.3+(12.7-12.3)/2,12.8+(13.2-12.8)/2)
p3.59a$f <- Table.3.8$Number.of.Cables
p3.59a$X.bar <- round(sum(p3.59a$f*p3.59a$X)/sum(p3.59a$f),2)
cat("Using the long method: ", p3.59a$X.bar)
rm(p3.59a)
```

```r
p3.59b <- new.env()
p3.59b$X <- c(9.3+(9.7-9.3)/2,9.8+(10.2-9.8)/2,10.3+(10.7-10.3)/2,10.8+(11.2-10.8)/2,
             11.3+(11.7-11.3)/2,11.8+(12.2-11.8)/2,12.3+(12.7-12.3)/2,12.8+(13.2-12.8)/2)
p3.59b$length <- length(p3.59b$X)
p3.59b$A.idx <- ceiling(p3.59b$length/2)
p3.59b$A <- p3.59b$X[p3.59b$A.idx]
p3.59b$u <- seq(1,p3.59b$length,by=1)-p3.59b$A.idx
p3.59b$f <- Table.3.8$Number.of.Cables
p3.59b$fu <- p3.59b$f*p3.59b$u
p3.59b$N <- sum(p3.59b$f)
p3.59b$c <- 9.7-9.3
p3.59b$X.bar <- round(p3.59b$A+(sum(p3.59b$fu)/p3.59b$N)*p3.59b$c,2)
cat("Using the coding method: ",p3.59b$X.bar)
rm(p3.59b)
rm(Table.3.8)
```

## Problem-3-60
Find $\bar{X}$ for the data in Table [(3.9)](#Table-3-9), using (a) the long method and (b) the coding method.

### Solution-3-60
The long method is defined in equation [(2)](../01%20Text/04%20The%20arithmetic%20mean.ipynb#eq2):
\begin{align}
\bar{X}=\frac{f_1X_1+f_2X_2+\ldots+f_KX_K}{f_1+f_2+\ldots+f_K}=\frac{\sum^K_{j=1}{f_jX_j}}{\sum^K_{j=1}{f_j}}=\frac{\sum{fX}}{\sum{f}}=\frac{\sum{fX}}{N}
\end{align}

The coding method is defined in equation [(7)](../01%20Text/07%20The%20arithmetic%20mean%20computed%20from%20grouped%20data.ipynb#eq7):
\begin{align}\label{eq7}\tag{7}
\bar{X}=A+\left(\frac{\sum^K_{j=1}{f_ju_j}}{N}\right)=A + \left(\frac{\sum{fu}}{N}\right)c
\end{align}

```r
p3.60a <- new.env()
p3.60a$X <- Table.3.9["X",]
p3.60a$f <- Table.3.9["f",]
p3.60a$X.bar <- round(sum(p3.60a$X*p3.60a$f)/sum(p3.60a$f),1)
cat("The long method: ", p3.60a$X.bar,"\n")
rm(p3.60a)
```

```r
p3.60b <- new.env()
p3.60b$X <- Table.3.9["X",]
p3.60b$f <- Table.3.9["f",]
p3.60b$length <- length(p3.60b$X)
p3.60b$c <- diff(p3.60b$X)[1]
p3.60b$A.idx <- ceiling(p3.60b$length/2)
p3.60b$A <- p3.60b$X[p3.60b$A.idx]
p3.60b$u <- seq(1,p3.60b$length,by=1)-p3.60b$A.idx
p3.60b$fu <- p3.60b$f*p3.60b$u
p3.60b$X.bar <- round(p3.60b$A+sum(p3.60b$fu)/sum(p3.60b$f)*p3.60b$c,1)
cat("The coded method: ", p3.60b$X.bar,"\n")
rm(p3.60b)
rm(Table.3.9)
```
