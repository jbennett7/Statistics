# The 10-90 percentile range
The _10&ndash;90 percentile range_ of a set of data is defined by

\begin{align}\label{eq1}\tag{1}
\text{10-90 percentile range} = P_{90}-P_{10}
\end{align}

where $P_{10}$ and $P_{90}$ are the 10th and 90th percentiles for the data (see Problem 4.8). The semi-10&ndash;90 percentile range, $\frac{1}{2}(P_{90}-P_{10})$, can also be used but is not commonly employed.
