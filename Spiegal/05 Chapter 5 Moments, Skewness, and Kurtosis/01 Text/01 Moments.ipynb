# Moments
If $X_1, X_2,\ldots,X_N$ are the $N$ values assumed by the variable $X$, we define the quantity  

\begin{align}
\overline{X^r}=\frac{X^r_1+X^r_2+\ldots+X^r_N}{N}
=\frac{\sum^N_{j=1}{X^r_j}}=\frac{\sum{X^r}}{N}\label{eq:rth_moment}\tag{1}
\end{align}

called the $r$th _moment_. The first moment with $r=1$ is the arithmeti mean $\bar{X}$.  
The $r$th _moment about the mean_ $\bar{X}$ is defined as

\begin{align}
m_r=\frac{\sum^N_{j=1}{(X_j-\bar{X})^r}}{N}
=\frac{\sum{(X-\bar{X})}}{N}=\overline{(X-\bar{X})^r}\label{eq:rth_moment_mean}\tag{2}
\end{align}

If $r=1$, then $m_1=0$ (see Problem 3.16). If $r=2$, then $m_2=s^2$, the variance.  
The $r$th _moment about any origin_ $A$ is defined as

\begin{align}
m^\prime_r=\frac{\sum^N_{j=1}{(X_j-A)^r}}{N}
=\frac{\sum{(X-A)^r}}{N}=\frac{\sum{d^r}}{N}=\overline{(X-A)^r}\label{eq:rth_moment_origin}\tag{3}
\end{align}

where $d=X-A$ are the deviations of $X$ from $A$. If $A=0$, equation $\eqref{eq:rth_moment_origin}$ reduces to equation $\eqref{eq:rth_moment}$. FOr this reason, equation $\eqref{eq:rth_moment}$ is often called the $r$th _moment about zero_.
