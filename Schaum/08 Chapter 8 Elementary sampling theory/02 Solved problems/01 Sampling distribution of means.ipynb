# Sampling distribution of means

## Problem 8-1
A population consists of the five numbers 2, 3, 6, 8, and 11. Consider all possible samples of size 2 that can be drawn with replacement from this population. Find (a) the mean of the population, (b) the standard deviation of the population, (c) the mean of the sampling distribution of means, and (d) the standard of the sampling distribution of means (i.e., the standard error of means).

### SOLUTION

```r
X <- c(2,3,6,8,11)
cat("(a) The mean of the population:", mean(X), "\n")
cat("(b) The standard deviation:", sd(X), "\n")
cat("(c) There are 5(5) = 25 samples of size 2 that can be drawn with replacement:\n")
s.means = c()
for (i in X) {
    cat("\t")
    for (j in X) {
        cat("(", i, ",", j, ")\t", sep="")
        s.means = c(s.means, mean(c(i,j)))
    }
    cat("\n")
}
cat("    The mean of sampling distribuion of means is:", mean(s.means), "\n")
cat("(d) The variance of the sampling distribution:", sum((s.means-mean(s.means))^2)/25)
rm(list=ls())
```
