# Combinations
A combination of $n$ different objects taken $r$ at a time is a selection of $r$ out of the $n$ objects, with no attention given to the order of arrangement. The number of combinations of $n$ objects taken $r$ at a time is denoted by the symbol $\binom{n}{r}$ and is given by

\begin{align}\label{eq11}\tag{11}
\binom{n}{r}=\frac{n(n-1)\cdots(n-r+1)}{r!}=\frac{n!}{r!(n-r)!}
\end{align}

__EXAMPLE 13.__ The number of combinations of the letters _a_, _b_, and _c_ taken two at a time is

\begin{align}
\binom{3}{2}=\frac{3\cdot2}{2!}=3
\end{align}

These are _ab_, _ac_, and _bc_. Note that _ab_ is the same combination as _ba_, but not the same permutation.

The number of combinations of $3$ things taken $2$ at at a time is given by the EXCEL command $=\text{COMBIN(3,2)}$ wahich gives the number 3.
