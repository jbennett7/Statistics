## Problem-3-69
Find the median of the maximum loads of the cables in Table 3.8 of Problem [(3.59)](02%20The%20Arithmetic%20Mean.ipynb#Problem-3-59).

### Solution-3-69
Formula for the median of grouped data, obtained by the interpolation equation [(8)](../01%20Text/08%20The%20median.ipynb#eq8):
\begin{align}
\text{Median}=L_1+\left(\frac{\frac{N}{2}-\left(\sum{f}\right)_1}{f_{median}}\right)c
\end{align}  
\begin{align}
\text{where} \ \ \ L_1 &= \text{lower class boundary of the median cliass (i.e., the class containing the median)} \\
N &= \text{number of items in the data (i.e., total frequency)} \\
\left(\sum{f}\right)_1 &= \text{sum of frequencies of all classes lower than the median class} \\
f_{median} &= \text{frequency of the median class} \\
c &= \text{sie of the median class interval}
\end{align}

```r
load("../../Saves/Table_3.8.Rdata")
p3.69 <- new.env()
p3.69$X.lower <- Table.3.8$Max.Load.low
p3.69$X.upper <- Table.3.8$Max.Load.high
p3.69$f <- Table.3.8$Number.of.Cables
p3.69$N <- sum(p3.69$f)
p3.69$idx <- 1
p3.69$sum.f <- 0
for(l in 1:length(p3.69$X.lower)){
    p3.69$sum.f <- p3.69$sum.f+p3.69$f[l]
    if(p3.69$sum.f>p3.69$N/2){
        p3.69$idx <- l
        break
    }
}
p3.69$L1 <- p3.69$X.lower[p3.69$idx]
p3.69$lower.sum.f <- sum(p3.69$f[1:(p3.69$idx-1)])
p3.69$f.median <- p3.69$f[p3.69$idx]
p3.69$c <- p3.69$X.upper[p3.69$idx]-p3.69$X.lower[p3.69$idx]
p3.69$median <- round(p3.69$L1+(((p3.69$N/2)-p3.69$lower.sum.f)/p3.69$f.median)*p3.69$c,2)
cat("The calculated median of the grouped data is:",p3.69$median)

rm(p3.69)
rm(Table.3.8)
```
