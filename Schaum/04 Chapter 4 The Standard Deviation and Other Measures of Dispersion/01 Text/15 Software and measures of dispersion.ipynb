# Software and measures of dispersion
The statisitical software gives a variety of measures for dispersion. The dispersion measures are usually given in descriptive statistics. EXCEL allows for the computation of all the measures discussed in this book. MINITAB and EXCEL are discussed here and outputs of other packages are given in the solved problems.

## Example 3

```r
ex3 <- new.env()
ex3$X <- c(32,  113, 70,  60,  84,
           114, 31,  58,  86,  102,
           113, 79,  86,  24,  40,
           44,  42,  54,  71,  25,
           42,  116, 68,  30,  63,
           121, 74,  77,  77,  100,
           51,  31,  61,  28,  26,
           47,  54,  74,  57,  35,
           77,  80,  125, 105, 61,
           102, 45,  115, 36,  52,
           58,  24,  24,  39,  40,
           95,  99,  54,  35,  31,
           77,  29,  69,  58,  32,
           49,  118, 44,  95,  65,
           71,  65,  74,  122, 99)

cat("The range of the following data set is:", paste(ex3$range <- range(ex3$X),collapse="-"))
cat("\nThe mean deviation is:", ex3$MD <- round(sum(abs(ex3$X-mean(ex3$X)))/length(ex3$X),2))
cat("\nThe semi-interquartile range is:", ex3$IRQ <- round(quantile(ex3$X,.75)-quantile(ex3$X,.25)/2))
cat("\nThe 10-90 percentile range is:", ex3$P1090 <- round(quantile(ex3$X,.90)-quantile(ex3$X,.10),1))
cat("\n\nThe standard deviation for samples is:", ex3$SD <- round(sd(ex3$X),4))
cat("\nThe variance for samples is:", ex3$VAR <- round(var(ex3$X),3))
cat("\n\nThe standard deviation for populations is:",
    {n<-length(ex3$X);ex3$SDp <- round(sqrt((n-1)/n) * sd(ex3$X),4)})
cat("\nThe variance for populations is:", 
    {n<-length(ex3$X);ex3$VAR <- round((n-1)/n * var(ex3$X),3)})
```

```r
d <- density(ex3$X)
d$y <- d$y * length(ex3$X)
plot(d,ylim=c(0,1))
```
