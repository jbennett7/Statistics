__4.33__ Find the range of the sets (a) $5, 3, 8, 4, 7, 6, 12, 4, 3$ and (b) $8.772, 6.453, 10.624, 8.628, 9.434, 6.351$.

```r
p33.a <- c(5, 3, 8, 4, 7, 6, 12, 4, 3)
# In R, the range is given by
(p33.a.R.range <- range(p33.a))

# To get the range as the difference
(p33.a.range <- diff(p33.a.R.range))
```

```r
p33.b <- c(8.772, 6.453, 10.624, 8.628, 9.434, 6.351)
(p33.b.R.range <- range(p33.b))
(p33.b.range <- diff(p33.b.R.range))
```
