# Moments in dimensionless form
To avoid particular units, we can define the _dimensionless moments_ about the mean as

\begin{align}
a_r=\frac{m_r}{s^r}=\frac{m_r}{(\sqrt{m_2})^r}=\frac{m_4}{\sqrt{m_2^r}}
\end{align}

where $s=\sqrt{m_2}$ is the standard deviation. Since $m_1=0$ and $m_2=s^2$, we have $a_1=0$ and $a_2=1$.
