# Charlier's check and sheppard's corrections
Charlier's check in computing moments by the coding method uses the identities:

\begin{align}
\sum{f(u+1)}=\sum{fu}+N
\end{align}

\begin{align}
\sum{f(u+1)^2}=\sum{fu^2}+2\sum{fu}+N
\end{align}

\begin{align}
\sum{f(u+1)^3}=\sum{fu^3}+3\sum{fu^2}+3\sum{fu}+N
\end{align}

\begin{align}
\sum{f(u+1)^4}=\sum{fu^4}+4\sum{fu^3}+6\sum{fu^2}+4\sum{fu}+N
\end{align}

Sheppard's corrections for moments are as follows:

\begin{align}
\text{Corrected }m_2=m_2-\frac{1}{12}c^2\ \ \ \ 
\text{Corrected }m_4=m_4-\frac{1}{2}c^2m_2+\frac{7}{240}c^4
\end{align}

The moments $m_1$ and $m_3$ need no correction.
