# Charlier's check

## Problem 5-8
Illustrate the use of Charlier's check for the computations in Problem [5.7](./02%20Computation%20of%20moments%20from%20grouped%20data.ipynb#Problem-5-7).

### Solution 5-8
To supply the required check, we add to Table [5.3](./02%20Computation%20of%20moments%20from%20grouped%20data.ipynb#Table-5-3) the columns shown in Table [5.4](#Table-5-4) (with the exception of column 2, which is repeated in Table [5.3](./02%20Computation%20of%20moments%20from%20grouped%20data.ipynb#Table-5-3) for convenience).  
In each of the following groupings, the first is taken from Table [5.4](#Table-5-4) and the second is taken from Table [5.2](#Table-5-2). Equality of results in each grouping provides the required check.

```r
load('../../Saves/Table_5.3.Rdata')
p5.8 <- new.env()
p5.8$u.p.1 <- Table.5.3$u+1
p5.8$f <- Table.5.3$f
p5.8$f.u.p.1 <- p5.8$f*p5.8$u.p.1
p5.8$f.u.p.1.2 <- p5.8$f*p5.8$u.p.1^2
p5.8$f.u.p.1.3 <- p5.8$f*p5.8$u.p.1^3
p5.8$f.u.p.1.4 <- p5.8$f*p5.8$u.p.1^4
Table.5.4 <- data.frame(u.p.1=p5.8$u.p.1, f=p5.8$f, f.u.p.1=p5.8$f.u.p.1, f.u.p.1.2=p5.8$f.u.p.1.2, f.u.p.1.3=p5.8$f.u.p.1.3, f.u.p.1.4=p5.8$f.u.p.1.4)
```
