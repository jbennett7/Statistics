# Conditional probability; independent and dependent events

If $E_1$ and $E_2$ are two events, the probability that $E_2$ occurs given that $E_1$ has occurred is denoted by $\text{Pr}\{E_2|E_1\}$, or $\text{Pr}\{E_2 \text{ given } E_2\}, and is called the _conditional probability_ of $E_2$ given that $E_1$ has occurred.

If the occurrence or nonoccurrence of $E_1$ does not effect the probability of occurrence of $E_2$, then $\text{Pr}\{E_2|E_1\}=\text{Pr}\{E_2\}$ and we say that $E_1$ and $E_2$ are _independent events_; otherwise, they are _dependent events_.

If we denote by $E_1E_2$ the event that "both $E_1$ and $E_2$ occur," sometimes called a _compound event_, then

<a id=eq1></a>
\begin{align}\label{eq1}\tag{1}
\text{Pr}\{E_1E_2\}=\text{Pr}\{E_1\}\ \text{Pr}\{E_2|E_1\}
\end{align}

In particular,
<a id=eq2></a>
\begin{align}\label{eq2}\tag{2}
\text{Pr}\{E_1E_2\}=\text{Pr}\{E_1\}\ \text{Pr}\{E_2\}\ \ \ \ \ \ \text{ for independent events}
\end{align}

For three events $E_1, E_2,$ and $E_3$, we have
<a id=eq3></a>
\begin{align}\label{eq3}\tag{3}
\text{Pr}\{E_1E_2E_3\}=\text{Pr}\{E_1\}\ \text{Pr}\{E_2|E_1\}\ \text{Pr}\{E_3|E_1E_2\}
\end{align}

That is, the probability of occurrence of $E_1, E_2,$ and $E_3$ is equal to $($the probability of $E_1)\times($the probability of $E_2$ given that $E_1$ has occurred$)\times($the probability of $E_3$ given that both $E_1$ and $E_2$  have occurred$)$.

<a id=eq4></a>
\begin{align}\label{eq4}\tag{4}
\text{Pr}\{E_1E_2E_3\}=\text{Pr}\{E_1\}\ \text{Pr}\{E_2\}\ \text{Pr}\{E_3\}\ \ \ \ \ \text{ for independent events}
\end{align}

In general, if $E_1, E_2, E_3,\ldots, E_n$ are $n$ independent events having respective probabilities $p_1, p_2, p_3,\ldots,p_n$, then the probability of occurrence of $E_1$ and $E_2$ and $E_3$ and $\ldots E_n$ is $p_1p_2p_3 \ldots p_n$.

### EXAMPLE 3.
Let $E_1$ and $E_2$ be the events "heads on fifth toss" and "heads on sixth toss" of a coin, respectively. Then $E_1$ and $E_2$ are independent events, and thus the probability of heads on both the fifth and sixth tosses is (assuming the coin to be fair)

\begin{align}
\text{Pr}\{E_1E_2\}=\text{Pr}\{E_1\}\ \text{Pr}\{E_2\}=\left(\frac{1}{2}\right)\left(\frac{1}{2}\right)=\frac{1}{4}
\end{align}

### EXAMPLE 4.
If the probability that $A$ will be alive in $20$ years is $0.7$ and the probability that $B$ will be alive in $20$ years is $0.5$, then the probability that they will both be alive in $20$ years is $(0.7)(0.5)=0.35$.

### EXAMPLE 5.
Suppose that a box contains $3$ white balls and $2$ black balls. Let $E_1$ be the event "first ball drawn is black" and $E_2$ the event "second ball drawn is black," where the balls are not replaced after being drawn. Here $E_1$ and $E_2$ are dependent events.

The probability that the first ball drawn is black is \text{Pr}\{E_1\}=2/(3+2)=\frac{2}{5}$. The probability that the second ball is drawn is black, given that the first ball drawn was black, is $\text{Pr}\{E_2|E_1\}=1/(3+1)=\frac{1}{4}$. Thus the probability that both balls drawn are black is

\begin{align}
\text{Pr}\{E_1E_2\}=\text{Pr}\{E_1\}\ \text{Pr}\{E_2|E_1\}=\frac{2}{5}\cdot\frac{1}{4}=\frac{1}{10}
\end{align}
