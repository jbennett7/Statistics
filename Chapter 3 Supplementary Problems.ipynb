## THE ARITHMETIC MEAN

53. A student received grades of $85, 76, 93, 82,$ and $96$ in the five  subjects. Determine the arithmetic mean of the grades.

```r
p53.s1 <- c(85, 76, 93, 82, 96)
round(mean(p53.s1))
```

```r
p54.s1 <- c(0.53, 0.46, 0.50, 0.49, 0.52, 0.53, 0.44, 0.55)
round(mean(p54.s1),2)
```

55. A set of numbers consists of six 6's, seven 7's, eight 8's, nine 9's and ten 10's. What is the arithmetic mean of the numbers?

```r
p55.s1 <- c(6,7,8,9,10)
weighted.mean(p55.s1,p55.s1)
```

56. A student's grades in the laboratory, lecture, and recitation parts of a physics course were $71, 78,$ and $89$, respectively.<p><p>
  a. If the weights accorded these grades are $2, 4,$ and $5$, respectively, what is an appropriate average grade?<p>
  b. What is the average grade if equal weights are used?

```r
p56.s1 <- c(71, 78, 89)
p56.w1 <- c(2, 4, 5)
round(weighted.mean(p56.s1, p56.w1))
round(mean(p56.s1))
```

57. Three teachers of economics reported mean examination grades of $79, 74,$ and $82$ in their classes, which consisted of $32, 25,$ and $17$ students, respectively. Determine the mean grade for all the classes.

```r
p57.grades <- c(79, 74, 82)
p57.students <- c(32, 25, 17)
round(weighted.mean(p57.grades, p57.students))
```

59. The distribution table shows the maximum loads in short tons ($1$ short ton = $2000$lb) supported by certain cables produced by a company. Determine the mean maximum loading, using (a) the long method and (b) the coding method.

```r
p59.X <- c(9.5, 10.0, 10.5, 11, 11.5, 12, 12.5, 13)
p59.f <- matrix(c(2, 5, 12, 17, 14, 6, 3, 1),ncol=8)
colnames(p59.f) <- c('[9.3-9.7]', '[9.8-10.2]', '[10.3-10.7]', '[10.8-11.2]', '[11.3-11.7]', '[11.8-12.2]', '[12.3-12.7]', '[12.8-13.2]')
p59.A <- 11
p59.u <- c(-3,   -2,    -1,  0,   1,   2,   3,   4)
p59.c <- diff(p59.X)[1]
(X.bar.long <- round(sum(p59.f*p59.X/sum(p59.f)),2))
(X.bar.coded <- p59.A + round(sum(p59.f*p59.u)/sum(p59.f)*p59.c,2))
#(X.bar.short <- p59.A + round(sum(p59.f*(p59.X - p59.A))/sum(p59.f),2))
```
