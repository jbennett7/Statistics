68. Find the median of the set of numbers in 55

```r
weighted.median <- function(X, f){
    f.sum <- sum(f)
    N <- length(f)
    c <- diff(X)[1]
    sf <- 0
    nfi <- 0
    for(i in 1:N){
        if(sf + f[i] > f.sum/2){
            nfi <- i
            break
        }
        sf <- sf + f[i]
    }
    missing <- f.sum/2 - sf
    nf <- f[nfi]
    lower <- X[i] - c/2
    upper <- X[i] + c/2
    return(lower + (missing/nf)*(upper-lower))
}
```

```r
p55.f <- c(6,7,8,9,10)
p55.X <- c(6,7,8,9,10)
round(weighted.median(p55.X, p55.f))
```
