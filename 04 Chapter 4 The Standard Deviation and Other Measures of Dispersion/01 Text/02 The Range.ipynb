# The Range
The _range_ of a set of numbers is the difference between the largest and smallest numbers in the set.

__EXAMPLE 1.__  The range of the set $2, 3, 3, 5, 5, 5, 8, 10, 12$ is $12-2=10$. Sometimes the range is given by simply quoting the smallest and largest numbers; in the above set, for instance, the range could be indicated as $2$ to $12$, or $2$&mdash;$12$.
