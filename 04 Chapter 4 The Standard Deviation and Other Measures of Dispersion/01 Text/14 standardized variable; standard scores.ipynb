# Standardized variable; standard scores
The variable that measures the deviation from the mean in units of the standard deviation is called a _standardized variable_, is a dimensionless quantity (i.e., is independent of the units used), and is given by

<a id=eq16></a>
\begin{align}\label{eq16}\tag{16}
z=\frac{X-\bar{X}}{s}
\end{align}

If the deviations from the mean are given in units of the standard deviation, they are said to be expressed in _standard units_, or _standard scores_. These are of great value in the compoarison of distributions (see Problem 4.31).
