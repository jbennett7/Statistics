# Relative Cumulative-Frequency Distributions and Percentage Ogives
The _relative cumulative frequency_, or _percentage cumulative frequency_, is the cumulative frequency divided by the total frequency. For example, the relative cumulative frequency of heights less than $68.5$in is $65/200=65\%$, signifying that $65\%$ of the students have heights less than $68.5$in. If the relative cumulative frequencies are used in Table 2.3, in place of cumulative frequencies, the results are called _relative cumulative-frequency distributions_ (or _percentage cumulative distributions_) and _relative cumulative-frequency polygons_ (or _percentage ogives_), respectively.
