# Fundamental rules of probability

## Problem-6-1
Determine the probability $p$, or an estimate of it, for each of the following events:  
(a) An odd number appears in a single toss of a fair die.  
(b) At least one head appears in two tosses of a fair coin.  
(c) An Ace, 10 of diamonds, or 2 of spades appears in drawing a single card from a well-shuffled ordinary deck of 52 cards.  
(d) The sum $7$ appears in a single toss of a pair of fair dice.  
(e) A tail appears in the next toss of a coin if out of $100$ tosses $56$ were heads.

### Solution-6-1
(a) Out of six possible equally likely cases, three cases (where the die comes up $1, 3,$ or $5$) are favorable to the event. Thus $p=\frac{3}{6}=\frac{1}{2}$.  

(b) If $H$ denotes "head" and $T$ denotes "tail," the two tosses can lead to four cases: HH, HT, TH, and TT, all equally likely. Only the first three cases are favorable to the event. Thus $p=\frac{3}{4}$.  

(c) The event can occur in six ways (ace of spades, ace of hearts, ace of clubs, ace of diamonds, 10 of diamonds, and 2 of spades) out of 52 equally likely cases. Thus $p=\frac{6}{52}=\frac{3}{26}$.  

(d) Each of the six faces of one die can be associated with each of the six faces of the other die, so that the total number of cases that can arise, all equally likely, is $6\cdot6=36$. These can be denoted by $(1,1), (2,1), (3,1),\ldots,(6,6)$. There are six ways of obtaining the sum $7$, denoted by $(1,6),(2,5),(3,4),(4,3),(5,2),$ and $(6,1)$. Thus $p=\frac{6}{36}=\frac{1}{6}$.  

(e) Since $100-56=44$ tails were obtained in $100$ tosses, the _estimated_ (or _empirical_) _probability_ of a tail is the relative frequency $44/100=0.44$.

## Problem-6-2
An experiment consists of tossing a coin and a die. If $E_1$ is the event that "head" comes up in tossing the coin and $E_2$ is the event that "$3$ or $6$" comes up in tossing the die, state in words the meaning of each of the following:  

(a)&ensp;$\bar{E_1}$&emsp;&emsp;&emsp;(c)&ensp;$E_1E_2$&emsp;&emsp;&emsp;(e)&ensp;$\text{Pr}\{E_1|E_2\}$  
(b)&ensp;$\bar{E_2}$&emsp;&emsp;&emsp;(d)&ensp;$\text{Pr}\{E_1\bar{E_2}\}$&nbsp;&ensp;(f)&ensp;$\text{Pr}\{E_1+\bar{E_2}\}$

### Solution-6-2
(a) Tails on the coin and anything on the die.  
(b) $1, 2, 4,$ or $5$ on the die, anything on the coin.  
(c) Heads on the coin and $3$ or $6$ on the die.  
(d) Probability of heads on the coin and $1, 2, 4,$ and $5$ on the die.

## Problem-6-3
A ball is drawn at random from a box containing $6$ red balls, $4$ white balls, and $5$ blue balls. Determine the probability that the ball drawn is (a) red, (b) white, (c) blue, (d) not red, and (e) red or white.

### Solution-6-3
Let $R, W,$ and $B$ denote the events of drawing a red ball, white ball, and blue ball, respectively. Then:  

(a)&emsp;&emsp;&emsp;&emsp;$\text{Pr}\{R\}=\frac{\text{ways of choosing a red ball}}{\text{total ways of choosing a ball}} = \frac{6}{6+4+5}=\frac{6}{15}=\frac{2}{5}$  

(b)&emsp;&emsp;&emsp;&emsp;$\text{Pr}\{W\}=\frac{4}{6+4+5}=\frac{4}{15}$  

(c)&emsp;&emsp;&emsp;&emsp;$\text{Pr}\{B\}=\frac{5}{6+4+5}=\frac{1}{3}$  

(d)&emsp;&emsp;&emsp;&emsp;$\text{Pr}\{\bar{R}\}=1-\text{Pr}\{R\}=1-\frac{2}{5}-\frac{3}{5}$&emsp;by part (a)  

(e)&emsp;&emsp;&emsp;&emsp;$\text{Pr}\{R+W\}=\frac{\text{ways of choosing a red or white ball}}{\text{total ways of choosing a ball}}=\frac{6+4}{6+4+5}=\frac{10}{15}=\frac{2}{3}$
##### Another method
\begin{align}
\text{Pr}\{R+W\} = \text{Pr}\{\bar{B}\}=1-\text{Pr}\{B\}=1-\frac{1}{3}=\frac{2}{3}\ \ \ \ \text{ by part (c)}
\end{align}

Note that $\text{Pr}\{R+W\}=\text{Pr}\{R\}+\text{Pr}\{W\}$ (i.e., $\frac{2}{3}=\frac{2}{5}+\frac{4}{15}$). This is an illustration of the general rule $\text{Pr}\{E_1+E_2\}=\text{Pr}\{E_1\}+\text{Pr}\{E_2\}$ that is true for _mutually exclusive_ events $E_1$ and $E_2$.

## Problem-6-4
A fair die is tossed twice. Find the probability of getting a $4, 5,$ or $6$ on the first toss and a $1, 2, 3,$ or $4$ on the second toss.

### Solution-6-4
Let $E_1=$ event "$4, 5,$ or $6$" on the first toss, and let $E_2=$ event "$1, 2, 3$ or $4$" on the second toss. Each of the six ways in which the die can fall on the first toss can be associated with each of the six ways in which it can fall on the second toss, a total of $6\cdot6=36$ ways, all equally likely. Each of the three ways in which $E_1$ can occur can be associated with each of the four ways in which $E_2$ can occur, to give $3\cdot4=12$ ways in which both $E_1$ and $E_2$, or $E_1E_2$ occur. Thus $\text{Pr}\{E_1E_2\}=12/36=1/3$.  

Note that $\text{Pr}\{E_1E_2\}=\text{Pr}\{E_1\}\text{Pr}\{E_2\}$ (i.e., $\frac{1}{3}=\frac{3}{6}\cdot\frac{4}{6}$) is valid for the _independent events_ $E_1$ and $E_2$.

## Problem-6-5
Two cards are drawn from a well-shuffled ordinary deck of 52 cards. Find the probability that they are both aces if the first card is (a) replaced and (b) not replaced.

### Solution-6-5
Let $E_1=$event "ace" on the first draw, and let $E_2=$ event "ace" on the second draw.  

(a)&ensp;If the first card is replaced, $E_1$ and $E_2$ are independent events. Thus $\text{Pr}\{$both cards drawn are aces$\}=\text{Pr}\{E_1E_2\}=\text{Pr}\{E_1\}\text{Pr}\{E_2\}=(\frac{4}{52})(\frac{4}{52})=\frac{1}{169}$.

(b)&ensp;The first card can be drawn in any one of $52$ ways, and the second card can be drawn in any one of $51$ ways since the first card is not replaced. Thus both cards can be drawn in $52\cdot51$ ways, all equally liekly.

There are four ways in which $E_1$ can occur and three ways in which $E_2$ can occur, so that both $E_1$ and $E_2$, or $E_1E_2$, can occur in $4\cdot3$ ways. Thus $\text{Pr}\{E_1E_2\}=(4\cdot3)/(52\cdot51)=\frac{1}{221}$.

Note that $\text{Pr}\{E_2|E_1\}=\text{Pr}\{$second card is an ace given that first card is an ace$\}=\frac{3}{51}$. Thus our result is an illustration of the general rule that $\text{Pr}\{E_1E_2\}=\text{Pr}\{E_1\}\text{Pr}\{E_2|E_1\}$ when $E_1$ and $E_2$ are dependent events.

## Problem-6-6
Three balls are drawn successively from the box of Problem [6.3](#Problem-6-3). FInd the probability that they are drawn in the order red, white, and blue if each ball is (a) replaced and (b) not replaced.

### Solution-6-6
Let $R=$ event "red" on the first draw, $W=$ event "white" on the second draw, and $B=$ event "blue" on the third draw. We require $\text{Pr}\{RWB\}$.

(a) If each ball is replaced, then $R, W,$ and $B$ are independent events and

\begin{align}
\text{Pr}\{RWB\}=\text{Pr}\{R\}\ \text{Pr}\{W\}\ \text{Pr}\{B\}
=\left(\frac{6}{6+4+5}\right)\left(\frac{4}{6+4+5}\right)\left(\frac{5}{6+4+5}\right)
=\left(\frac{6}{15}\right)\left(\frac{4}{15}\right)\left(\frac{5}{15}\right)=\frac{8}{225}
\end{align}

(b) If each ball is not replaced, the $R, W,$ and $B$ are dependent events and

\begin{align}
\text{Pr}\{RWB\}=\text{Pr}\{R\}\ \text{Pr}\{W|R\}\ \text{Pr}\{B|WR\}
=\left(\frac{6}{6+4+5}\right)\left(\frac{4}{5+4+5}\right)\left(\frac{5}{5+3+5}\right)
=\left(\frac{6}{15}\right)\left(\frac{4}{14}\right)\left(\frac{5}{13}\right)=\frac{4}{91}
\end{align}

## Problem-6-7
Find the probability of a $4$ turning up at least once in two tosses of a fair die.

### Solution-6-7
Let $E_1 = \text{event}$ "$4$" on the first toss, $E_2 = \text{event}$ "$4$" on the second toss, and $E_1 + E_2 = \text{event}$ "$4$" on the first toss of "$4$" on the second toss or $\text{both} = \text{event}$ that at least one $4$ turns up. We require $\text{Pr}\{E_1 + E_2\}$.
#### First method
The total number of equally likely ways in which both dice can fall is $6 \cdot 6 = 36$. Also,
$$
\begin{align*}
\text{Number of ways in which } E_1 \text{ occurs but not } E_2 &= 5 \\
\text{Number of ways in which } E_2 \text{ occurs but not } E_1 &= 5 \\
\text{Number of ways in which both } E_1 \text{ and } E_2 \text{ occur } &= 1
\end{align*}
$$
Thus the number of ways in which at least one of the events $E_1$ or $E_2$ occurs is $5 + 5 + 1 = 11$, and thus $\text{Pr}\{E_1 + E_2\} = \frac{11}{36}$.  
#### Second Method
Since $E_1$ and $E_2$ are not mutually exclusive, $\text{Pr}\{E_1 + E_2\}=\text{Pr}\{E_1\} + \text{Pr}\{E_2\} - \text{Pr}\{E_1E_2\}$. Also, since $E_1$ and $E_2$ are independent, $\text{Pr}\{E_1E_2\}=\text{Pr}\{E_1\}\text{Pr}\{E_2\}$. Thus $\text{Pr}\{E_1 + E_2\}=\text{Pr}\{E_1\} + \text{Pr}\{E_2\}-\text{Pr}\{E_1\}\text{Pr}\{E_2\} = \frac{1}{6} + \frac{1}{6} - (\frac{1}{6})(\frac{1}{6}) = \frac{11}{36}$.
#### Third method

$$
\text{Pr}\{\text{at least one }4\text{ comes up }\} + \text{Pr}\{\text{no } 4 \text{ comes up}\} = 1
$$
Thus
$$
\begin{align*}
\text{Pr}\{\text{at least one} \}4\text{ comes up}\} &= 1 - \text{Pr}\{\text{ no }4\text{ comes up}\} \\
                                              &= 1 - \text{Pr}\{\text{ no }4
                                                     \text{ on first toss and no }4 \text{ on second toss}\} \\
                                              &= 1 - \text{Pr}\{\bar{E_1}\bar{E_2}\} = 1 -
                                                     \text{Pr}\{\bar{E_1}\} \text{Pr}\{\bar{E_2}\} \\
                                              &= 1 - \left(\frac{5}{6}\right)\left(\frac{5}{6}\right) 
                                                      = \frac{11}{36}
\end{align*}
$$

## Problem 6-8
One bag contains 4 white balls and 2 black balls; another contains 3 white balls and 5 black balls. If one ball is drawn from each bag, find the probability that (a) both are white, (b) both are black, and (c) one is white and one is black.

### Solution
Let $W_1 = $ event "white" ball from the first bag, and let $W_2 =$ event "white" ball from second bag.  
(a)
\begin{align}
\text{Pr}\{W_1 W_2\} = \text{Pr}\{W_1\}\text{Pr}\{W_2\} = \left(\frac{4}{4+2}\right)\left(\frac{3}{3+5}\right) = \frac{1}{4}
\end{align}

(b)
\begin{align}
\text{Pr}\{\bar{W_1}\bar{W_2}\} = \text{Pr}\{\bar{W_1}\}\text{Pr}\{\bar{W_2}\} = \left(\frac{2}{4+2}\right)\left(\frac{5}{3+5}\right) = \frac{5}{24}
\end{align}

(c)  
The event "one is white and one is black" is the same as the event "either the first is white and the second is black _or_ the first is black and the second is white"; that is, $W_1 \bar{W_2} + \bar{W_1}W_2$. Since events $W_1 \bar{W_2}$ and $\bar{W_1} W_2$ are mutually exclusive, we have
\begin{align}
\text{Pr}\{W_1 \bar{W_2} + \bar{W_1} W_2\} &= \text{Pr}\{W_1 \bar{W_2}\} + \text{Pr}\{\bar{W_1} W_2\} \\
                                           &= \text{Pr}\{W_1\}\text{Pr}\{\bar{W_2}\} + \text{Pr}\{\bar{W_1}\}\text{Pr}\{W_2\} \\
                                           &= \left(\frac{4}{4+2}\right)\left(\frac{5}{3+5}\right) + \left(\frac{2}{4+2}\right)\left(\frac{3}{3+5}\right) = \frac{13}{24}
\end{align}

#### Another method
The required probability is $1 - \text{Pr}\{W_1W_2\} - \text{Pr}\{\bar{W_1}\bar{W_2}\} = 1 - \frac{1}{4} - \frac{5}{24} = \frac{13}{24}$.

## Problem 6-9
$A$ and $B$ play 12 games of chess, of which 6 are won by $A$, 4 are won by $B$, and 2 end in a draw. They agree to play a match consisting of 3 games. Find the probability that (a) $A$ wins all 3 games, (b) 2 games end in a draw, (c) $A$ and $B$ win alternatiely, and (d) $B$ wins at least 1 game.

### Solution
Let $A_1,A_2,$ and $A_3$ denote the events "A wins" in the first, second, and third games, respectively; let $B_1,B_2,$ and $B_3$ denote the events "B wins" in the first, second, and third games, respectively; and let, $D_1, D_2,$ and $D_3$ denote the events "there is a draw" in the first, second, and third games, respectively.  
On the basis of their past experience (empirical probability), we shall assume that $\text{Pr}\{A\text{ wins any one game}\} = \frac{6}{12} = \frac{1}{2}$, that $\text{Pr}\{B\text{ wins any one game}\} = \frac{4}{12} = \frac{1}{3}$, and that $\text{Pr}\{\text{any one game ends in a draw}\} = \frac{2}{12} = \frac{1}{6}$.

(a) $\text{Pr}\{A\text{ wins all 3 games}\} = \text{Pr}\{A_1A_2A_3\} = \text{Pr}\{A_1\}\text{Pr}\{A_2\}\text{Pr}\{A_3\} =
                                              \left(\frac{1}{2}\right) \left(\frac{1}{2}\right) \left(\frac{1}{2}\right) = \frac{1}{8}$

assuming that the results of each game are independent of the results of any others, which appears to be justifiable (unless, of course, the players happen to be _psychologically influenced_ by the other one's winning or losing).

(b)
\begin{align}
\text{Pr}\{2\text{ games end in a draw}\} &= \text{Pr}\{\text{1st and 2nd or 1st and 3rd or 2nd and 3rd games end in a draw}\} \\
                                          &= \text{Pr}\{D_1D_2D_3\} + \text{Pr}\{D_1\bar{D_2}D_3\} + \text{Pr}\{\bar{D_1}D_2D_3\} \\
                                          &= \text{Pr}\{D_1\}\text{Pr}\{D_2\}\text{Pr}\{\bar{D_3}\} + \text{Pr}\{D_1\}\text{Pr}\{\bar{D_2}\}\text{Pr}\{D_3\} \\
                                          &\ \ +\text{Pr}\{\bar{D_1}\}\text{Pr}\{D_2\}\text{Pr}\{D_3\} \\
                                          &= \left(\frac{1}{6}\right)\left(\frac{1}{6}\right)\left(\frac{5}{6}\right) + 
                                             \left(\frac{1}{6}\right)\left(\frac{5}{6}\right)\left(\frac{1}{6}\right) +
                                             \left(\frac{5}{6}\right)\left(\frac{1}{6}\right)\left(\frac{1}{6}\right) = \frac{15}{216} = \frac{5}{72}
\end{align}

(c)
\begin{align}
\text{Pr}\{A \text{ and } B \text{ win alternately}\} &= \text{Pr}\{\text{ A wins then B wins then A wins or B wins then A wins then B wins}\} \\
                                                      &= \text{Pr}\{A_1B_2A_3 + B_1A_2B_3\} = \text{Pr}\{A_1B_2A_3 + B_1A_2B_3\} \\
                                                      &= \text{Pr}\{A_1\}\text{Pr}\{B_2\}\text{Pr}\{A_3\} + \text{Pr}\{B_1\}\text{Pr}\{A_2\}\text{Pr}\{B_3\} \\
                                                      &= \left(\frac{1}{2}\right)\left(\frac{1}{3}\right)\left(\frac{1}{2}\right) +
                                                          \left(\frac{1}{3}\right)\left(\frac{1}{2}\right)\left(\frac{1}{3}\right) = \frac{5}{36}
\end{align}

(d)
\begin{align}
\text{Pr}\{\text{B wins at least 1 game}\} &= 1 - \text{Pr}\{\text{B wins no game}\} \\
                                           &= 1 - \text{Pr}\{\bar{B_1}\bar{B_2}\bar{B_3}\} = 1 - \text{Pr}\{B_1\}\text{Pr}\{A_2\}\text{Pr}\{B_3\} \\
                                           &= 1 - \left(\frac{2}{3}\right)\left(\frac{2}{3}\right)\left(\frac{2}{3}\right) = \frac{19}{27}
\end{align}
