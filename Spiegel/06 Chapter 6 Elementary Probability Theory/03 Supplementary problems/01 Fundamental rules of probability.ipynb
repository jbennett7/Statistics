## Problem-6-40
Determine the probability $p$, or an estimate of it, for each of the following events.

&emsp;&emsp;(a)&ensp;A king, ace, jack of clubs, or queen of diamonds appears in drawing a single card from a well-shuffled ordinary deck of cards.  
&emsp;&emsp;(b)&ensp;The sum $8$ appears in a single toss of a pair of fair dice.  
&emsp;&emsp;(c)&ensp;A nondefective bolt will be found if out of $600$ bolts already examined, 12 were defective.  
&emsp;&emsp;(d)&ensp;A $7$ or $11$ comes up in a single toss of a pair of fair dice.  
&emsp;&emsp;(e)&ensp;At least one head appears in three tosses of a fair coin.

### Solution-6-40
&emsp;&emsp;(a)&ensp;There are 4 kings, 4 aces, 1 jack of clubs and 1 queen of diamonds for a total of 10 cards. $p=\frac{10}{52}=\frac{5}{26}$.  
&emsp;&emsp;(b)&ensp;For a pair of dice, there are five different combinations to get $8$: (2,6), (3,5), (4,4), (5,3), (2,6). There are a total of $6\times6=36$ combinations. Therefore, $p=\frac{5}{36}$.  
&emsp;&emsp;(c)&ensp;Using the previous $600$ observations, we can conclude that the probability of finding a defective bolt is $p=\frac{12}{600}=\frac{1}{50}$.  
&emsp;&emsp;(d)&ensp;For a pair of dice, there are six different combinations to get $7$: (1,6), (2,5), (3,4), (4,3), (5,2), (6,1);
and there are two different combinations to get $11$: (5,6), (6,5).  
&emsp;&emsp;&emsp;&ensp;With $36$ total combinations, the probability is $p=\frac{8}{36}=\frac{2}{9}$.  
&emsp;&emsp;(e)&ensp;For three tosses of a fair coin, there are $8$ different combinations:
(H,H,H), (H,H,T), (H,T,H), (H,T,T), (T,H,H), (T,H,T), (T,T,H), (T,T,T).  
&emsp;&emsp;&emsp;&ensp;There is only one combination where a head does not appear, therefore the probability is $p=\frac{7}{8}$.

## Problem-6-42
A ball is drawn at random from a box containing $10$ red, $30$ white, $20$ blue, and $15$ orange marbles. Find the probability that the ball drawn is (a) orange or red, (b) not red or blue, (c) not blue, (d) white, and (e) red, white, or blue.

### Solution-6-42
There are a total of $10+30+20+15=75$ balls.  Event red = $R$; event white $W$; event blue $B$; and event orange $O$.  
&emsp;(a)&ensp;The probability of "orange" or "red" is:
$$
\text{Pr}\{O+R\}=\frac{15}{75}+\frac{10}{75}=\frac{25}{75}=\frac{1}{3}\text{.}
$$
&emsp;(b)&ensp;The probability of not red or blue:  
&emsp;&emsp;The probability of not red or blue is the same as the sum of all the other colors:
$$
\text{Pr}\{\bar{R}+\bar{B}\}=\text{Pr}\{W+O\}=\frac{45}{75}=\frac{3}{5}\text{.}
$$
&emsp;(c)&ensp;The probability of "not blue":
$$
\text{Pr}\{\bar{B}\}=\frac{75-20}{75}=\frac{11}{15}\text{.}
$$
&emsp;(d)&ensp;The probability of "white":
$$
\text{Pr}\{W\}=\frac{30}{75}=\frac{2}{5}\text{.}
$$
&emsp;(e)&ensp;The probability of "red", "white", or "blue":
$$
\text{Pr}\{R+W+B\}=\frac{10}{75}+\frac{30}{75}+\frac{20}{75}=\frac{60}{75}=\frac{4}{5}\text{.}
$$

## Problem-6-45
Find the probability of scoring a total of $7$ points (a) once, (b) at least once, and (c) twice in two tosses of a pair of fair dice.

### Solution-6-45
For one toss of a pair of fair dice the chances of getting $7$ is $\frac{6}{36}=\frac{1}{6}$ which is defined as event $E$.  
&emsp;(a)&ensp;The probability of scoring $7$ on either toss of the dice  
&emsp;&emsp;States:&ensp;(1)&ensp;$\bar{E}_1\bar{E}_2$&ensp;(2)&ensp;$\bar{E}_1E_2$&ensp;(3)&ensp;$E_1\bar{E}_2$&ensp;(4)&ensp;$E_1E_2$  
&emsp;&emsp;We are interested in (2) and (3).
$$
\text{Pr}\{\bar{E}_1E_2\}+\text{Pr}\{E_1\bar{E}_2\}=\frac{5}{6}\times\frac{1}{6}+\frac{1}{6}\times\frac{5}{6}=\frac{10}{36}=\frac{5}{18}
$$
&emsp;(b)&ensp;The probability of scoring $7$ at least once is equal to the complement of not scoring $7$ on either toss.  
$$
\text{Pr}\{\bar{E}_1\bar{E}_2\} = \frac{5}{6}\times\frac{5}{6} = \frac{25}{36}
$$
&emsp;&emsp;Therefore the probability of scoring $7$ at least once is
$$
\text{Pr}\{\text{7 at least once}\}=\frac{36-25}{36} = \frac{11}{36}
$$
&emsp;(c)&ensp;The probability of scoring $7$ both times is
$$
\text{Pr}\{E_1E_2\}=\frac{1}{6}\times\frac{1}{6}=\frac{1}{36}
$$

## Problem-6-46
Two cards are drawn successively from an ordinary deck of 52 well-shuffled cards. Find the probability that (a) the first card is not a $10$ of clubs or an ace, (b) the first card is an ace but the second is not, (c) at least one card is a diamond, (d) the cards are not of the same suit, (e) not more than one card is a picture card given that the first was a picture card, and (h) the cards are picture cards or spades or both.

### Solution-6-46
$E_1$ is the first card drawn; $E_2$ is the second card drawn.  
&emsp;(a)&ensp; The probability that $E_1=\text{is not a 10 of clubs or an ace}$. There are four aces and one 10 of clubs. This means that the complement $\bar{E}_1$ is $\frac{5}{52}$.
$$
\text{Pr}\{E_1\text{ is not a 10 of clubs or an ace}\} = \frac{52 - 5}{52} = \frac{47}{52}\text{.}
$$
&emsp;(b)&ensp;The probability that the first card is an ace but the second is not. The probability that the first card is an ace is $\frac{4}{52}$. The probability that the second card is not an ace, given that the first card is an ace is
$$
\text{Pr}\{E_2|E_1\}=\frac{4}{52}\times\frac{48}{51}=\frac{192}{2652}=\frac{16}{221}\text{.}
$$
&emsp;(c)&ensp;The probability that at least one card is a diamond. The complement is that none of the two draws are diamonds.  
$$
\text{Pr}\{\bar{E}_1\bar{E}_2\}=\frac{52-13}{52}\times\frac{51-13}{51}=\frac{39}{52}\times\frac{38}{51}=\frac{1482}{2652}=\frac{57}{102}\text{.}
$$
&emsp;&emsp;Therefore, the probability that at least one card is a diamond is
$$
\text{Pr}\{E_1E_2\}=\frac{102-57}{102} = \frac{45}{102} = \frac{15}{34}\text{.}
$$
&emsp;(d)&ensp;The probabilities of the two drawn cards are not in the same suit.  There are four suits. Each suit has $\frac{1}{4}$ chance on the first.
On the second the same suit has the probability of $\frac{12}{51}$ being drawn. Therefore the probability that the second card is not the same suit as the first is
$$
\text{Pr}\{\text{the second card is not the same suit as the first}\}=\frac{51-12}{51}=\frac{13}{17}\text{.}
$$
&emsp;(e)&ensp;The probability that not more than one drawn card is a picture card (king, queen, or jack). The probability of drawing a picture card on both draws is
$$
\text{Pr}\{\text{both draws are picture cards}\}=\frac{12}{52}\times\frac{11}{51}=\frac{132}{2652}=\frac{11}{221}\text{.}
$$
&emsp;&emsp;Therefore, the probability that no more than one picture card is drawn is the complement.
$$
\text{Pr}\{\overline{\text{both draws are picture cards}}\} = \frac{221-11}{221} = \frac{210}{221}\text{.}
$$
&emsp;(f)&ensp;The second card is not a picture card
$$
\text{Pr}\{E_2\}=\frac{52-12}{52}=\frac{40}{52}=\frac{10}{13}\text{.}
$$
&emsp;(g)&ensp;The second card is not a picture card given that the first is a picture card. The probability that the first card is a picture card is
$$
\text{Pr}\{E_1\}=\frac{12}{52}=\frac{3}{4}\text{.}
$$
&emsp;&emsp;Therefore, the probability that the next card is a picture card, given that the first is a picture card is
$$
\text{Pr}\{E_2|E_1\}=\frac{51-11}{51}=\frac{40}{51}\text{.}
$$
&emsp;(h)&ensp;The cards are picture cards, spades, or both.  
&emsp;&emsp;The probability that the two cards are picture cards is
$$
\text{Pr}\{E_1E_2\}=\frac{12}{52}\times\frac{11}{51}=\frac{132}{2652}=\frac{22}{442}\text{.}
$$
&emsp;&emsp;The probability that they are spades is
$$
\text{Pr}\{E_1E_2\}=\frac{13}{52}\times\frac{12}{51}=\frac{156}{2652}=\frac{26}{442}.
$$
&emsp;&emsp;The probability that they are both.
$$
\text{Pr}\{E_1E_2\}=\left(\frac{12}{52}\times\frac{11}{51}+\frac{13}{52}\times\frac{12}{51}\right)\times\frac{3}{52}
$$

## Problem-6-48
The odds in favor of $A$ winning a game of chess against $B$ are $3:2$. If three games are to be played, what are the odds (a) in favor of $A$'s winning at least two games out of the three and (b) against $A$ losing the first two games to $B$?

### Solution-6-48

\begin{align*}
P(\{A \text{ winning }\}) = \frac{3}{5} \\ \\ \\
P(\{B \text{ winning }\}) = \frac{2}{5}
\end{align*}

(a)<br><br>
The odds that $A$ wins at least two games is the same as the odds of $A$ winning two games plus the odds of $A$ winning three games.<br><br>
There are three out of eight ways $A$ can win twice: (A, A, B), (A, B, A), and (B, A, A). 
\begin{align*}
\text{Pr}\{A \text{ wins two games}\} &= \left( \text{Pr}\{A\} \times \text{Pr}\{A\} \times \text{Pr}\{B\} \right) +
\left( \text{Pr}\{A\} \times \text{Pr}\{B\} \times \text{Pr}\{A\} \right)
+ \left( \text{Pr}\{B\} \times \text{Pr}\{A\} \times \text{Pr}\{A\} \right) \\ \\
&= \left( \frac{3}{5} \times \frac{3}{5} \times \frac{2}{5} \right)
+ \left( \frac{3}{5} \times \frac{2}{5} \times \frac{3}{5} \right)
+ \left( \frac{2}{5} \times \frac{3}{5} \times \frac{3}{5} \right) \\ \\
&= \left( 3 \times \left( \frac{3}{5} \right)^2 \times \frac{2}{5} \right) \\ \\
&= \frac{54}{125} \\ \\ \\ \\
\text{Pr}\{A \text{ wins all three games}\} &= \left( \text{Pr}\{A\} \times \text{Pr}\{A\} \times \text{Pr}\{B\} \right) \\ \\
&= \left( \frac{3}{5} \right)^3 \\ \\
&= \frac{27}{125} \\ \\ \\ \\
\text{Pr}\{A \text{ wins at least two out of three games}\} &= \frac{54}{125} + \frac{27}{125} = \frac{81}{125}
\end{align*}

In ratio form: $81:44$<br>

(b)<br><br>
The odds against $A$ losing the first two games to $B$ is the same as for $B$ not winning the first two games.
\begin{align*}
\text{Pr}\{B \text{ not winning first two games}\} = 1 - \text{Pr}\{B \text{ winning the first two games}\} = \left( \frac{2}{5} \right)^2 
= 1 - \frac{4}{25} = \frac{21}{25}
\end{align*}

In ratio form: $21:4$


#### Alternate Method to solve the previous problem

To solve these problems, we'll first calculate the probabilities of the different outcomes, and then convert those probabilities into odds.

Given that the odds in favor of $A$ winning a game against $B$ are $3:2$, we can interpret this as follows: out of every 5 games, $A$ wins 3 and $B$ wins 2. Let's denote $P(A)$ as the probability of $A$ winning a single game and $P(B)$ as the probability of $B$ winning a single game.

We can set up the following equation based on the given odds:

\[ \frac{P(A)}{P(B)} = \frac{3}{2} \]

Solving this equation, we get:

\[ P(A) = \frac{3}{3 + 2} = \frac{3}{5} \]

\[ P(B) = \frac{2}{3 + 2} = \frac{2}{5} \]

(a) Odds in favor of $A$ winning at least two games out of the three:

To find this probability, we need to consider the different combinations of wins $A$ can have: 2 wins, 3 wins. We'll calculate the probabilities for each of these scenarios and then add them together.

1. Probability of $A$ winning exactly 2 games:
\[ P(\text{2 wins for } A) = \binom{3}{2} \times (P(A))^2 \times (1 - P(A))^1 \]
\[ P(\text{2 wins for } A) = \binom{3}{2} \times \left(\frac{3}{5}\right)^2 \times \left(\frac{2}{5}\right)^1 \]

2. Probability of $A$ winning all 3 games:
\[ P(\text{3 wins for } A) = (P(A))^3 \]
\[ P(\text{3 wins for } A) = \left(\frac{3}{5}\right)^3 \]

Now, we can calculate these probabilities:

\[ P(\text{2 wins for } A) = \binom{3}{2} \times \left(\frac{3}{5}\right)^2 \times \left(\frac{2}{5}\right)^1 = 3 \times \frac{9}{25} \times \frac{2}{5} = \frac{54}{125} \]

\[ P(\text{3 wins for } A) = \left(\frac{3}{5}\right)^3 = \frac{27}{125} \]

Adding these probabilities together gives us the probability of $A$ winning at least two games out of three:

\[ P(\text{at least 2 wins for } A) = P(\text{2 wins for } A) + P(\text{3 wins for } A) = \frac{54}{125} + \frac{27}{125} = \frac{81}{125} \]

Now, let's convert this probability into odds:

\[ \text{Odds in favor of } A = \frac{P(\text{winning})}{P(\text{losing})} \]

\[ \text{Odds in favor of } A = \frac{\frac{81}{125}}{1 - \frac{81}{125}} = \frac{\frac{81}{125}}{\frac{44}{125}} = \frac{81}{44} \]

So, the odds in favor of $A$ winning at least two games out of the three are $81:44$.

(b) Odds against $A$ losing the first two games to $B$:

To calculate this, we find the probability of $A$ losing the first two games and then subtract that from 1 to get the probability of $A$ not losing the first two games. Finally, we convert this probability into odds.

1. Probability of $A$ losing the first two games:
\[ P(\text{lose first 2 games}) = P(B) \times P(B) = \left(\frac{2}{5}\right)^2 \]

Now, let's calculate this probability:

\[ P(\text{lose first 2 games}) = \left(\frac{2}{5}\right)^2 = \frac{4}{25} \]

Now, the probability of $A$ not losing the first two games:

\[ P(\text{not lose first 2 games}) = 1 - P(\text{lose first 2 games}) = 1 - \frac{4}{25} = \frac{21}{25} \]

Now, let's convert this probability into odds:

\[ \text{Odds against } A = \frac{P(\text{losing})}{P(\text{winning})} \]

\[ \text{Odds against } A = \frac{1 - \frac{21}{25}}{\frac{21}{25}} = \frac{\frac{4}{25}}{\frac{21}{25}} = \frac{4}{21} \]

So, the odds against $A$ losing the first two games to $B$ are $4:21$.

## Problem-6-49

A purse contains 2 silver coins and 4 copper coins, and a second purse contains 4 silver coins and 3 copper coins. If a coin is selected at random from one of the two purses, what is the probability that it is a silver coin?

### Solution-6-49

There is a fifty-fifty chance of picking a coin out of either purse. From the first purse you have a $\frac{2}{6}=\frac{1}{3}$ chance of selecting a silver coin. You have a $\frac{4}{7}$ chance of selecting a silver coin from the second purse.<be>

\begin{align*}
\text{Pr}\{\text{Selecting a silver coin from two purses}\} &=
\frac{1}{2} \times \text{Pr}\{ \text{Selecting a silver coin from the first purse} \} +
\frac{1}{2} \times \text{Pr}\{\text{Selecting a silver coin from the second purse} \} \\ &=
\left( \frac{1}{2} \times \frac{1}{3} \right) + \left( \frac{1}{2} \times \frac{4}{7} \right) \\ &=
\frac{1}{6} + \frac{4}{14} \\ &=
\frac{19}{42}
\end{align*}
