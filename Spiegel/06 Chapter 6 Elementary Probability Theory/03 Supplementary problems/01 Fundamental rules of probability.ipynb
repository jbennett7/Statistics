# Fundamental rules of probability

## Problem-6-40
Determine the probability $p$, or an estimate of it, for each of the following events.

&emsp;&emsp;(a)&ensp;A king, ace, jack of clubs, or queen of diamonds appears in drawing a single card from a well-shuffled ordinary deck of cards.  
&emsp;&emsp;(b)&ensp;The sum $8$ appears in a single toss of a pair of fair dice.  
&emsp;&emsp;(c)&ensp;A nondefective bolt will be found if out of $600$ bolts already examined, 12 were defective.  
&emsp;&emsp;(d)&ensp;A $7$ or $11$ comes up in a single toss of a pair of fair dice.  
&emsp;&emsp;(e)&ensp;At least one head appears in three tosses of a fair coin.

### Solution-6-40
&emsp;&emsp;(a)&ensp;Each of the events: A king, ace, or jack of clubs, or a queen of diamonds have an equal chance of occurring $p=\frac{1}{52}$.  
&emsp;&emsp;(b)&ensp;For a pair of dice, there are five different combinations to get $8$: (2,6), (3,5), (4,4), (5,3), (2,6). There are a total of $6\times6=36$ combinations. Therefore, $p=\frac{5}{36}$.  
&emsp;&emsp;(c)&ensp;Using the previous $600$ observations, we can conclude that the probability of finding a defective bolt is $p=\frac{12}{600}=\frac{1}{50}$.  
&emsp;&emsp;(d)&ensp;For a pair of dice, there are six different combinations to get $7$: (1,6), (2,5), (3,4), (4,3), (5,2), (6,1);
and there are two different combinations to get $11$: (5,6), (6,5).  
&emsp;&emsp;&emsp;&ensp;With $36$ total combinations, the probability is $p=\frac{8}{36}=\frac{2}{9}$.  
&emsp;&emsp;(e)&ensp;For three tosses of a fair coin, there are $8$ different combinations:
(H,H,H), (H,H,T), (H,T,H), (H,T,T), (T,H,H), (T,H,T), (T,T,H), (T,T,T).  
&emsp;&emsp;&emsp;&ensp;There is only one combination where a head does not appear, therefore the probability is $p=\frac{7}{8}$.

## Problem-6-41
An experiment consists of drawing three cards in succession from a well-shuffled ordinary deck of cards. Let $E_1$ be the event "king" on the first draw, $E_2$ the event "king" on the second draw, and $E_3$ the event "king" on the third draw. State in words the meaning of each of the following:  
(a)&ensp;$\text{Pr}\{E_1\bar{E_2}\}$
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;(c)&ensp;$\bar{E}_1+\bar{E}_2$
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;(e)&ensp;$\bar{E}_1\bar{E}_2\bar{E}_3$  
(b)&ensp;$\text{Pr}\{E_1+E_2\}$
&emsp;&emsp;&emsp;&emsp;&ensp;(d)&ensp;$\text{Pr}\{E_3|E_1\bar{E}_2\}$
&emsp;&emsp;&emsp;&ensp;(f)&ensp;$\text{Pr}\{E_1E_2+\bar{E}_2E_3\}$

### Solution-6-41
&emsp;(a)&ensp;The probability of "king" on first draw, and the probability of not "king" on the second draw.  
&emsp;(b)&ensp;The probability of "king" on first draw, or the probability of "king" on the second draw.  
&emsp;(c)&ensp;Not "king" on first draw, or not "king" on second draw.  
&emsp;(d)&ensp;The probability of "king" on first draw and not "king" on second draw, with the condition of "king" on third draw.  
&emsp;(e)&ensp;Not "king" on first draw, and not "king" on second draw, and not "king" on third draw.    
&emsp;(f)&ensp;The probability of "king" on first draw and "king" on second draw, or the probability of not "king" on second draw and "king" on third draw.

## Problem-6-42
A ball is drawn at random from a box containing $10$ red, $30$ white, $20$ blue, and $15$ orange marbles. Find the probability that the ball drawn is (a) orange or red, (b) not red or blue, (c) not blue, (d) white, and (e) red, white, or blue.

### Solution-6-42
There are a total of $10+30+20+15=75$ balls.  Event red = $R$; event white $W$; event blue $B$; and event orange $O$.  
&emsp;(a)&ensp;The probability of "orange" or "red" is:
$$
\text{Pr}\{O+R\}=\frac{15}{75}+\frac{10}{75}=\frac{25}{75}=\frac{1}{3}\text{.}
$$
&emsp;(b)&ensp;The probability of not red or blue:  
&emsp;&emsp;The probability of not red or blue is the same as the sum of all the other colors:
$$
\text{Pr}\{\bar{R}+\bar{B}\}=\text{Pr}\{W+O\}=\frac{45}{75}=\frac{3}{5}\text{.}
$$
&emsp;(c)&ensp;The probability of "not blue":
$$
\text{Pr}\{\bar{B}\}=\frac{75-20}{75}=\frac{11}{15}\text{.}
$$
&emsp;(d)&ensp;The probability of "white":
$$
\text{Pr}\{W\}=\frac{30}{75}=\frac{2}{5}\text{.}
$$
&emsp;(e)&ensp;The probability of "red", "white", or "blue":
$$
\text{Pr}\{R+W+B\}=\frac{10}{75}+\frac{30}{75}+\frac{20}{75}=\frac{60}{75}=\frac{4}{5}\text{.}
$$

## Problem-6-43
Two marbles are drawn in succession from the box of Problem 6.42, replacement being made after each drawing. Find the probability that (a) both are white, (b) the first is red and the second is white, (c) neither is orange, (d) they are either red or white or both (red and white), (e) the second is not blue, (f) the first is orange (g) at least one is blue, (h) at most one is red, (i) the first is white but the second is not, and (j) only one is red.

### Solution-6-43
The same for problem 6.42 applies: $75$ marbles. The two events: first draw $E_1$, and second draw $E_2$ are
independent since they are drawn with replacement.  
&emsp;(a)&ensp;Both events are white:
$$
\text{Pr}\{WW\} =\frac{30}{75}\times\frac{30}{75}=\frac{900}{5625}=\frac{4}{25}\text{.}
$$
&emsp;(b)&ensp;The first is red and the second is white
$$
\text{Pr}\{RW\} = \frac{10}{75}\times\frac{30}{75}=\frac{4}{75}\text{.}
$$
&emsp;(c)&ensp;neither is orange:
$$
\text{Pr}\{\bar{O}\bar{O}\}=\frac{60}{75}\times\frac{60}{75}=\frac{16}{25}\text{.}
$$
&emsp;(d)&ensp;either red or white or both (red and white).  
&emsp;&emsp;There are several possibilities here:  
&emsp;&emsp;&emsp;(1)&ensp; both red
$$
\text{Pr}\{RR\}=\frac{10}{75}\times\frac{10}{75}=\frac{100}{5625}=\frac{4}{225}\text{.}
$$
&emsp;&emsp;&emsp;(2)&ensp; both white
$$
\text{Pr}\{WW\}=\frac{30}{75}\times\frac{30}{75}=\frac{900}{5625}=\frac{36}{225}\text{.}
$$
&emsp;&emsp;&emsp;(3)&ensp; either red or white
$$
\text{Pr}\{WR+RW\}=2\times\frac{30}{75}\times\frac{10}{75}=\frac{24}{225}\text{.}
$$
&emsp;&emsp;&emsp;The sum is:
$$
\frac{4}{225}+\frac{36}{225}+\frac{24}{225}=\frac{64}{225}\text{.}
$$
&emsp;(e)&ensp; the second is not blue:
$$
\text{Pr}\{B\} = \frac{75}{75} - \frac{20}{75} = \frac{55}{75} = \frac{11}{15}\text{.}
$$
&emsp;(f)&ensp; the first is orange:
$$
\text{Pr}\{O\} = \frac{15}{75} = \frac{1}{5}\text{.}
$$
&emsp;(g)&ensp; at least one is blue:  
&emsp;&emsp;To find the probability of drawing at least one blue, we need to find the probability of drawing no blue
$$
\text{Pr}\{\bar{B}\} = \frac{75-20}{75}=\frac{55}{75}\text{.}
$$
&emsp;&emsp;The probability of drawing no blue balls in two successive draws
$$
\text{Pr}\{\bar{B}\bar{B}\}=\frac{55}{75}\times\frac{55}{75}=\frac{121}{225}\text{.}
$$
&emsp;&emsp;The probability of drawing at least one blue ball is the complement of drawing no blue balls in two successive draws
$$
\text{Pr}\{\text{at least one blue}\} = \frac{225-121}{225} = \frac{104}{225}\text{.}
$$
&emsp;(h)&ensp; at most, one is red:  
&emsp;&emsp;To find the probability of at most one red, we need to find the probability of drawing two red balls
$$
\text{Pr}\{RR\}=\frac{10}{75}\times\frac{10}{75}=\frac{4}{225}\text{.}
$$
&emsp;&emsp;The probability of drawing at most one red ball is the complement of drawing two red balls
$$
\text{Pr}\{\text{at most one red}\} = \frac{225-4}{225} = \frac{221}{225}\text{.}
$$
&emsp;(i)&ensp;The probability of the first draw is White but the second is not
$$
\text{Pr}\{W\bar{W}\} = \frac{30}{75}\times\frac{45}{75} = \frac{1350}{5625} = \frac{6}{25}\text{.}
$$
&emsp;(j)&ensp;The probability that only one of the draws is red, consider the following two scenarios:  
&emsp;&emsp;Drawing one red and not drawing red, and not drawing red and drawing red:
$$
\text{Pr}\{R\bar{R}\}=\text{Pr}\{\bar{R}R\}=2\times\frac{10}{75}\times\frac{65}{75}=\frac{1300}{5625}=\frac{52}{225}\text{.}
$$
