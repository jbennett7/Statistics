## Problem-6-53

A continuous random variable $X$ that can assume values only between $X=2$ and $8$ inclusive has a density function given by $a(X+3)$, where $a$ is a constant. (a) Calculate $a$. Find (b) $\text{Pr}\{3 < X < 5\}$, (c) $\text{Pr}\{X \ge 4\}$, and (d) $\text{Pr}\{|X - 5| < 0.5\}$.

### Solution-6-53

(a)
We need to integrate $\int_2^8 a (x+3)dx$

\begin{align*}
\int_2^8 a (x+3) dx &= 1 \\ \\
a \left[ \frac{1}{2} x^2 + 3x \right]_2^8 &= \\ \\
a \left( (32 + 24) - (2 + 6) \right) &= 48a \\ \\
48a &= 1 \\ \\
a &= \frac{1}{48}
\end{align*}

(b) To find $\text{Pr}\{3 < X < 5\}$<br><br>

\begin{align*}
\frac{1}{48} \int_3^5 (x+3)dx &= \\ \\
\frac{1}{48} \left[ \frac{1}{2} x^2 + 3x \right]_3^5 &= \\ \\
\frac{1}{48} \left( \frac{25}{2} + 15 - \frac{9}{2} - 9 \right) &= \\ \\
\frac{14}{48} &=  \\ \\
\frac{7}{24}
\end{align*}

(c) To find $\text{Pr}\{X \ge 4\}$<br><br>

\begin{align*}
\frac{1}{48} \int_4^8 (x+3)dx &= \\ \\
\frac{1}{48} \left[ \frac{1}{2} x^2 + 3x \right]_4^8 &= \\ \\
\frac{1}{48} \left( 56 - 20 \right) &= \\ \\
\frac{36}{48} &=  \\ \\
\frac{18}{24}
\end{align*}

(d) To find $\text{Pr}\{|X - 5| < 0.5\}$.

The interval can be rewritten
\begin{align*}
|X - 5| < 0.5 \\ \\
-0.5 < X - 5 < 0.5 \\ \\
4.5 < X < 5.5
\end{align*}

Then we solve
\begin{align*}
\frac{1}{48} \int_{4.5}^{5.5} (x+3)dx &= \\ \\
\frac{1}{48} \left[ \frac{1}{2} x^2 + 3x \right]_{4.5}^{5.5} &= \\ \\
\frac{1}{48} \left( \frac{30.25}{2} + 16.5 + \frac{20.25}{2} - 13.5 \right) &= \\ \\
\frac{8}{48} &=  \\ \\
\frac{1}{6}
\end{align*}

## Problem-6-54

Three marbles are drawn without replacement from an urn containing 4 red and 6 white marbles. If $X$ is a random variable that denotes the total number of red marbles drawn, (a) construct a table showing the probability distribution of $X$ and (b) graph the distribution.

### Solution-6-54

(a)<br>

\begin{align*}
\text{Pr}\{\text{zero red marbles}\} = \text{Pr}\{\text{three white marbles}\} =
\frac{6}{10} \times \frac{5}{9} \times \frac{4}{8} = \frac{120}{720} = \frac{1}{6} \\ \\ \\
\text{Pr}\{\text{one red marble}\} =
3 \times \frac{4}{10} \times \frac{6}{9} \times \frac{5}{8} = \frac{360}{720} = \frac{1}{2} \\ \\ \\
\text{Pr}\{\text{two red marbles}\} =
3 \times \frac{4}{10} \times \frac{3}{9} \times \frac{6}{8} = \frac{216}{720} = \frac{3}{10} \\ \\ \\
\text{Pr}\{\text{three red marbles}\} =
\frac{4}{10} \times \frac{3}{9} \times \frac{2}{8} = \frac{24}{720} = \frac{1}{30}
\end{align*}

```r
data <- c(rep(0,120),rep(1,360),rep(2,216),rep(3,24))
table(data)
barplot(table(data))
```

To construct the probability distribution of \( X \), which represents the total number of red marbles drawn when three marbles are drawn without replacement from the urn, we need to consider all possible outcomes.

There are a total of \( \binom{10}{3} \) ways to draw three marbles from the urn.

Let's construct the table:

| X (Number of Red Marbles) | Probability (P(X)) |
|----------------------------|--------------------|
| 0                          | P(X = 0)           |
| 1                          | P(X = 1)           |
| 2                          | P(X = 2)           |
| 3                          | P(X = 3)           |

Now, we calculate the probabilities for each value of \( X \):

1. \( P(X = 0) \): Probability of drawing all three white marbles.
\[ P(X = 0) = \frac{\binom{6}{3}}{\binom{10}{3}} \]

2. \( P(X = 1) \): Probability of drawing two white marbles and one red marble.
\[ P(X = 1) = \frac{\binom{6}{2} \times \binom{4}{1}}{\binom{10}{3}} \]

3. \( P(X = 2) \): Probability of drawing one white marble and two red marbles.
\[ P(X = 2) = \frac{\binom{6}{1} \times \binom{4}{2}}{\binom{10}{3}} \]

4. \( P(X = 3) \): Probability of drawing all three red marbles.
\[ P(X = 3) = \frac{\binom{4}{3}}{\binom{10}{3}} \]

Let's calculate these probabilities:

1. \( P(X = 0) \):
\[ P(X = 0) = \frac{\binom{6}{3}}{\binom{10}{3}} = \frac{20}{120} = \frac{1}{6} \]

2. \( P(X = 1) \):
\[ P(X = 1) = \frac{\binom{6}{2} \times \binom{4}{1}}{\binom{10}{3}} = \frac{15 \times 4}{120} = \frac{60}{120} = \frac{1}{2} \]

3. \( P(X = 2) \):
\[ P(X = 2) = \frac{\binom{6}{1} \times \binom{4}{2}}{\binom{10}{3}} = \frac{6 \times 6}{120} = \frac{36}{120} = \frac{3}{10} \]

4. \( P(X = 3) \):
\[ P(X = 3) = \frac{\binom{4}{3}}{\binom{10}{3}} = \frac{4}{120} = \frac{1}{30} \]

Therefore, the probability distribution of \( X \) is as follows:

| X (Number of Red Marbles) | Probability (P(X)) |
|----------------------------|------------------|
| 0                          | $\frac{1}{6}$    |
| 1                          | $ \frac{1}{2}$   |
| 2                          | $ \frac{3}{10} $ |
| 3                          | $ \frac{1}{30} $ |

## Problem-6-55

(a) Three dice are rolled and $X =$ the sum on the three upturned faces. Give the probability distribution for $X$. (b) Find $\text{Pr}\{7 \le X \le 11\}$.

### Solution-6-55

```r
Die1 <- c()
Die2 <- c()
Die3 <- c()
for (i in 1:6){
    for (j in 1:6) {
        for (k in 1:6) {
            Die1 <- c(Die1,k)
            Die2 <- c(Die2,j)
            Die3 <- c(Die3,i)
        }
    }
}
Sum <- Die1 + Die2 + Die3
df <- data.frame(Die3, Die2, Die1, Sum)
(t <- table(df$Sum))
barplot(t)
sum(t[5:9])/sum(t)
```
