## Problem-6-58

$A$ and $B$ play a game in which they toss a fair coin 3 times. The one obtaining heads first wins the game. If $A$ tosses the coin first and if the total value of the stakes is $\$20$, how much should be contributed by each in order that the game be considered fair?

### Solution-6-58

```r
cat("A should contribute: ", (a <- 20*(.5+.5^2+.5^3)),
    "\nB should contribute: ", 20 - a, "\n")
```

To determine a fair contribution for each player, we need to calculate the probability of A winning and the probability of B winning.

Let's denote:

- \( p \) as the probability of A winning the game.
- \( q \) as the probability of B winning the game.

Since A tosses the coin first, they can win in the following scenarios:

1. A wins on the first toss.
2. A wins on the second toss and B gets tails on the first toss.
3. A wins on the third toss, B gets tails on the first two tosses.

Given that the coin is fair, the probability of getting heads or tails in each toss is \( \frac{1}{2} \).

So, the probability of A winning is:

\[ p = \frac{1}{2} + \frac{1}{2} \times \frac{1}{2} + \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} = \frac{1}{2} + \frac{1}{4} + \frac{1}{8} = \frac{7}{8} \]

Similarly, B can only win if A does not win in the first three tosses. So, the probability of B winning is:

\[ q = 1 - p = 1 - \frac{7}{8} = \frac{1}{8} \]

To make the game fair, the expected value for both players should be equal. Thus, the expected payoff for both players should be zero. Since the total value of the stakes is $20, and there are two players, each player's expected payoff should be $10.

Let \( x \) be the amount contributed by A and \( y \) be the amount contributed by B.

The expected payoff for A is:

\[ E(A) = 10 \times p - x = 10 \times \frac{7}{8} - x \]

The expected payoff for B is:

\[ E(B) = 10 \times q - y = 10 \times \frac{1}{8} - y \]

For the game to be fair, \( E(A) = E(B) = 0 \).

\[ 10 \times \frac{7}{8} - x = 0 \]
\[ 10 \times \frac{1}{8} - y = 0 \]

Solving these equations for \( x \) and \( y \):

\[ x = 10 \times \frac{7}{8} = \frac{70}{8} = \frac{35}{4} \]
\[ y = 10 \times \frac{1}{8} = \frac{10}{8} = \frac{5}{4} \]

So, A should contribute \( \frac{35}{4} \) dollars and B should contribute \( \frac{5}{4} \) dollars to make the game fair.

## Problem-6-59

Find (a) $E(X)$, (b) $E(X^2)$, (c) $E[(X - \bar{X})^2]$, and (d) $E(X^3)$ for the probability distribution of Table 6.4.

### Notes

In probability theory, $E[(X - \bar{X})]$ represents the expected value of the deviation of a random variable $X$ from its mean $\bar{X}$. 

Here, $X$ is a random variable and $\bar{X}$ is the mean (or expected value) of $X$.

The expression $(X - \bar{X})$ calculates the deviation of each individual value of $X$ from its mean. If $X$ is consistently greater than its mean, $(X - \bar{X})$ will be positive. If $X$ is consistently less than its mean, $(X - \bar{X})$ will be negative. By taking the expected value of this deviation, $E[(X - \bar{X})]$, we're essentially finding the average deviation of $X$ from its mean over all possible outcomes.

This expression is important in understanding the dispersion or variability of the random variable $X$ around its mean. If $E[(X - \bar{X})] = 0$, it indicates that, on average, the random variable $X$ doesn't deviate from its mean. If $E[(X - \bar{X})] > 0$, it suggests that, on average, $X$ tends to be greater than its mean, and vice versa for $E[(X - \bar{X})] < 0$.

$\bar{X} = \sum_{i} X_i \cdot p(X_i)$<br>

To prove that $E(X) = p$, where $X$ is a random variable that takes the value 1 with probability $p$ and 0 with probability $q = 1 - p$, we'll use the definition of expected value.

The expected value $E(X)$ of a random variable $X$ is given by the sum of each possible value of $X$ multiplied by its probability.

For this random variable, $X$, we have:

- $P(X = 1) = p$
- $P(X = 0) = q = 1 - p$

So, the expected value \( E(X) \) can be expressed as:

\begin{align*}
E(X) &= (1) \times P(X = 1) + (0) \times P(X = 0) \\

\end{align*}

Thus, we have proved that $E(X) = p$.

To prove that $E[(X - \bar{X})^2] = pq$, where $X$ is a random variable that takes the

- $P(X = 0) = q = 1 - p$

Now, let's calculate $E[(X - \bar{X})^2]$:

$E[(X - \bar{X})^2] = \sum_{x} (x - \bar{X})^2 \cdot p(x)$

$E[(X - \bar{X})^2] = (1 - p)^2 \cdot p + (0 - p)^2 \cdot q$

$E[(X - \bar{X})^2] = p - p^2$
