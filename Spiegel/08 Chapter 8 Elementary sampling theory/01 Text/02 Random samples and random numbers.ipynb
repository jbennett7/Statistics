# Random samples and random numbers
In order that the conclusions of sampling theory and statistical inference be valid, samples must be chosen so as to be _representative_ of a population. A study of sampling methods and of related problems that arise is called the _design of the experiment_.  

One way in which a representative sample may be obtained is by a process called _random sampling_, according to which each member of a population has an equal chance of being included in the sample. One technique for obtaining a random sample is to assign numbers to each member of the population, write these numbers on samll pieces of paper, place them in an urn, and then draw numbers from the urn, being careful to mix thoroughly before each drawing. An alternative method is to use a table of _random numbers_ (see Appendix IX) specially constructed for such purposes. See Problem 8.6.
