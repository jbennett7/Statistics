# The standard deviation
The _standard deviation_ of a set of $N$ numbers $X_1,X_2,\ldots,X_N$ is denoted by $s$ and is defined by

<a id=eq5></a>
\begin{align}\tag{5}
s = \sqrt{\frac{\sum^N_{j=1}{(X_j-\bar{X}})^2}{N}} = \sqrt{\frac{\sum{(X-\bar{X})^2}}{N}}
  = \sqrt{\frac{\sum{x^2}}{N}} = \sqrt{\overline{(X-\bar{X})^2}}
\end{align}

where $x$ represents the deviations of each of the numbers $X_j$ from the mean $\bar{X}$. Thus $s$ is the root mean square ($\text{RMS}$) of the deviations from the mean, or, as it is sometimes called, the _root-mean-square deviation_.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;If $X_1,X_2,\ldots,X_K$ occur with frequencies $f_1,f_2,\ldots,f_K$, respectively, the standard deviation can be written

<a id=eq6></a>
\begin{align}\tag{6}
s = \sqrt{\frac{\sum^N_{j=1}{f_j(X_j-\bar{X}})^2}{N}} = \sqrt{\frac{\sum{f(X-\bar{X})^2}}{N}}
  = \sqrt{\frac{\sum{fx^2}}{N}} = \sqrt{\overline{(X-\bar{X})^2}}
\end{align}

where $N=\sum^K_{j=1}=\sum{f}$. In this form it is useful for grouped data.  

### NOTE
Sometimes the standard deviation of a sample's data is defined with $(N-1)$ replacing $N$ in the denominators of the expressions in equations(5) and (6) because the resulting value represents a better estimate of the standard deviation of a population from which the sample is taken. For large values of $N$ (certainly $N>30$), there is practically no difference between the two definitions. Also, when the better estimate is needed we can always obtain it by multiplying the standard deviation computed according to the first definition by $\sqrt{N/(N-1)}$. Hence we shall adhere to the form (5) and (6).

## The standard deviation in R

The `sd()` replaces $N$ with $(N - 1)$. Therefore, we need to define our own standard deviation function

```r
sd.pop <- function(x) { sqrt(sum((x - mean(x))^2)/length(x)) }
```

```r
x.a <- c(12, 6, 7, 3, 15, 10, 18, 5)
x.b <- c(9, 3, 8, 8, 9, 8, 9, 18)
sd(x.a)
sd.pop(x.a)
sd(x.b)
sd.pop(x.b)
```
