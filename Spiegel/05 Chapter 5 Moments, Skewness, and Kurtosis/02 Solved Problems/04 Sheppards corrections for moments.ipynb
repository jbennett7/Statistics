# Sheppard's corrections for moments

## Problem 5-9
Apply Sheppard's corrections to determine the moments about the mean for the data in (a) Problem [5.6](#Problem-5-6) and (b) Problem [5.7](#Problem-5-7).

### Solution 5-9

```r
p5.9 <- new.env()
p5.9$m2.a.orig <- 8.5275
p5.9$m4.a.orig <- 199.3759
p5.9$C.a <- 3
cat('Corrected m2 = m2 - c^2/12 = ', p5.9$m2.a.orig - p5.9$C.a^2/12, '\n')
cat('Corrected m4 = m4 - (1/2)*c^2*m2 + (7/240)*c^4 = ', p5.9$m4.a.orig - .5*p5.9$C.a^2*p5.9$m2.a.orig + 7/240*p5.9$C.a^4, '\n')
```
