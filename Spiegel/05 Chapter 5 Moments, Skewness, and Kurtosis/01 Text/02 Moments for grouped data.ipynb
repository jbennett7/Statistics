# Moments for grouped data
If $X_1, X_2,\ldots, X_K$ occur with frequencies $f_1, f_2,\ldots, f_K$, respectively, the above moments are given by

\begin{align}
\overline{X^r}=\frac{f_1X^r_1+f_2X^r_2+\ldots+f_KX^r_K}{N}
=\frac{\sum^K_{j=1}{f_jX_j^r}}{N}=\frac{\sum{fX^r}}{N}\label{eq:rth_moment_grouped}\tag{4}
\end{align}

\begin{align}
m_r=\frac{\sum^K_{j=1}{f_j(X_j=\bar{X})^r}}{N}
=\frac{\sum{f(X-A)^r}}{N}=\overline{(X-\bar{X})^r}\label{eq:rth_moment_mean_grouped}\tag{5}
\end{align}

\begin{align}
m^\prime_r=\frac{\sum^K_{j=1}{f_j(X_j-A)^r}}{N}
=\frac{\sum{f(X-A)^r}}{N}=\overline{(X-A)^r}\label{eq:rth_moment_origin_grouped}\tag{6}
\end{align}

where $N=\sum^K_{j=1}{f_j}=\sum{f}$. The formulas are suitable for calculating moments from grouped data.
