# The arithmetic mean

## Problem 3-6
The grades of a student on six examinations were $84, 91, 72, 68, 87,$ and $78$. Find the arithmetic mean of the grades.

### Solution 3-6

\begin{align}
\bar{X}=\frac{\sum{X}}{N}=\frac{84+91+72+68+87+78}{6}=\frac{480}{6}=80
\end{align}

Frequently one uses the term _average_ synonymously with _arithmetic mean_. Strictly speaking, however, this is incorrect since there are averages other than the arithmetic mean.

```r
p3.6 <- c(84, 91, 72, 68, 87, 78)
mean(p3.6)
```

## Problem 3-9
Find the arithmetic mean of the numbers $5, 3, 6, 5, 4, 5, 2, 8, 6, 5, 4, 8, 3, 4, 5, 4, 8, 2, 5,$ and $4$.

### Solution 3-9
#### First Method
\begin{align}
\bar{X}=\frac{\sum{X}}{N}=\frac{5+3+6+5+4+5+2+8+6+5+4+8+3+4+5+4+8+2+5+4}{20}=\frac{96}{20}=4.8
\end{align}
#### Second method
There are six 5's, two 3's, two 6's, five 4's, two 2's, and three 8's. Thus

\begin{align}
\bar{X}=\frac{\sum{fX}}{\sum{f}}=\frac{\sum{fX}}{N}=\frac{(6)(5)+(2)(3)+(2)(6)+(5)(4)+(2)(2)+(3)(8)}{6+2+2+5+2+3}=\frac{96}{20}=4.8
\end{align}

```r
p3.9 <- c(5, 3, 6, 5, 4, 5, 2, 8, 6, 5, 4, 8, 3, 4, 5, 4, 8, 2, 5, 4)
mean(p3.9)
```

## Problem 3-10
Out of $100$ numbers, $20$ were $4$'s, $40$ were $5$'s, $30$ were $6$'s and the remainder were $7$'s. Find the arithmetic mean of the numbers.

### Solution 3-10
\begin{align}
\bar{X}=\frac{\sum{fX}}{\sum{f}}=\frac{\sum{fX}}{N}=\frac{(20)(4)+(40)(5)+(30)(6)+(10)(7)}{100}=\frac{530}{100}=5.30
\end{align}

## Problem 3-11
A student's final grades in mathematics, physics, English, and hygiene are, respectively, $82, 86, 90,$ and $70$. If the respective credits received for these courses are $3, 5, 3,$ and $1$, determine an appropriate average grade.

### Solution 3-11
We use a weighted arithmetic mean, the weights associated with each grade being taken as the number of credits received. Thus

\begin{align}
\bar{X}=\frac{\sum{wX}}{\sum{w}}=\frac{(3)(82)+(5)(86)+(3)(90)+(1)(70)}{3+5+3+1}=85
\end{align}

## Problem 3-12
In a company having $80$ employees, $60$ earn $\$10.00$ per hour and $20$ earn $\$13.00$ per hour.  
(a) Determine the mean earnings per hour.  
(b) Would the answer in part (a) be the same if the $60$ employees earn a mean hourly wage of $\$10.00$ per hour? Prove your answer.  
(c) Do you believe the mean hourly wage to be typical?

### Solution 3-12
(a)
\begin{align}
\bar{X}=\frac{\sum{fX}}{N}=\frac{(60)(\$10.00)+(20)(\$13.00)}{60+20}=\$10.75
\end{align}

(b) Yes, the result is the same. To prove this, suppose that $f_1$ numbers have mean $m_1$ and that $f_2$ numbers have mean $m_2$. We must show that the mean of all the numbers is

\begin{align}
\bar{X}=\frac{f_1m_1+f_2m_2}{f_1+f_2}
\end{align}

Let the $f_1$ numbers add up to $M_1$ and the $f_2$ numbers add up to $M_2$. Then by definition of the arithmetic mean,

\begin{align}
m_1=\frac{M_1}{f_1}\ \ \text{ and }\ \ 
m_2=\frac{M_2}{f_2}
\end{align}

or $M_1=f_1m_1$ and $M_2=f_2m_2$. Since all $(f_1+f_2)$ numbers add up to $(M_1+M_2)$, the arithmetic mean of all numbers is

\begin{align}
\bar{X}=\frac{M_1+M_2}{f_1+f_2}=\frac{f_1m_1+f_2m_2}{f_1+f_2}
\end{align}

as required. The result is easily extended.

(c) We can say that $\$10.75$ is a "typical" hourly wage in the sense that most of the employees earn $\$10.00$, which is not too far from $\$10.75$ per hour. It must be remembered that, whenever we summarize numerical data into a single number (as is true in an average), we are likely to make some error. Certainly, however, the result is not as misleading as that in Problem [3.8](02%20The%20arithmetic%20mean.ipynb#Problem-3-8).

Actually, to be on safe ground, some estimate of the "spread," or "variation," of the data about the mean (or other average) should be given. This is called the _dispersion_ of the data. Various measures of dispersion are given in Chapter 4.

## Problem 3-13
For groups of students, consisting of $15, 20, 10,$ and $18$ individuals, reported mean weights of $162, 148, 153,$ and $140$ pounds (lb), respectively. Find the mean weight of all the students.

### Solution 3-13
\begin{align}
\bar{X}=\frac{\sum{fX}}{\sum{f}}=\frac{(15)(162)+(20)(148)+(10)(153)+(18)(140)}{15+20+10+18}=150\text{ lb}
\end{align}

## Problem 3-14
If the mean annual incomes of agricultural and nonagricultural workers are $\$25,000$ and $\$35,000$, respectively, would the mean annual income of both groups together be $\$30,000$?

### Solution 3-14
It would be $\$30,000$ only if the numbers of agricultural and nonagricultural workers were the same. To determine the true mean annual income, we would have to know the relative numbers of workers in each group. Suppose that $10\%$ of all workers are agricultural workers. Then the mean would be $(0.10)(25,000)+(0.90)(35,000)+=\$34,000$. If there were equal numbers of both types of workers, the mean would be $(0.50)(25,000)+(0.50)(35,000)=\$30,000$.
