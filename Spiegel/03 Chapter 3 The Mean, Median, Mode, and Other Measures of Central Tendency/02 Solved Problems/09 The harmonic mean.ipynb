# The harmonic mean

## Problem 3-39
Find the harmonic mean $H$ of the numbers $3, 5, 6, 6, 7, 10,$ and $12$.

### Solution 3-39
\begin{align}
\frac{1}{H}   & = \frac{1}{N}\sum{\frac{1}{X}}=\frac{1}{7}\left(\frac{1}{3}+\frac{1}{5}+\frac{1}{6}+\frac{1}{6}+\frac{1}{7}+\frac{1}{10}+\frac{1}{12}\right)=\frac{1}{7}\left(\frac{140+84+70+70+60+42+35}{420}\right) \\
              & = \frac{501}{2940} \\
\text{and } H & = \frac{2940}{501}=5.87
\end{align}

It is often convenient to express the fractions in decimal form first. Thus

\begin{align}
\frac{1}{H}   & = \frac{1}{7}(0.3333+0.2000+0.1667+0.1429+0.1000+0.0833) \\
              & = \frac{1.1929}{7} \\
\text{and } H & = \frac{7}{1.1929}=5.87
\end{align}

Comparison with Problem [3.35](08%20The%20geometric%20mean.ipynb#Problem-3-35) illustrates the fact that the harmonic ean of several positive numbers not all equal is less than their geometric mean, which is in turn less than their arithmetic mean.

# Problem 3-40
During four successive years, a home ownerpurchased oil for her furnace at respective costs of $\$0.80, \$0.90, \$1.05,$ and $\$1.25$ per gallon (gal). What was the average cost of oil over the 4-year period?

### Solution 3-40
<a id='Solution-3-40-1'></a>
__Case 1__  
Suppose that the home owner purchased the same quantity of oil each year, say $1000$ gal. Then

\begin{align}
\text{Average cost} = \frac{\text{total cost}}{\text{total quantity purchased}}=\frac{\$800+\$900+\$1050+\$1250}{4000\text{ gal}}=\$1.00/\text{gal}
\end{align}

This is the same as the arithmetic mean of the cost per gallon; that is, $\frac{1}{4}(\$0.80+\$0.90+\$1.05+\$1.25)=1.00/\text{gal}$. This result would be the same even if $x$ gallons were used each year.

<a id='Solution-3-40-2'></a>
__Case 2__  
Suppose that the home owner spends the same amount of money each year, say $\$1000$. Then

\begin{align}
\text{Average cost}=\frac{\text{total cost}}{\text{total quantity purchased}}=\frac{\$4000}{(1250+1111+952+800)\text{gal}}=\$0.975/\text{gal}
\end{align}

This is the same as the harmonic mean of the costs per gallon:

\begin{align}
\frac{4}{\frac{1}{0.80}+\frac{1}{0.90}+\frac{1}{1.05}+\frac{1}{1.25}}=0.975
\end{align}

The result would be the same even if $y$ dollars were spent each year.

Both averaging processes are correct, each average being computed under different prevailing conditions.

It should be noted that in case the number of gallons used changes from one year to another instead of remaining the same, the ordinary arithmetic mean of Case [1](#Solution-3-40-1) is replaced by a weighted arithmetic mean. Similarly, if the total amount spent changes from one year to another, the ordinary harmonic mean of Case [2](#Solution-3-40-2) is replaced by a weighted harmonic mean.

# Problem 3-41
A car travels $25$ miles at $25$ miles per hour (mi/h), $25$ miles at $50$ mph, and $25$ mph, and $25$ miles at $75$ mph. Find the arithmetic mean of the three velocities and the harmonic mean of the three velocities. Which is correct?

### Solution 3-41
The average velocity is equal to the distance traveled divided by the total time and is equal to the following:
\begin{align}
\frac{75}{\left(1+\frac{1}{2}+\frac{1}{3}\right)}=40.9\text{ mi/h}
\end{align}

The arithmetic mean of the three velocities is:
\begin{align}
\frac{25+50+75}{3}=50\text{ mi/h}

The harmonic mean is found as follows:
\begin{align}
\frac{1}{H}=\frac{1}{N}\sum{\frac{1}{X}}=\frac{1}{3}\left(\frac{1}{25}+\frac{1}{50}+\frac{1}{75}\right)=\frac{11}{450} && \text{ and } && H = \frac{450}{11}=40.9
\end{align}

The harmonic mean is the correct measure of the average velocity.
