# Summation notation
The symbol $\sum^N_{j=1}{X_j}$ is used to denote the sum of all the $X_j$'s from $j=1$ to $j=N$; by definition,

\begin{align}
\sum^N_{j=1}{X_j}=X_1+X_2+X_3+\ldots+X_N
\end{align}

When no confusion can result, we often denote this sum simply by $\sum{X}, \sum{X_j}, \sum_j{X_j}$. The symbol $\sum$ is the Greek capital letter _sigma_, denoting sum.

__EXAMPLE 1.__
\begin{align}
\sum^N_{j=1}{X_jY_j}=X_1Y_1+X_2Y_2+X_3Y_3+\ldots+X_NY_N
\end{align}
__EXAMPLE 2.__
\begin{align}
\sum^N_{j=1}{aX_j}=aX_1+aX_2+\ldots+aX_N=a(X_1+X_2+\ldots+X_N)=a\sum^N_{j=1}{X_j}
\end{align}  
where $a$ is a constant. More simply, $\sum{aX}=a\sum{X}$.  
__EXAMPLE 3.__
If $a, b,$ and $c$ are any constants, then $\sum{(aX+bY-cZ)}=a\sum{X}+b\sum{Y}-c\sum{Z}$. See [Problem 3.3](../02%20Solved%20Problems/01%20Summation%20Notation.ipynb#Problem-3-3).
