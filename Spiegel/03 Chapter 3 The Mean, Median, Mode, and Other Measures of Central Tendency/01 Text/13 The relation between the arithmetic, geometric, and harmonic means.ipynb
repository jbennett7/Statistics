# The relation between the arithmetic, geometric, and harmonic means
The geomtric mean of a set of positive numbers $X_1, X_2,\ldots, X_N$ is less than or equal to their arithmetic mean but is greater than or equal to their harmonic mean. In symbols,

<a id=eq14></a>
\begin{align}\label{eq14}\tag{14}
H\le G\le\bar{X}
\end{align}

The equality signs hold only if all the numbers $X_1, X_2,\ldots, X_N$ are identical.

__EXAMPLE 15.__ The set $2, 4, 8$ has arithmetic mean $4.67$, geomtric mean $4$, and harmonic mean $3.43$.
