# The weighted arithmetic mean
Sometimes we associate with the numbers $X_1, X_2,\ldots,X_K$ certain _weighting factors_ (or _weights_) $w_1, w_2,\ldots, w_K$, depending on the significance or importance attached to the numbers. In this case,

<a id=eq3></a>
\begin{align}\label{eq3}\tag{3}
\bar{X}=\frac{w_1X_1+w_2X_2+\ldots+w_KX_K}{w_1+w_2+\ldots+w_K}=\frac{\sum{wX}}{\sum{w}}
\end{align}

is called the _weighted arithmetic mean_. Note the similarity to equation ([<u>2</u>](04%20The%20arithmetic%20mean.ipynb#eq2)), which can be considered a weighted arithmetic mean with weights $f_1, f_2,\ldots, f_K$.

__EXAMPLE 6.__ If a final examination in a course is weighted 3 times as much as a quiz and a student has a final examination grade of $85$ and quiz grades of $70$ and $90$, the mean grade is

\begin{align}
\bar{X}=\frac{(1)(70)+(1)(90)+(3)(85)}{1+1+3}=\frac{415}{5}=83
\end{align}

---

```r
x <- c(70, 90, 85)
wx <- c(1, 1, 3)
weighted.mean(x, wx)
```
