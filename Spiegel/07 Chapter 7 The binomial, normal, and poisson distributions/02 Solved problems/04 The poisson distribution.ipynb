# The poisson distribution

## Problem 7.27
Ten percent of the tools produced in a certain manufacturing process turn out to be defective. Find the probability that in a sample of 10 tools chosen at random exactly 2 will be defective by using (a) the binomial distribution and (b) the Poisson approximation to the binomial distribution.

### SOLUTION

```r
p7.27 <- new.env()
p7.27$p <- 0.1
cat("The binomial distribution: ", dbinom(2, 10, p7.27$p), "\n")
p7.27$lambda <- 10*p7.27$p
cat("The poisson approximation: ", dpois(2, p7.27$lambda))
rm(p7.27)
```

## Problem 7.28
If the probability that an individual suffers a bad reaction from injection of a given serum is 0.001, determine the probability that out of 2000 individuals (a) exactly 3 and (b) more than 2 individuals will suffer a bad reaction.

### SOLUTION

```r
p7.28 <- new.env()
cat("(a) The binomial distribution: ", dbinom(3, 2000, 0.001), "\n")
cat("(a) The poisson approximation: ", dpois(3, 2000*0.001), "\n")
cat("\n\n(b)The binomial distribution: ", pbinom(2, 2000, 0.001), "\n")
cat("(b)The poisson approximation: ", ppois(2, 2000*0.001), "\n")
rm(p7.28)
```

## Problem 7.29
A poisson distribution is given by
\begin{align}
p(X) = \frac{(0.72)^X e^{-0.72}}{X!}
\end{align}

Find (a) p(0), (b) p(1), (c) p(2), and (d) p(3).

### SOLUTION

```r
p7.29 <- new.env()
p7.29$pd <- function(x) {
    for(idx in x) {
        cat("p(", idx, ") = ", (0.72^idx * exp(-0.72))/(factorial(idx)), "\n")
    }
}
p7.29$pd(c(0,1,2,3))
rm(p7.29)
```
