### Properties of the arithmetic mean

#### 3.16
Prove that the sum of the deviations of $X_1, X_2,...,X_N$ from their mean $\bar{X}$ is equal to zero.

#### Solution
Let $d_1 = X_1 - \bar{X}, d_2 = X_2 - \bar{X},...,d_N = X_N - \bar{X}$ be the deviations of
$X_1, X_2,...,X_N$ from their mean $\bar{X}$. Then

$$
\begin{align*}
\text{Sum of deviations}
= \sum{d_j}
= \sum{(X_j - \bar{X})}
= \sum{X_j} - N\bar{X}
= \sum{X_j} - N\left(\frac{\sum{X_j}}{N}\right)
= \sum{X_j} - \sum{X_j} = 0
\end{align*}
$$

where we have used $\sum$ in place of $\sum_{j=1}^N$. We could, if desired, have omitted the subscript $j$ in $X_j$, provided that it is _understood_.

#### 3.17
If $Z_1 = X_1 + Y_1, Z_2 = X_2 + Y_2,...,Z_N = X_N + Y_N$, prove that $\bar{Z} = \bar{X} + \bar{Y}$.

#### Solution
By definition,
$$
\begin{align*}
\bar{X} = \frac{\sum{X}}{N}
\ \ \ \ \ \ \ \ \ \ \ 
\bar{Y} = \frac{\sum{Y}}{N}
\ \ \ \ \ \ \ \ \ \ \ 
\bar{Z} = \frac{\sum{Z}}{N}
\end{align*}
$$
Thus
$$
\begin{align*}
\bar{Z}
= \frac{\sum{Z}}{N}
= \frac{\sum{(X+Y)}}{N}
= \frac{\sum{X}+\sum{Y}}{N}
= \frac{\sum{X}}{N}+\frac{\sum{Y}}{N}
= \bar{X} + \bar{Y}
\end{align*}
$$

where the subscripts $j$ on $X$, $Y$, and $Z$ have been omitted and where $\sum$ means $\sum_{j=1}^N$.

#### 3.18
(a) If $N$ numbers $X_1, X_2,...,X_N$ have deviations from any number $A$ given by $d_1 = X_1 - A$, 
$d_2 = X_2 - A,...,d_N = X_N - A$, respectively, prove that:

$$
\begin{align*}
\bar{X} = A + \frac{\sum_{j=1}^N{d_j}}{N} = A + \frac{\sum{d}}{N}
\end{align*}
$$


(b) In case $X_1, X_2,...,X_K$ have respective frequencies $f_1, f_2,...,f_K$ and 
$d_1 = X_1 - A,...,d_K = X_K - A$, show that the result in part (a) is replaced with

$$
\begin{align*}
\bar{X} = A + \frac{\sum_{j=1}^K{f_jd_j}}{\sum_{j=1}^K{f_j}} = A + \frac{\sum{fd}}{N}\text{   where   }
\sum_{j=1}^K{f_j} = \sum{f} = N
\end{align*}
$$



#### Solution (a) First method
$$
\begin{align*}
\bar{X} = A + \frac{\sum{X_j}}{N} = \frac{\sum(A+d_j)}{N} = \frac{\sum{A}+\sum{d_j}}{N} = \frac{NA+\sum{d_j}}{N}=
A+\frac{\sum{d_j}}{N}
\end{align*}
$$

where we have used $\sum$ in place of $\sum_{j=1}^N$ for brevity.

#### Solution (a) Second method
We have $d=X-A$, or $X=A+d$, omitting the subscripts on $d$ and $X$. Thus, by Problem 3.17,

$$
\begin{align*}
\bar{X} = \bar{A}+\bar{d} = A + \frac{\sum{d}}{N}
\end{align*}
$$

since the mean of a number of constants all equal to $A$ is $A$.

#### Solution (b)
$$
\begin{align*}
\bar{X}
= \frac{\sum_{j=1}^K{f_jX_j}}{\sum_{j=1}^K{f_j}}
= \frac{\sum{f_jX_j}}{N}
= \frac{\sum{f_j(A+d_j)}}{N}
= \frac{\sum{Af_j} + \sum{f_jd_j}}{N}
= \frac{A\sum{f_j} + \sum{f_jd_j}}{N}
= \frac{AN + \sum{f_jd_j}}{N}
= A + \frac{\sum{f_jd_j}}{N}
= A + \frac{\sum{fd}}{N}
\end{align*}
$$

Note that _formally_ the result is obtained from part (a) by replacing $d_j$ with $f_jd_j$ and summing from
$j = 1$ to $K$ instead of from $j = 1$ to $N$. The result is equialent to $\bar{X} = A + \bar{d}$, where
$\bar{d} = \left(\sum{fd}\right)/N$

#### 3.21
Let $d_j - X_j - A$ denote the deviations of any class mark $X_j$ in a frequency distribution from a given class mark $A$. Show that if all class intervals have equal _size_ $c$, then (a) the deviations are all multiples 0f $c$ (i.e., $d_j = cu_j$, where $u_j = 0$, $\pm1$, $\pm2$,...) and (b) the arithmetic mean can be computed from the formula

$$
\begin{align*}
\bar{X} = A + \left(\frac{\sum{fu}}{N}\right)c
\end{align*}
$$

#### Solution
(a) The result is illustrated in the following table, where it is observed that the deviations in column 2 are all multiples ofof the class-interval size $c = 3$ in.

To see that the result is true in general, note that if $X_1$, $X_2$, $X_3$,... are successive class marks, their common difference will for this case be equal to $c$, so that $X_2 = X_1 + c$, $X_3 = X_1 + 2c$, and in general $X_j = X_1 + (j-1)c$. Then any two class marks $X_p$ and $X_q$, for example, will differ by

$$
\begin{align*}
X_p - X_q = [X_1 + (p - 1)c]-[X_1 + (q - 1)c] = (p - q)c
\end{align*}
$$

which is a multiple of $c$.

(b) By part (a), the deviations of all the calss marks from any given one are multiples of $c$ (i.e., $d_j=cu_j$). Then, using Problem 3.18(b), we have

$$
\begin{align*}
\bar{X}
= A + \frac{\sum{f_jd_j}}{N}
= A + \frac{\sum{f_j(cu_j)}}{N}
= A + c\frac{\sum{f_ju_j}}{N}
= A + \left(\frac{\sum{fu}}{N}\right)c
\end{align*}
$$

Note that this is equivalent to the result $\bar{X} = A + c\bar{u}$, which can be obtained from $\bar{X} = A + \bar{d}$ by placing $d = cu$ and observing that $\bar{d} = c\bar{u}$ (see Problem 3.18)

```r
X <- c(61, 64, 67, 70, 73)
A <- X[3]
d <- c(-6, -3, 0, 3, 6)
f <- c(5, 18, 42, 27, 8)
fd <- f*d
df <- data.frame(X,d,f,fd)
df
N <- sum(df$f)
X.bar <- A + sum(df$fd)/N
X.bar
```

#### 3.31
Find the mean, median, and mode for the sets (a) $3$, $5$, $2$, $6$, $5$, $9$, $5$, $2$, $8$, $6$ and
(b) $51.6$, $48.7$, $50.3$, $49.5$, $48.9$.

```r
set.a <- c(3, 5, 2, 6, 5, 9, 5, 2, 8, 6)
set.b <- c(51.6, 48.7, 50.3, 49.5, 48.9)

mean.a.long <- sum(set.a)/length(set.a)
mean.a.quick <- mean(set.a)

median.a.long <- 0
sorted.set.a <- sort(set.a)
if(length(sorted.set.a)%%2 != 0){
    i <- length(sorted.set.a)/2
    median.a.long <- sorted.set.a[i]
} else {
    i <- (length(sorted.set.a)-1)/2
    median.a.long <- (sorted.set.a[i]+sorted.set.a[i+1])/2
}
median.a.quick <- median(set.a)

getmode <- function(v) {
    uniqv <- unique(v)
    uniqv[which.max(tabulate(match(v, uniqv)))]
}
mode.a <- getmode(set.a)
c(mean.a.quick, median.a.quick, mode.a)

mean.b <- mean(set.b)
median.b <- median(set.b)
# This is wrong
mode.b <- getmode(set.b)
c(mean.b, median.b, mode.b)
```

#### 3.36
The numbers $X_1$, $X_2$,...,$X_K$ occur with frequencies $f_1$, $f_2$,...,$f_K$, where $f_1 + f_2 + ... + f_K = N$ is the total frequency. \
\
(a) Find the geometric mean $G$ of the numbers. \
(b) Derive an expression for $\log{G}$. \
(c) How can the results be used to find the geometric mean for data grouped into a frequency distribution?

#### Solution
(a)
$$
\begin{align*}
G = \sqrt[N]{\underbrace{X_2X_2...X_2}_{f_2 \text{times}}\ \underbrace{X_2X_2...X_2}_{f_2 \text{times}}\ldots\ \underbrace{X_KX_K...X_K}_{f_K \text{times}}} = \sqrt[N]{X_1^{f_1}X_2^{f_2}\ldots X_K^{f_K}}
\end{align*}
$$
\
where $N = \sum{f}$. This is sometimes called the _weighted geometric mean_.

(b)
$$
\begin{align*}
\log{G}
= \frac{1}{N}\log{\left(X_1^{f_1}X_2^{f_2}\ldots X_K^{f_K}\right)}
= \frac{1}{N}\left(f_1\log{X_1} + f_2\log{X_2} + \ldots + f_K\log{X_K}\right)
= \frac{1}{N}\sum_{j=1}^Kf_j\log{X_j} = \frac{\sum{f}\log{X}}{N}
\end{align*}
$$
\
where we assume that all the numbers are positive; otherwise, the logarithms are not defined.\
Note that the logarithm of the geometric mean of a set of positive numbers is the arithmetic mean of the logarithms of the numbers.

(c)\
The result can be used to find the geometric mean for grouped data by taking $X_1$, $X_2$,$\ldots$,$X_K$ as class marks and $f_1$, $f_2$,$\ldots$, $f_K$ as the corresponding class frequencies.
