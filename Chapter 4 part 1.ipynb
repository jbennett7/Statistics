## Dispersion, or variation
The degree to which numerical data tend to spread about an average value is called the _dispersion_, or _variation_, of the data. Various measures of this dispersion (or variation) are available, the most common being the range, mean deviation, semi-interquartile reange, $10$&mdash;$90$ percentile range, and standard deviation.

---

## The Range
The _range_ of a set of numbers is the difference between the largest and smallest numbers in the set.

---

__EXAMPLE 1.__ The range of the set $2, 3, 3, 5, 5, 5, 8, 10, 12$ is $12-2=10$. Sometimes the range is given by simply quoting the smallest and largest numbers; in the above set, for instance, the range could be indicated as $2$ to $12$, or $2-12$.

---

__4.1.__ Find the range of the sets (a) 12, 6, 7, 3, 15, 10, 18, 5 and (b) 9, 3, 8, 8, 9, 8, 9, 18

### Solution
In both cases, range $=$ largest number $-$ smallest number $=18-3=15$. However, as seen from the arrays of sets (a) and (b),<p>
$$
\begin{align*}
\text{(a)}\ 3, 5, 6, 7, 10, 12, 15, 18\ \ \ \ \ \ \ \ \text{(b)}\ 3, 8, 8, 9, 9, 9, 18
\end{align*}
$$<p>
there is much more variation or dispersion, in (a) than in (b). In fact, (b) consists mainly of $8$'s and $9$'s.<p>
Since the range indicates no difference between the sets, it is not a very good measure of dpsersion in this case. Where extreme values are present, the range is generally a poor measure of dispersion.<p>
An improvement is achieved by throwing out the extreme cases, $3$, and $18$. Then for set (a) the range is $(15 - 5) = 10$, while for set (b) the range is $(9-8)=1$, clearly showing that (a) has greater dispersion than (b). However, this is not the way the range is defined. The semi-interquartile range and the 10&mdash;90 percentile range were designed to improve on the range by eliminating extreme cases.

```r
p1.a <- c(12, 6, 7, 3, 15, 10, 18, 5)
p1.b <- c(9, 3, 8, 8, 9, 8, 9, 18)
(p1.a.range <- diff(range(p1.a)))
(p1.b.range <- diff(range(p1.b)))
```

## The mean deviation
The _mean deviation_, or _average deviation_, of a set of $N$ numbers $X_1, X_2,\ldots,X_N$ is abbreviaged $\text{MD}$ and is defined.<p>
\begin{align}
\text{MD}
= \frac{\sum^N_{j=1}{\lvert X_j-\bar{X}\rvert}}{N}
= \frac{\sum{\lvert X-\bar{X}\rvert}}{N}
= \frac{\sum{|X - \bar{X}|}}{N}
= \overline{\lvert X-\bar{X}\rvert}\label{eq1}\tag{1}
\end{align}<P>
where $\bar{X}$ is the arithmetic mean of the numbers and $\lvert X_j-\bar{X}\rvert$ is the absolute value of the deviation of $X_j$ from $\bar{X}$. (The _absolute value_ of a number is the number without the associated sign and is indicated by two vertical lines placed around the number; thus $\lvert-4\rvert=4,\lvert+3\rvert=3,\lvert6\rvert=6,$ and $\vert-0.84\rvert=0.84$.)<p>
__EXAMPLE 2.__  Find the mean deviation of the set $2, 3, 6, 8, 11$.<p>
\begin{align}
\text{Arithmetic mean (}\bar{X}\text{)}=\frac{2+3+6+8+11}{5}=6
\end{align}<p>
\begin{align}
\text{MD}
=\frac{\lvert2-6\rvert+\lvert3-6\rvert+\lvert6-6\rvert+\lvert8-6\rvert+\lvert11-6\rvert}{5}
=\frac{\lvert-4\rvert+\lvert-3\rvert+\lvert0\rvert+\lvert2\rvert+\lvert5\rvert}{5}
=\frac{4+3+0+2+5}{5}=2.8
\end{align}<p>
If $X_1, X_2,\ldots,X_K$ occur with frequencies $f_1,f_2,\ldots,f_K$, respectively, the mean deviation can be written as<p>
\begin{align}
\text{MD}
=\frac{\sum^K_{j=1}{f_j\lvert X_j-\bar{X}\rvert}}{N}
=\frac{\sum{f\lvert X-\bar{X}\rvert}}{N}
=\overline{\lvert X-\bar{X}\rvert}\label{eq2}\tag{2}
\end{align}<p>
where $N=\sum^K_{j=1}{f_j}=\sum{f}$. This form is useful for grouped data, where the $X_j$'s represent class marks and the $f_j$'s are the corresponding class frequencies.<p>
Occasionally the mean deviation is defined in terms of absolute deviations from the median or other average instead of from the mean. An interesting property of the sum $\sum^N_{j=1}\lvert X_j-a\rvert$ is that it is a minimum where $a$ is the median (i.e., the mean deviation about the median is a minimum).<p>
Note that it would be more appropriate to use the terminology _mean absolute deviation_ than _mean deviation_.

Below is a function that can be used to find the mean deviation. It is also found in the saved data file `Saves/mean_deviation.Rdata`.

```r
md <- function(x) {
    return (sum(abs(x - mean(x)))/length(x))
}
```

## The Semi-Interquartile Range
The _semi-interquartile range_, or _quartile deviation_, of a set of data is denoted by $Q$ and is defined by<p>
\begin{align}
Q = \frac{Q_3 - Q_1}{2}\label{eq3}\tag{3}
\end{align}<p>
where $Q_1$ and $Q_3$ are the first and third quartiles for the data (see Problems 4.6 and 4.7). The interquartile range $Q_3-Q_1$ is sometimes used, but the semi-interquartile range is more common as a measure of dispersion.

---

__4.6.__ Find the semi-interquartile range for the height distribution of the students at XYZ University.

```r
Table_2.1
```

### Solution
The lower and upper quartiles are $Q_1=65.5+\frac{2}{42}(3)=65.64$in and $Q_3=68.5+\frac{10}{27}(3)=69.61$in, respectively, and the semi-interquartile range (or quartile deviation) is $Q=\frac{1}{2}(Q_3-Q_1)=\frac{1}{2}(69.61-65.64)=1.98$in. Note that $50\%$ of the cases lie between $Q_1$ and $Q_3$ (i.e., $50$ students have heights between $65.64$ and $69.61$in).<p>
We can consider $\frac{1}{2}(Q_1+Q_3)=67.63$in to be a measure of central tendency (i.e., average height). It follwos that $50\%$ of the heights lie in the range $67.63\pm1.98$in.

```r
p6.ci <- class.interval <- 3
p6.boundaries <- seq(59.5, 74.5, by=class.interval)
p6.cumsum <- cumsum(as.vector(Table_2.1))
p6.f.sum <- sum(Table_2.1)
p6.Q.breaks <- p6.f.sum/4

p6.Q1.index <- match(Table_2.1[p6.cumsum > p6.Q.breaks][1],Table_2.1)
p6.Q1.close.class <- p6.boundaries[p6.Q1.index]
p6.Q1.class <- as.vector(Table_2.1)[p6.Q1.index]
p6.Q1.remainder <- p6.Q.breaks - p6.cumsum[p6.Q1.index-1]
p6.Q1 <- round(p6.Q1.close.class + p6.Q1.remainder/p6.Q1.class*p6.ci,2)

p6.Q3.index <- match(Table_2.1[p6.cumsum > 3*p6.Q.breaks][1], Table_2.1)
p6.Q3.close.class <- p6.boundaries[p6.Q3.index]
p6.Q3.class <- as.vector(Table_2.1)[p6.Q3.index]
p6.Q3.remainder <- 3*p6.Q.breaks - p6.cumsum[p6.Q3.index-1]
p6.Q3 <- round(p6.Q3.close.class + p6.Q3.remainder/p6.Q3.class*p6.ci,2)

c((p6.Q.central <- round((p6.Q1 + p6.Q3)/2,2)), (p6.Q.pm <- round((p6.Q3 - p6.Q1)/2,2)))
(p6.Q.range <- c(p6.Q.central-p6.Q.pm, p6.Q.central+p6.Q.pm))
```

## The 10-90 Percentile Range
The $10$&mdash;$90$ _percentile range_ of a set of data is defined by<p>
\begin{align}
10-90\ \text{percentile range}=P_{90}-P_{10}\label{eq4}\tag{4}
\end{align}<p>
where $P_{10}$ and $P_{90}$ are the $10$th and $90$th percentiles for the data (see Problem 4.8). The semi-$10$&mdash;$90$ percentile range, $\frac{1}{2}(P_{90}-P_{10})$, can also be used but is not commonly employed.

---

__4.8.__ Find the $10$&mdash;$90$ percentile range of the heights of the students at XYZ University (Table 2.1).

```r
Table_2.1
```

### Solution
Here $P_{10}=62.5+\frac{5}{18}(3)=63.33$in, and $P_{90}=68.5+\frac{25}{27}(3)=71.27in$. Thus the $10$&mdash;$90$ percentile range is $P_{90}-P_{10}=71.27-63.33=7.94$in. Since \frac{1}{2}(P_{10}+P_{90})=67.30$in and \frac{1}{2}(P_{90}-P_{10})=3.97$in, we can conclude that $80\%$ of the students have heights in the range $67.30\pm3.97$in.

```r
p8.ci <- 3
p8.boundaries <- seq(59.5, 74.5, by=3)
p8.D.cumsum <- cumsum(as.vector(Table_2.1))
p8.f.sum <- sum(Table_2.1)
p8.D.breaks <- p8.f.sum/10

p8.D.10.index <- match(TRUE, p8.D.cumsum > p8.D.breaks)[1]
p8.D.10.close.class <- p8.boundaries[p8.D.10.index]
p8.D.10.class <- as.vector(Table_2.1)[p8.D.10.index]
p8.D.10.remainder <- p8.D.breaks - p8.D.cumsum[p8.D.10.index-1]
p8.D.10 <- round(p8.D.10.close.class + p8.D.10.remainder/p8.D.10.class*p8.ci,2)

p8.D.90.index <- match(TRUE, p8.D.cumsum > 9*p8.D.breaks)[1]
p8.D.90.close.class <- p8.boundaries[p8.D.90.index]
p8.D.90.class <- as.vector(Table_2.1)[p8.D.90.index]
p8.D.90.remainder <- 9*p8.D.breaks - p8.D.cumsum[p8.D.90.index-1]
p8.D.90 <- round(p8.D.90.close.class + p8.D.90.remainder/p8.D.90.class*p8.ci,2)

p8.D.90 - p8.D.10
c((p8.D.central <- round((p8.D.10 + p8.D.90)/2,2)), (p8.D.pm <- round((p8.D.90 - p8.D.10)/2,2)))
(p8.D.10.90.range <- c(p8.D.central - p8.D.pm, p8.D.central + p8.D.pm))
```
