Table 2.6 from solved problem 2.3

```r
breaks <- seq(25000, 32000, by=1000)
f <- c(8, 10, 16, 14, 10, 5, 2)
s1 <- c()
for(i in 1:length(f)){
    s2 <- sample(breaks[i]:(breaks[i+1]-1), f[i], replace=TRUE)
    s1 <- sample(c(s1, s2))
    rm(s2)
}
s1 <- s1/100
breaks <- breaks/100
table(cut(s1, breaks=breaks, right=FALSE))
```
