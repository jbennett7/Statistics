# Histograms and Frequency Polygons
Histograms and frequency polygons are two graphic representations of frequency distributions.

1. A _histogram_ or _frequency histogram_, consists of a set of rectangles having (a) bases on a horizontal axis (the X axis), with centers at the class marks and lengths equal to the class interval sizes, and (b) areas proportinal to the class frequencies. It is one of the oldest constructs in use which shows distributions.
2. A _frequency polygon_ is a line graph of the class freqencies plotted against class marks. It can be obtained by connecting the midpoints of the tops of the rectangles in the histogram.

The histogram shows the researcher how the data is distributed. In the case for `Table 2.2`, iPhone hours has a distribution we call the _normal distribution_. Most of the data is near the middle. There are a few values that are down in the $10$ and $11$ range and a few that are up in the $22$ and $23$ range. Most of the data is concentrated in the middle.

---

## Histogram and Frequency Polygon from raw data

```r
load("../../Saves/Table_2.2.Rdata")
h <- hist(Table_2.2, breaks=14, right=FALSE)
marks <- seq(10.5, 22.5, by=1)
h
plot(marks, h$counts, type="l", col="red", xlab="iPhone hours", ylab="Frequency", main="Polygon iPhone hours")
points(marks, h$counts, pch=20)
```

## Histogram and Frequency Polygon from a frequency distribution

```r
load('../../Saves//Table_2.1.Rdata')
counts <- as.vector(Table_2.1)
breaks <- seq(59.5, 74.5, by=3)
marks <- seq(61, 73, by=3)
density <- counts/diff(breaks)
xname <- "Height (in)"
hist <- list(breaks=breaks, counts=counts, density=density, xname=xname)
class(hist) <- "histogram"
hist
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8)
plot(hist)
plot(marks, counts, type="l", col="red", xlab="Height (in)", ylab="Frequency", main="Polygon of Height (in)")
points(marks, counts, pch=20)
options(repr.plot.width=7, repr.plot.height=7)
```
