__2.1.__  
(a) Arrange the numbers $17, 45, 38, 27, 6, 48, 11, 57, 34,$ and $22$ in an array.  
(b) Determine the range of these numbers.

### Solution
(a) In ascending order of magnitude, the array is: $6, 11, 17, 22, 27, 34, 38, 48, 57$. In descending order of magnitude, the array is: $57, 48, 38, 34, 27, 22, 17, 11, 6$.  
(b) Since the smallest number is $6$ and the largest number is $57$, the range is $57-6=51$.

```r
p1.data <- c( 17, 38, 27, 6, 48, 11, 57, 34, 22)
(p1.array <- sort(p1.data))
(p1.rev.array <- rev(p1.array))
(p1.range <- diff(range(p1.array)))
```

__2.2__ The final grades in mathematics of $80$ students at State University are recorded in the accompanying table.

```r
p2.data   <- c( 68, 84, 75, 82, 68, 90, 62, 88, 76, 93,
             73, 79, 88, 73, 60, 93, 71, 59, 85, 75,
             61, 65, 75, 87, 74, 62, 95, 78, 63, 72,
             66, 78, 82, 75, 94, 77, 69, 74, 68, 60,
             96, 78, 89, 61, 75, 95, 60, 79, 83, 71,
             79, 62, 67, 97, 78, 85, 76, 65, 71, 75,
             65, 80, 73, 57, 88, 78, 62, 76, 53, 74,
             86, 67, 73, 81, 72, 63, 76, 75, 85, 77)
```

With reference to this table, find:  
(a) The highest grade.  
(b) The lowest grade.  
(c) The range.  
(d) The grades of the five highest-ranking students.  
(e) The grades of the five lowest-ranking students.  
(f) The grade of the student ranking tenth highest.  
(g) The number of students who received grades of 75 or higher.  
(h) The number of students who received grades belwo 85.  
(i) The percentage of students who received grades higher than $65$ but not higher than $85$.  
(j) The grades that did not appear at all.

### Solution
Some of these questions are so detailed that they are best answered by first constructing an array. This can be done by subdividing the data into convenient classes and placing each number taken from the table into the appropriate class, as in Table 2.4, called an _entry table_. By then arranging the numbers of each class into an array, as in Table 2.5, the required array is obtained. From Table 2.5 it is relatively easy to answer the above questions.

```r
# The following will construct table 2.4 and 2.5 from page 44 of the text.
p2.c <- 5
p2.range <- range(p2.data)
lower.limit <- floor(p2.range[1]/10)*10
upper.limit <- ceiling(p2.range[2]/10)*10
breaks <- seq(lower.limit, upper.limit, by=p2.c)
tags <- c("50-54", "55-59", "60-64", "65-69", "70-74", "75-79", "80-84", "85-89", "90-94", "95-99")
group_tags <- cut(p2.data, breaks=breaks, include.lowest=TRUE, right=FALSE, labels=tags)
for(tag in tags){
    cat(tag, "   ", p2.data[group_tags == tag], "\n")
}
cat("\n")
for(tag in tags){
    cat(tag, "   ", sort(p2.data[group_tags == tag]), "\n")
}
```

```r
# (a) The highest grade.
max(p2.data)
# (b) The lowest grade.
min(p2.data)
# (c) The range.
diff(range(p2.data))
# (d) The grades of the five highest-ranking students.
tail(sort(p2.data),5)
# (e) The grades of the five lowest-ranking students.
head(sort(p2.data),5)
# (f)  The grade of the student ranking tenth highest.
sort(p2.data, decreasing=TRUE)[10]
# (g) The number of students who received grades of 75 or higher.
sum(p2.data >= 75)
# (h) The number of students who received grades below 85.
sum(p2.data < 85)
# (i) The percentage of students who received grades higher than 65 but not higher than 85.
sum(p2.data > 65 & p2.data <= 85)/length(p2.data)*100
# (j) The grades that did not appear at all.
setdiff(0:100, p2.data)
```
