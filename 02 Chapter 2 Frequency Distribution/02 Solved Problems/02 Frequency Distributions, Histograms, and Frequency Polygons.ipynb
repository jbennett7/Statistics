(c) The class mark (or class midpoint) of the third class.  
(d) The class boundaries of the fifth class.  
(e) The size of the fifth-class interval.  
(f) The frequency of the third class.  
(g) The relative frequency of the third class.  
(h) The class interval having the largest frequency. This is sometimes called the _modal class interval_; its frequency is then called the _modal class frequency_.  
(i) The percentage of employees earning less that $\$280.00$ per week.  
(j) The percentage of employees earning less than $\$300.00$ per week but at least $\$260.00$ per week.

__2.4__ If the class marks in a frequency distribution of the weights of students are $128, 137, 146, 155, 164, 173,$ and $182$ pounds (lb), find (a) the class-interval size, (b) the class boundaries, and (c) the class limits, assuming that the weights were measured to the nearest pound.

### Solution
(a) Class-interval size = common difference between successive class markes = $137-128=146-137=etc.=9$lb.  
(b) Since the class intervals all have equal size, the class boundaries are midway between the class marks and thus have the values

\begin{align}
\frac{1}{2}(128+137), \frac{1}{2}(137+146), \ldots, \frac{1}{2}(173+182) \ \text{ or }\ 132.5, 141.5, 150.5, \ldots, 177.5\text{ lb}
\end{align}

The first class boundary is $132.5-9=123.5$ and the last class boundary is $177.5+9=186.5$, since the common class interval size is $9$lb. Thus all the class boundaries are given by

\begin{align}
123.5; 132; 141.5; 150.5; 159.5; 168.5; 177.5; 186.5 \text{ lb}
\end{align}

(c) Since the class limits are integers, we choose them as the integers nearest to the class boundaries, namely, $123, 132, 133, 141, 142, \ldots$ Thus the first class has limits $124$&mdash;$132$, the next $133$&mdash;$141$, etc.

```r
(p4.X <- c( 128, 137, 146, 155, 164, 173, 182))
# (a) class interval size
(p4.c <- diff(p4.X)[1])
# (b) the class boundaries
(p4.boundaries <- unique(sort(c(p4.X - p4.c/2, p4.X + p4.c/2))))
# (c) the class limits
(p4.lower.limits <- p4.X - floor(p4.c/2))
(p4.upper.limits <- p4.X + floor(p4.c/2))
```
