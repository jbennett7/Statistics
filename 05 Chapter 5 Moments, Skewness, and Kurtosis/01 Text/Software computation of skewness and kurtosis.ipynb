# Software computation of skewness and kurtosis
The software that we have discussed in the text so far may be used to compute skewness and kurtosis measures for sample data. The data in [Table 5.1](#Table-5-1) samples of size $50$ from a normal distriution, a skewed-right distribution, a skewed-left distribution, and a uniform distribution.

The normal data are female height measurements, the skewed-right data are age at marriage for females, the skewed-left data are obituary data that give the age at death for females, and the uniform data are the amount of cola put into a $12$ ounce container by a soft drinks machine. The distribution of the four sets of sample data is shown in [Fig. 5-1](#Figure-5-1). The distributions of the four samples are illustrated by dotplots.

```r
if (! "moments" %in% as.vector(installed.packages()[,1])){
    install.packages("moments")
    library("moments")
} else {
    library("moments")
}
```

```r
range.normal <- range(normal)
breaks.normal <- seq(range.normal[1], range.normal[2], by=1)
range.skewed.right <- range(skewed.right)
breaks.skewed.right <- seq(range.skewed.right[1], range.skewed.right[2], by=1)
range.skewed.left <- range(skewed.left)
breaks.skewed.left <- seq(range.skewed.left[1], range.skewed.left[2], by=1)
range.uniform <- range(uniform)
breaks.uniform <- seq(range.uniform[1], range.uniform[2], by=.1)
```

The variable Height is the height of $50$ audlt female, the variable Wedage is the age at the wedding of $50$ females, the variable Obitage is the age at death of $50$ females, and the variable Cola-fill is the amount of cola put into $12$ ounce containers. Using the terminology that we have learned in this chapter: the height distribution is mesokurtic, the cola-fill distribution is platykurtic, the wedage distribution is skewed to the right, and the obitage distribution is skewed to the left.

#### Example 5-2
Use EXCEL to find the skewness and kurtosis values for the data in [Fig. 5-1](#Figure-5-1). If the variable names are entered into `A1:D1` and sample data are entered into `A2:D51` and any open sell is used to enter `=SKEW(A2:A51)` the value $-0.0203$ is returned. The function `=SKEW(B2:B51)` returns $1.9774$, the function `=SKEW(C2:C51)` returns $-1.4986$, and `=SKEW(D2:D51)` returns $0.0156$. It can be seen that MINITAB and EXCEL give the same values for kurtosis and skewness.

#### R Programming using the moments library

```r
normal.descriptive.statistics <- c(length(normal), mean(normal), round(sd(normal), 3), round(skewness(normal),2), round(kurtosis(normal),2))
skewed.right.descriptive.statistics <- c(length(skewed.right), mean(skewed.right), round(sd(skewed.right), 3), round(skewness(skewed.right), 2), round(kurtosis(skewed.right), 2))
skewed.left.descriptive.statistics <- c(length(skewed.left), mean(skewed.left), round(sd(skewed.left), 3), round(skewness(skewed.left), 2), round(kurtosis(skewed.left), 2))
uniform.descriptive.statistics <- c(length(uniform), mean(uniform), round(sd(uniform), 3), round(skewness(uniform),2), round(kurtosis(uniform), 2))
```

```r
m <- t(matrix( c(normal.descriptive.statistics, skewed.right.descriptive.statistics, skewed.left.descriptive.statistics, uniform.descriptive.statistics), ncol=4))
rownames(m) <- c("Height", "Wedage", "Obitage", "Cola-fill")
colnames(m) <- c("N", "Mean", "StDev", "Skewness", "Kurtosis")
m
```
