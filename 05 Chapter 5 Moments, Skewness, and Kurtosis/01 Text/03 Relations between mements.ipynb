# Relations between moments
The following relations exist between moments about the mean $m_r$ andmoments about an arbitrary origin $m^\prime_r$:

\begin{align}
m_2=m^\prime_2-m^{\prime2}_1
\end{align}
\begin{align}
m_3=m^\prime_3-3m^\prime_1m^\prime_2+2m^{\prime3}_1\label{eq:moment_relations}\tag{7}
\end{align}
\begin{align}
m_4=m^\prime_4-4m^\prime_4m^\prime_3+6m^{\prime2}_1m^\prime_2-3m^{\prime4}_1
\end{align}

etc. (see Problem 5.5). Note that $m^\prime_1=\bar{X}-A$.
