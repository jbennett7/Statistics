# Skewness

## Problem 5-10
Find Pearson's (a) first and (b) second coefficients of skewness for the wage distribution of the 65 employees at the P&R Company (see Problems [3.44](#Problem-3-4) and [4.18](#Problem-4-18)).

### Solution 5-10
Mean = $\$279.76$, median = $\$279.06$, mode = $\$277.50$, and standard deviation $s = \$15.60$. THus:

(a)
\begin{align}
\text{First coefficient of skewness} = \frac{\text{mean}-\text{mode}}{s} = \frac{\$279.76 - \$277.50}{\$15.60} = 0.1448 \text{, or } 0.14
\end{align}
(b)
\begin{align}
\text{Second coefficient of skewness} = \frac{3(\text{mean} - \text{median})}{s} = \frac{3(\$279.76 - \$279.06)}{\$15.60} = 0.1346\text{, or }0.13
\end{align}

If the corrected standard deviation is usd \[see Problem [4.21(b)](#Problem-4-21-b), these coefficients become, respectively:

(a)
\begin{align}
\frac{\text{Mean} - \text{mode}}{\text{Corrected }s} = \frac{\$279.76 - \$277.50}{\$15.33} = 0.1474\text{, or }0.15
\end{align}
(b)
\begin{align}
\frac{3(\text{mean} - \text{median})}{\text{Corrected }s} = \frac{3(\$279.76 - \$279.06)}{\$15.33} = 0.1370\text{, or }0.14
\end{align}

Since the coefficients are positive, the distribution is skewed positively (i.e., to the right).
