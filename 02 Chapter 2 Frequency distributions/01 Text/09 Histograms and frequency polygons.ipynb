# Histograms and frequency polygons
Histograms and frequency polygons are two graphic representations of frequency distributions.

1. A _histogram_ or _frequency histogram_, consists of a set of rectangles having (a) bases on a horizontal axis (the X axis), with centers at the class marks and lengths equal to the class interval sizes, and (b) areas proportinal to the class frequencies. It is one of the oldest constructs in use which shows distributions.
2. A _frequency polygon_ is a line graph of the class freqencies plotted against class marks. It can be obtained by connecting the midpoints of the tops of the rectangles in the histogram. This information is represented graphically in [Fig. 2-9](#Figure_2_9).

The histogram and frequency polygon corresponding to the frequency distribution of heights in [Table 2.1](03%20frequency%20distributions.ipynb#Table_2_1) are shown in [Figs. 2-1](#Figure_2_1) and [2.9](#Figure_2_9).

Notice in [Fig. 2.9](#Figure_2_9) how the frequency polygon is tied down at the ends, that is, at $58$ and $76$.

#### Figure 2-1
<a id="Figure_2_1"></a>

```r
load("../../Saves//Table_2.1.Rdata")
Freq <- as.vector(Table.2.1$Freq)
hist <- list(breaks=seq(60, 75, by=3), counts=Freq, density=Freq/3, xname="Height (in)")
class(hist) <- "histogram"
plot(hist)
```

#### Figure 2.9
<a id="Figure_2_9"></a>

```r
marks <- seq(58, 76, by=3)
plot(marks, c(0, hist$counts, 0), type="l", col="red", xlab="iPhone hours", ylab="Frequency", main="Polygon iPhone hours")
points(marks, c(0, hist$counts, 0), pch=20)
```
