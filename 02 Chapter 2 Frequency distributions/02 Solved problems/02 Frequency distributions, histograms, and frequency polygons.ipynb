## Problem 2-4
If the class marks in a frequency distribution of the weights of students are $128, 137, 146, 155, 164, 173,$ and $182$ pounds (lb), find (a) the class-interval size, (b) the class boundaries, and (c) the class limits, assuming that the weights were measured to the nearest pound.

```r
cat("Class Marks: ", (X <- c( 128, 137, 146, 155, 164, 173, 182)))
```

### Solution 2-4
(a) Class-interval size = common difference between successive class markes = $137-128=146-137=etc.=9$lb.  
(b) Since the class intervals all have equal size, the class boundaries are midway between the class marks and thus have the values

\begin{align}
\frac{1}{2}(128+137), \frac{1}{2}(137+146), \ldots, \frac{1}{2}(173+182) \ \text{ or }\ 132.5, 141.5, 150.5, \ldots, 177.5\text{ lb}
\end{align}

The first class boundary is $132.5-9=123.5$ and the last class boundary is $177.5+9=186.5$, since the common class interval size is $9$lb. Thus all the class boundaries are given by

\begin{align}
123.5; 132; 141.5; 150.5; 159.5; 168.5; 177.5; 186.5 \text{ lb}
\end{align}

(c) Since the class limits are integers, we choose them as the integers nearest to the class boundaries, namely, $123, 132, 133, 141, 142, \ldots$ Thus the first class has limits $124$&mdash;$132$, the next $133$&mdash;$141$, etc.

```r
cat("(a) Class interval size: ", (c <- diff(X)[1]), "\n")
cat("(b) The class boundaries:", (boundaries <- unique(sort(c(X - c/2, X + c/2)))), "\n")
lower.limits <- X - floor(c/2);upper.limits <- X + floor(c/2)
cat("(c) The class limits:     ")
for( i in 1:length(lower.limits)) {
    if(i==length(lower.limits)){
        cat(lower.limits[i], "-", upper.limits[i], "\n", sep="")
    } else {
        cat(lower.limits[i], "-", upper.limits[i], ", ", sep="")
    }
}
```

## Problem 2-7
In answering [Problem 2.6(a)](#problem-2-6), a student chose the class intervals $5.10$&ndash;$5.40, 5.40$&ndash;$5.70,\ldots, 6.90$&ndash;$7.20, and 7.20$&ndash;$7.50$. Was there anything wrong with this choice?

### Solution 2-7
These class intervals overlap at $5.40, 5.70,\ldots, 7.20$. Thus a measurement recorded as $5.40$, for example, could be placed in either of the first two class intervals. Some statisticians justify this cchoice by agreeing to place half of such ambiguous cases in one class and half in the other.

The ambiguity is removed by writing the class intervals as $5.10$ to under $5.40$, $5.40$ to under $5.70$, etc. In this case, the class limits coincide with the class boundaries, and the class marks can coincide with the observed data.

In general it is desirable to avoid overlapping class intervals whenever possible and to choose then so that the class boundaries are values not coinciding with actual observed data. For example, the class intervals for [Problem 2.6](#problem-2-6) could have been chosen as $5.095$&ndash;$5.395, 5.395$&ndash;$5.695$, etc., without ambiguity. A disadvantage of this particular choice is that the class marks do not coincide with the observed data.

### Solution 2-8
The largest weight is $176$ lb and the smallest weight is $119$ lb, so that the range is $176$&ndash;$119=57$ lb. If five class intervals are used, the calss-interval size is $57/5=11$ approximately; if $20$ class intervals are used, the class-interval size is $57/20=3$ approximately.

One convenient choice for the class-interval size is $5$ lb. Also, it is convenient to choose the class marks as $120, 125, 130, 135,\ldots$ lb. Thus the class intervals can be taken as $118$&ndash;$122, 123$&ndash;$127, 128$&ndash;$132,\ldots$ With this choice the class boundaries are $117.5, 122.5, 127.5,\ldots$, which do not coincide with the observed data.

The required frequency distribution is shown in [Table 2.7](#Table-2-7). The center column, called a _tally_, or _score sheet_, is used to tabulate the class frequencies from the raw data and is usually omitted in the final presentation of the frequency distribution. It is unnecessary to make an array, although if it is available it can be used in tabulating the frequencies.

## Problem 2-12
At the P&R Company ([Problem 2.3](#Problem-2-3)), five new employees were hired at weekly wages of $\$285.34, \$316.83, \$335.78, \$356.21,$ and $\$374.50$. Construct a frequency distribution of wages for the $70$ employees.

### Solution 2-12
Possible frequency distributions are shown in [Table 2.10](#Table-2-10).

In [Table 2.10(a)](#Table-2-10-a), the same class-interval size, $\$10.00$, has been maintained throughout the table. As a result, there are too many empty classes and the detail is much too fine at the upper end of the wage scale.

In [Table 2.10(b)](#Table-2-10-b), empty classes and fine detail have been avoided by the use of the open class interval "$\$320.00$ and over." A disadvantae of this is that the table becomes useless in performing certain mathematical calculations. For example, it is impossible to determine the total maount of wages paid per week, since "over $\$320.00$" might conceiveably imply that individuals could earn as high as $\$1400.00$ per week.

In [Table 2.10(c)](#Table-2-10-c), a class-interval size of $\$20.00$ has been used. A disadvantage is that much information is lost at the lower end of the wage scale and the detail is still fine at the upper end of the scale.

In [Table 2.10(d)](#Table-2-10-d), unequal class-interval sizes have been used. A disadvantage is that certain mathematical calculations to be made later lose a simplicity that is available when class intervals have the same size. Also, the larger the class-interval size, the greater will be the grouping error.
