## Problem 2-14
Construct (a) a cumulative-frequency distribution, (b) a percentage cumulative distribution, (c) an ogive, and (d) a percentage ogive from the frequency distribution in [Table 2.6](./02%20Frequency%20Distributions,%20Histograms,%20and%20Frequency%20Polygons.ipynb#Table-2-6) of [Problem 2.3](./02%20Frequency%20Distributions,%20Histograms,%20and%20Frequency%20Polygons.ipynb#Problem-2-3)

### Solution 2-14
(a) and (b) The cumulative-frequency distribution and percentage cumulative distribution (or cumulative relative-frequency distribution) are shown in [Table 2.11](#Table-2-11).

Note that each entry in column 2 is obtained by adding sucessive entries from column 2 of [Table 2.6](./02%20Frequency%20Distributions,%20Histograms,%20and%20Frequency%20Polygons.ipynb#Table-2-6), thus, $18=8+10, 34=8+10+16$, etc.

Each entry in column 3 is obtained from the previous column by dividing by $65$, the total frequency, and expressing the result as apercentage. Thus, $34/65=52.3\%$. Entries in this column can also be obtained by adding successive entries from column 2 of [Table 2.9](./02%20Frequency%20Distributions,%20Histograms,%20and%20Frequency%20Polygons.ipynb#Table-2-9). THus, $27.7=12.3+15.4, 52.3=12.3+15.4+24.6$, etc.

(c) and (d) The ogive (or cumulative-frequency polygon) is shown in [Fig. 2-16(a)](#Figure-2-16-a) and the percentage ogive (relative cumulative-frequency polygon) is shown in [Fig. 2-16(b)](#Figure-2-16-b). Both of these are MINTAB-generated plots.

```r
Table.2.11 <- matrix(c(0, cumsum(Table.2.6)), nrow=length(Table.2.6)+1)
rownames(Table.2.11) <- c("Less than $250.00", "Less than $260.00", "Less than $270.00", "Less than $280.00", "Less than $290.00",
                            "Less than $300.00", "Less than $310.00", "Less than $320.00")
Table.2.11 <- cbind(Table.2.11, c(0, round(cumsum(Table.2.6)/sum(Table.2.6)*100,1)))
colnames(Table.2.11) <- c("Cumulative Frequency", "Percentage Cumulative Distribution")
```

## Problem 2-16
From the ogives in Figs. [2-16](#Figure-2-16) and [2-17](#Figure-2-17) (of Problems [2.14](#Problem-2-14) and [2.15](#Problem-2-15), respectively), estimate the number of employees earning (a) less than $\$288.00$ per week, (b) $\$296.00$ or more per week, and (c) at least $\$263.00$ per week but less than $\$275.00$ per week

### Solution 2-16
(a) Referring to the "less than" ogive of [Fig. 2-16](#Figure-2-16), construct a vertical line intersecting the "Wages" axis at $\$288.00$. This line meets the ogive at the point with coordinates $(288, 45)$; hence $45$ employees earn less than $\$288.00$ per week.

(b) In the "or more" ogive of [Fig. 2-17](#Figure-2-17), construct a vertical line at $\$296.00$. This line meets the ogive at the point $(296, 11)$; hence $11$ employees earn $\$296.00$ or more per week.

This can also be obtained from the "less than" ogive of [Fig. 2-16](#Figure-2-16). By constructing a line at $\$296.00$, we find that $54$ employees earn less than $\$296.00$ per week; hence $65-54=11$ employees earn $\$296.00$ or more per week.

(c) Using the "less than" ogive of [Fig. 2-16](#Figure-2-16), we have: $\text{Required number of employees}=\text{number earning less than }\$275.0\text{ per week}-\text{number earning less than }\$263.00\text{ per week}=26-11=15$.

Note that the above results could just as well have been obtained by the process of _interpolation_ in the cumulative-frequency tables. In part (a), for example, since $\$288.00$ is $8/10$, or $4/5$, of the way between $\$280.00$ and $\$290.00$, the required number of employees should be $4/5$ of the way between the corresponding values $34$ and $48$ (see [Table 2.11](#Table-2-11). But $4/5$ of the way between $34$ and $48$ is $\frac{4}{5}(48-34)=11$. Thus the required number of employees is $34+11=45$.

## Problem 2-17
Five pennies were tossed $1000$ times, and at each toss the number of heads was observed. The number of tosses during which $0, 1, 2, 3, 4,$ and $5$ heads were obtained is shown in [Table 2.13](#Table-2-13).

```r
Table.2.13 <- matrix(c(0, 1, 2, 3, 4, 5, 38, 144, 342, 287, 164, 25), ncol=2)
colnames(Table.2.13) <- c("Number of Heads", "Number of Tosses (frequency)")
```

#### Table 2-13

```r
Table.2.13
```

(a) Graph the data of [Table 2.13](#Table-2-13)  
(b) Construct a table showing the percentage of tosses resulting  in a number of heads less than $0, 1, 2, 3, 4, 5,$ or $6$.  
(c) Graph the data of the table in part (b).

### Solution 2-17
<a id="Problem-2-17-a"></a>
(a) The data can be shown graphically either as in [Fig. 2-18](#Figure-2-18) or as in [Fig. 2-19](#Figure-2-19).

#### Figure 2-18
MINITAB dot plot of the number of heads.

#### Figure 2-19

```r
barplot(Table.2.13[,2], names.arg=Table.2.13[,1])
```

[Figure 2-18](#Figure-2-18) seems to be a more natural graph to use&mdash;since, for example, the number of heads cannot be $1.5$ or $3.2$. The graph is called a _dot plot_. It is especially used when the data are discrete.

[Figure 2-19](#Figure-2-19) shows a histogram of the data. Note that the total area of the histogram is the total frequency $1000$, as it should be. In using the histogram representation or the corresponding frequency polygon, we are essentially treating the data _as if_ they were continuous. This will later be found useful. Note that we have already used the histogram and frequency polygon for discrete data in [Problem 2.10](./02%20Frequency%20Distributions,%20Histograms,%20and%20Frequency%20Polygons.ipynb#Problem-2-10).

(b) Referring to the required [Table 2.14](#Table-2-14), note that it shows simply a cumulative-frequency distribution and percentage cumulative distribution of the number of heads. It should be observed that the entries "Less than 1," "Less than 2," etc., could just as well have been "Less than or equal to 0," "Less than or equal to 1," etc.

```r
Table.2.14 <- matrix(c(0, cumsum(Table.2.13[,2]), 0, cumsum(Table.2.13[,2])/10), nrow=length(Table.2.13[,2])+1)
colnames(Table.2.14) <- c("Number of Tosses (cumulative frequency)", "Percentage Number of Tosses (percentage cumulative frequency)")
rownames(Table.2.14) <- c("Less than 0", "Less than 1", "Less than 2", "Less than 3", "Less than 4", "Less than 5", "Less than 6")
p17.X <- c(0, 1, 2, 3, 4, 5, 6)
```

(c) The required graph can be presented either as in [Fig. 2-20](#Figure-2-20) or as in [Fig. 2-21](#Figure-2-21).

#### Figures 2-20 and 2-21

```r
par(mfrow=c(1,2))
options(repr.plot.width=24, repr.plot.height=8)
plot(stepfun(p17.X[2:7], Table.2.14[,2]), main="Fig. 2-20 MINITAB step function")
plot(p17.X, Table.2.14[,2], type="l", col="red", main="Fig. 2-21 MINITAB cumulative frequency polygon")
points(p17.X, Table.2.14[,2], pch=20)
options(repr.plot.width=7, repr.plot.height=7)
```

[Figure 2-20](#Figure-2-20) is the most natural for presenting discrete data&mdash;since, for example, the percentage of tosses in which there will be less than $2$ heads is equal to the percentage in which there will be less than $1.75, 1.56,$ or $1.23$ heads, so that the same percentage ($18.2\%$) should be shown for these values (indicated by the horizontal line).

[Figure 2-21](#Figure-2-21) shows the cumulative-frequency polygon, or ogive, for the data and essentially treats the data as if they were continuous.

Note that [Figs. 2-20 and 2-21](#Figures-2-20-and-2-21) correspond, respectively, to Figs. [2-18](#Figure-2-18) and [2-19](#Figure-2-19) of [part (a)](#Problem-2-17-a).
