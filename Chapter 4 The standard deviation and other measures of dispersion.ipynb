# Chapter 4 The Standard Deviation and Other Measures of Dispersion

## Dispersion, or variation
The degree to which numerical data tend to spread about an average value is called the _dispersion_, or _variation_, of the data. Various measures of this dispersion (or variation) are available, the most common being the range, mean deviation, semi-interquartile reange, $10$&mdash;$90$ percentile range, and standard deviation.

---

## The Range
The _range_ of a set of numbers is the difference between the largest and smallest numbers in the set.

---

__EXAMPLE 1.__ The range of the set $2, 3, 3, 5, 5, 5, 8, 10, 12$ is $12-2=10$. Sometimes the range is given by simply quoting the smallest and largest numbers; in the above set, for instance, the range could be indicated as $2$ to $12$, or $2-12$.

---

__4.1.__ Find the range of the sets (a) 12, 6, 7, 3, 15, 10, 18, 5 and (b) 9, 3, 8, 8, 9, 8, 9, 18

### Solution
In both cases, range $=$ largest number $-$ smallest number $=18-3=15$. However, as seen from the arrays of sets (a) and (b),<p>
$$
\begin{align*}
\text{(a)}\ 3, 5, 6, 7, 10, 12, 15, 18\ \ \ \ \ \ \ \ \text{(b)}\ 3, 8, 8, 9, 9, 9, 18
\end{align*}
$$<p>
there is much more variation or dispersion, in (a) than in (b). In fact, (b) consists mainly of $8$'s and $9$'s.<p>
Since the range indicates no difference between the sets, it is not a very good measure of dpsersion in this case. Where extreme values are present, the range is generally a poor measure of dispersion.<p>
An improvement is achieved by throwing out the extreme cases, $3$, and $18$. Then for set (a) the range is $(15 - 5) = 10$, while for set (b) the range is $(9-8)=1$, clearly showing that (a) has greater dispersion than (b). However, this is not the way the range is defined. The semi-interquartile range and the 10&mdash;90 percentile range were designed to improve on the range by eliminating extreme cases.

````r
p1.a <- c(12, 6, 7, 3, 15, 10, 18, 5)
p1.b <- c(9, 3, 8, 8, 9, 8, 9, 18)
(p1.a.range <- diff(range(p1.a)))
(p1.b.range <- diff(range(p1.b)))
````

## The mean deviation
The _mean deviation_, or _average deviation_, of a set of $N$ numbers $X_1, X_2,\ldots,X_N$ is abbreviaged $\text{MD}$ and is defined.<p>
\begin{align}
\text{MD}
= \frac{\sum^N_{j=1}{\lvert X_j-\bar{X}\rvert}}{N}
= \frac{\sum{\lvert X-\bar{X}\rvert}}{N}
= \frac{\sum{|X - \bar{X}|}}{N}
= \overline{\lvert X-\bar{X}\rvert}\label{eq1}\tag{1}
\end{align}<P>
where $\bar{X}$ is the arithmetic mean of the numbers and $\lvert X_j-\bar{X}\rvert$ is the absolute value of the deviation of $X_j$ from $\bar{X}$. (The _absolute value_ of a number is the number without the associated sign and is indicated by two vertical lines placed around the number; thus $\lvert-4\rvert=4,\lvert+3\rvert=3,\lvert6\rvert=6,$ and $\vert-0.84\rvert=0.84$.)<p>
__EXAMPLE 2.__  Find the mean deviation of the set $2, 3, 6, 8, 11$.<p>
\begin{align}
\text{Arithmetic mean (}\bar{X}\text{)}=\frac{2+3+6+8+11}{5}=6
\end{align}<p>
\begin{align}
\text{MD}
=\frac{\lvert2-6\rvert+\lvert3-6\rvert+\lvert6-6\rvert+\lvert8-6\rvert+\lvert11-6\rvert}{5}
=\frac{\lvert-4\rvert+\lvert-3\rvert+\lvert0\rvert+\lvert2\rvert+\lvert5\rvert}{5}
=\frac{4+3+0+2+5}{5}=2.8
\end{align}<p>
If $X_1, X_2,\ldots,X_K$ occur with frequencies $f_1,f_2,\ldots,f_K$, respectively, the mean deviation can be written as<p>
\begin{align}
\text{MD}
=\frac{\sum^K_{j=1}{f_j\lvert X_j-\bar{X}\rvert}}{N}
=\frac{\sum{f\lvert X-\bar{X}\rvert}}{N}
=\overline{\lvert X-\bar{X}\rvert}\label{eq2}\tag{2}
\end{align}<p>
where $N=\sum^K_{j=1}{f_j}=\sum{f}$. This form is useful for grouped data, where the $X_j$'s represent class marks and the $f_j$'s are the corresponding class frequencies.<p>
Occasionally the mean deviation is defined in terms of absolute deviations from the median or other average instead of from the mean. An interesting property of the sum $\sum^N_{j=1}\lvert X_j-a\rvert$ is that it is a minimum where $a$ is the median (i.e., the mean deviation about the median is a minimum).<p>
Note that it would be more appropriate to use the terminology _mean absolute deviation_ than _mean deviation_.

Below is a function that can be used to find the mean deviation. It is also found in the saved data file `Saves/mean_deviation.Rdata`.

````r
md <- function(x) {
    return (sum(abs(x - mean(x)))/length(x))
}
````

## The Semi-Interquartile Range
The _semi-interquartile range_, or _quartile deviation_, of a set of data is denoted by $Q$ and is defined by<p>
\begin{align}
Q = \frac{Q_3 - Q_1}{2}\label{eq3}\tag{3}
\end{align}<p>
where $Q_1$ and $Q_3$ are the first and third quartiles for the data (see Problems 4.6 and 4.7). The interquartile range $Q_3-Q_1$ is sometimes used, but the semi-interquartile range is more common as a measure of dispersion.

---

__4.6.__ Find the semi-interquartile range for the height distribution of the students at XYZ University.

````r
Table_2.1
````

### Solution
The lower and upper quartiles are $Q_1=65.5+\frac{2}{42}(3)=65.64$in and $Q_3=68.5+\frac{10}{27}(3)=69.61$in, respectively, and the semi-interquartile range (or quartile deviation) is $Q=\frac{1}{2}(Q_3-Q_1)=\frac{1}{2}(69.61-65.64)=1.98$in. Note that $50\%$ of the cases lie between $Q_1$ and $Q_3$ (i.e., $50$ students have heights between $65.64$ and $69.61$in).<p>
We can consider $\frac{1}{2}(Q_1+Q_3)=67.63$in to be a measure of central tendency (i.e., average height). It follwos that $50\%$ of the heights lie in the range $67.63\pm1.98$in.

````r
p6.ci <- class.interval <- 3
p6.boundaries <- seq(59.5, 74.5, by=class.interval)
p6.cumsum <- cumsum(as.vector(Table_2.1))
p6.f.sum <- sum(Table_2.1)
p6.Q.breaks <- p6.f.sum/4

p6.Q1.index <- match(Table_2.1[p6.cumsum > p6.Q.breaks][1],Table_2.1)
p6.Q1.close.class <- p6.boundaries[p6.Q1.index]
p6.Q1.class <- as.vector(Table_2.1)[p6.Q1.index]
p6.Q1.remainder <- p6.Q.breaks - p6.cumsum[p6.Q1.index-1]
p6.Q1 <- round(p6.Q1.close.class + p6.Q1.remainder/p6.Q1.class*p6.ci,2)

p6.Q3.index <- match(Table_2.1[p6.cumsum > 3*p6.Q.breaks][1], Table_2.1)
p6.Q3.close.class <- p6.boundaries[p6.Q3.index]
p6.Q3.class <- as.vector(Table_2.1)[p6.Q3.index]
p6.Q3.remainder <- 3*p6.Q.breaks - p6.cumsum[p6.Q3.index-1]
p6.Q3 <- round(p6.Q3.close.class + p6.Q3.remainder/p6.Q3.class*p6.ci,2)

c((p6.Q.central <- round((p6.Q1 + p6.Q3)/2,2)), (p6.Q.pm <- round((p6.Q3 - p6.Q1)/2,2)))
(p6.Q.range <- c(p6.Q.central-p6.Q.pm, p6.Q.central+p6.Q.pm))
````

## The 10-90 Percentile Range
The $10$&mdash;$90$ _percentile range_ of a set of data is defined by<p>
\begin{align}
10-90\ \text{percentile range}=P_{90}-P_{10}\label{eq4}\tag{4}
\end{align}<p>
where $P_{10}$ and $P_{90}$ are the $10$th and $90$th percentiles for the data (see Problem 4.8). The semi-$10$&mdash;$90$ percentile range, $\frac{1}{2}(P_{90}-P_{10})$, can also be used but is not commonly employed.

## The Standard Deviation
The _standard deviation_ of a set of $N$ numbers $X_1,X_2,\ldots,X_N$ is denoted by $s$ and is defined by<p>
\begin{align}
s
= \sqrt{\frac{\sum^N_{j=1}{(X_j - \bar{X})^2}}{N}}
= \sqrt{\frac{\sum{(X=\bar{X})^2}}{N}}
= \sqrt{\frac{\sum{x^2}}{N}}
= \sqrt{(X - \bar{X})^2}\label{eq5}\tag{5}
\end{align}<p>
where $x$ represents the deviations of each of the numbers $X_j$ from the mean $\bar{X}$. Thus $s$ is the root mean square (RMS) of the deviations from the mean, or, as it is sometimes called, the _root-mean-square deviation_.<p>
If $X_1,X_2,\ldots,X_K$ occur with frequencies $f_1,f_2,\ldots,f_k$, respectively, the standard deviation can be written<p>
\begin{align}
s
=\sqrt{\frac{\sum^K_{j=1}{f_j(X_j-\bar{X})^2}}{N}}
=\sqrt{\frac{\sum{f(X-\bar{X})^2}}{N}}
=\sqrt{\frac{\sum{fX^2}}{N}}
=\sqrt{\overline{(X-\bar{X})^2}}\label{eq6}\tag{6}
\end{align}<p>
where $N=\sum^K_{j=1}{f_j}=\sum{f}$. In this form it is useful for grouped data.<p>
Sometimes the standard deviation of a sample's data is defined with $(N-1)$ replacing $N$ in the denominators of the expressions in equations (\ref{eq5}) and (\ref{eq6}) because the resulting value represents a better estimate of the standard deviation of a populiation from which the sample is taken. For large values of $N$ (certainly $N>30$), there is practically no difference between the two definitions. Also, when the better estimate is needed we can always obtain it by multiplying the standard deviation computed according to the first definition by $\sqrt{\frac{N}{(N-1)}}$. Hence we shall adhere to the form (5) and (6).

````r
sd <- function(x) {
        return (sqrt(sum((x - mean(x))^2)/length(x)))
}
````

__When should the SD be computed with a denominator of n instead of (n-1)?__

Statistics books often show two equations to compute the SD, one using n, and the other using n-1, in the denominator. Some calculators have two buttons. 

The n-1 equation is used in the common situation where you are analyzing a sample of data and wish to make more general conclusions. The SD computed this way (with n-1 in the denominator) is your best guess for the value of the SD in the overall population.

If you simply want to quantify the variation in a particular set of data, and don't plan to extrapolate to make wider conclusions, then you can compute the SD using n in the denominator. The resulting SD is the SD of those particular values. It makes no sense to compute the SD this way if you want to estimate the SD of the population from which those points were drawn. It only makes sense to use n in the denominator when there is no sampling from a population, there is no desire to make general conclusions. 

The goal of science is always to generalize, so the equation with n in the denominator should not be used. The only example I can think of where it might make sense is in quantifying the variation among exam scores. But much better would be to show a scatterplot of every score, or a frequency distribution histogram. 

GraphPad Prism and InStat always compute the SD with the n-1 denominator.

## The variance
The _variance_ of a set of data is defined as the square of the standard deviation and is thus given by $s^2$ in equations (5) and (6).<p>
When it is necessary to distinguish the standard deviation of a population from the standard deviation of a sample drawn from this population, we often use the symbol $s$ for the latter and $\sigma$ (lowercase Greek _sigma_) for the former. Thus $s^2$ and $\sigma^2$ would represent the _sample variance_ and _population variance_, respectively.

---

## Short methods for computing the standard deviation
Equations (5) and (6) can be written, respectively, in the equivalent forms
\begin{align}
s
=\sqrt{\frac{\sum^N_{j=1}{X_j^2}}{N}-\left(\frac{\sum^N_{j=1}{X_j}}{N}\right)^2}
=\sqrt{\frac{\sum{X^2}}{N}-\left(\frac{\sum{X}}{N}\right)^2}
=\sqrt{\overline{X^2}-\bar{X}^2}\label{eq7}\tag{7}
\end{align}<p>
\begin{align}
s
=\sqrt{\frac{\sum^K_{j=1}{f_jX_j^2}}{N}-\left(\frac{\sum^N_{j=1}{X_j}}{N}\right)^2}
=\sqrt{\frac{\sum{fX^2}}{N}-\left(\frac{\sum{fX}}{N}\right)^2}
=\sqrt{\overline{X^2}-\bar{X}^2}\label{eq8}\tag{8}
\end{align}<p>
where $\overline{X^2}$ denotes the mean of the squares of the various values of $X$, while $\bar{X}^2$ denotes the square of the mean of the various values of X (see Problems 4.12 to 4.14).<p>
If $d_j=X_j-A$ are the deviations of $X_j$ from some arbitrary constant A, results (\ref{eq7}) and (\ref{eq8}) become, resqpectively,<p>
\begin{align} 
s
=\sqrt{\frac{\sum^N_{j=1}{d_j^2}}{N}-\left(\frac{\sum^N_{j=1}{d_j}}{N}\right)^2}
=\sqrt{\frac{\sum{d^2}}{N}-\left(\frac{\sum{d}}{N}\right)^2}
=\sqrt{\overline{d^2}-\bar{d}^2}\label{eq9}\tag{9}
\end{align}<p>
\begin{align} 
s
=\sqrt{\frac{\sum^N_{j=1}{f_jd_j^2}}{N}-\left(\frac{\sum^N_{j=1}{f_jX_j}}{N}\right)^2}
=\sqrt{\frac{\sum{fd^2}}{N}-\left(\frac{\sum{fd}}{N}\right)^2}
=\sqrt{\overline{d^2}-\bar{d}^2}\label{eq10}\tag{10}
\end{align}<p>
(See Problems 4.15 and 4.17.)<p>
When data is grouped into a frequency distribution whose class intervals have equal size $c$, we have $d_j=cu_j$ or $X_j=A+cu_j$ and result (10) becomes<p>
\begin{align}
s
=c\sqrt{\frac{\sum^K_{j=1}{f_ju_j^2}}{N}-\left(\frac{\sum^K_{j=1}{f_ju_j}}{N}\right)^2}
=c\sqrt{\frac{\sum{fu^2}}{N}-\left(\frac{\sum{fu}}{N}\right)^2}
=c\sqrt{\overline{u^2}-\bar{u}^2}\label{eq11}\tag{11}
\end{align}<p>
This last formula provides a very short method for computing the standard deviation and should always be used for grouped data when the class-interval sizes are equal. It is called the _coding method_ and is exactly analogous to that used in Chapter 3 for computing the arithmetic mean of grouped data. (See Problems 4.16 to 4.19.)

## Properties of the standard deviation
1. The standard deviation can be defined as

\begin{align}
s=\sqrt{\frac{\sum^N_{j=1}{(X_j-a)^2}}{N}}
\end{align}

where $a$ is an average besides the arithmetic mean. Of all such standard deviations, the minimum is that for which $a=\bar{X}$, because of Property 2 in Chapter 3. This property provides an important reason for defining the standard deviation as above. For a proof of this property see Problem 4.27.
