# Set operations

## Equality and Difference
Two sets $A$ and $B$ are _equal_ if all elements in $A$ are present in $B$ and all elements in $B$ are present in $A$; that is, if $A \subseteq B$ and $B \subseteq A$. For equal sets we write $A=B$.

The _difference- of two sets $A$ and $B$, denoted $A-B$, is the set containing all elements of $A$ that are not present in$B$. For example, with $A = \{ 0.6 \lt a \le 1.6 \}$ and $B = \{1.0 \le b \le 2.5\}$, then $A-B=\{0.6 \lt c \le 1.0\}$ or $B-A=\{1.6 \lt d \le 2.5\}$. Note that $A-B \ne B-A$.

## Union and intersection
The _union_ (called it $C$) of two sets $A$ and $B$ is written
\begin{align*}\tag{1.2-1}
C = A \cup B
\end{align*}
It is the set of all elements of $A$ or $B$ or both. The union is sometimes called the _sum_ of two sets.

The _intersection_ (call it $D$) of two sets $A$ and $B$ is written
\begin{align*}\tag{1.2-2}
D = A \cap B
\end{align*}
It is the set of all elements common to both $A$ and $B$. Intersection is sometimes called the _product_ of two sets. For mutually exclusive sets $A$ and $B$, $A \cap B = \emptyset$.

By repeated application of (1.2-1) or (1.2-2), the union and intersection of $N$ sets $A_n, n=1, 2, \ldots, N$, become
\begin{align*}
C = A_1 \cup A_2 \cup \cdots \cup A_N = \bigcup^N_{n=1}A_n\tag{1.2-3} \\
\\
D = A_1 \cap A_2 \cap \cdots \cap A_N = \bigcap^N_{n=1}A_n\tag{1.2-4}
\end{align*}

### Complement
The _complement_ of a set $A$, denoted by $\bar{A}$, is the set of all elements not in $A$.  
Thus,
\begin{align*}\tag{1.2-5}
\bar{A} = S - A
\end{align*}
It is also easy to see that $\bar{\emptyset} = S$, $\bar{S} = \emptyset$, $A \cup \bar{A} = S$, and $A \cap \bar{A} = \emptyset$.

__EXAMPLE 1.2-1__. We illustrate intersection, union, and complement by taking an example with the four sets
\begin{align*}
S &= \{1 \le \text{integers} \le 12\}     &B &= \{2, 6, 7, 8, 9, 10, 11\} \\
A &= \{1, 3, 5, 12\}                      &C &= \{1, 3, 4, 6, 7, 8\}
\end{align*}
Applicable unions and intersections here are:
\begin{align*}
A \cup B &= \{1, 2, 3, 5, 6, 7, 8, 9, 10, 11, 12\}     &A \cap B &= \emptyset \\
A \cup C &= \{1, 3, 4, 5, 6, 7, 8, 12\}                &A \cap C &= \{1, 3\} \\
B \cup C &= \{1, 2, 3, 4, 6, 7, 8, 9, 10, 11\}         &B \cap C &= \{6, 7, 8\}
\end{align*}
Complements are
\begin{align*}
\bar{A} &= \{2, 4, 6, 7, 8, 9, 10, 11\} \\
\bar{B} &= \{1, 3, 4, 5, 12\} \\
\bar{C} &= \{2, 5, 9, 10, 11, 12\}
\end{align*}

### Algebra of Sets
#### The commutative law
\begin{align*}
A \cap B &= B \cap A\tag{1.2-6} \\
\\
A \cup B &= B \cup A\tag{1.2-7}
\end{align*}
#### The distributive law
\begin{align*}
A \cap (B \cup C) = (A \cap B) \cup (A \cap C)\tag{1.2-8} \\
\\
A \cup (B \cap C) = (A \cup B) \cap (A \cup C)\tag{1.2-9}
\end{align*}
#### De Morgan's laws
\begin{align*}
(\overline{A \cup B}) = \bar{A} \cap \bar{B}\tag{1.2-12} \\
\\
(\overline{A \cap B}) = \bar{A} \cup \bar{B}\tag{1.2-13}
\end{align*}
#### Duality principle
\begin{align*}
A \cap (B \cup C) = (A \cap B) \cup (A \cap C)\tag{1.2-14} \\
\\
A \cup (B \cap C) = (A \cup B) \cap (A \cup C)\tag{1.2-15}
\end{align*}

__EXAMPLE 1.2-3__. We demonstrate (1.2-14) for three sets $A=\{1, 2, 4, 6\}$, $B = \{2, 6, 8, 10\}$, and $C=\{3 \le c \le 4\}$. Here $B \cup C = \{2, 3 \le c \le 4, 6, 8, 10\}$, $A \cap B = \{2, 6\}$, $A \cap C = \{4]$. For the left side of (1.2-14) $A \cap (B \cup C) = \{2, 4, 6\}$. For the right side of (1.2-14) $(A \cap B) \cup (A \cap C) = \{2, 4, 6\}$. This result is the same as for the left side, demonstrating the validity of (1.2-14) for these sets.
