# Set definitions

A _set_ is a collection of objects. If $a$ is an element of set $A$, then we write
\begin{align*}\tag{1.1-1}
a \in A
\end{align*}
If $a$ is not an element of $A$, we write
\begin{align*}\tag{1.1-2}
a \notin A
\end{align*}

A set is specified by the content of two braces: $\{\cdot\}$.

Two methods for specifying content in a set:
1. _tabular method_ - elements are enumerated explicitly. I.e. $\{1, 2, 3, 4\}$.
2. _rule_method_ - elements are specified by some rule. I.e. $\{\text{integers between 5 and 10}\}$.

A set is _countable_ if its elements can be put in one-to-one correspondence with the natural numbers, which are the integers $1, 2, 3, \ldots$. If a set is not countable, it is called _uncountable_.

A set is empty if it has no elements. The empty set symbol is $\emptyset$.

A set is _finite_ if it is either empty, or has elements tha can be counted, with the counting process terminating. If a set is not finite, then it is called _infinite_. An infinite set having countable elements is called _countably infinite_.

If every element of a set $A$ is also an elment in antoher set $B$, $A$ is said to be contained in $B$. $A$ is known as a _subset_ of $B$ and we write
\begin{align*}\tag{1.1-3}
A \subseteq B
\end{align*}
If at least one element exists in $B$ which is not in $A$, then $A$ is a _proper subset_ of $B$, denoted by
\begin{align*}\tag{1.1-4}
A \subset B
\end{align*}

The null set $\{\emptyset\}$ is clearly a subset of all other sets.

Two sets $A$ and $B$, are called _disjoint_ or _mutually exclusive_ if they have no common elements.

__EXAMPLE 1.1-1__. To illustrate the topics discussed above, we identify the sets listed below,
\begin{align*}
A &= \{1, 3, 5, 7\}         &D &= {0.0} \\
B &= \{1, 2, 3, \ldots\}    &E &= \{2, 4, 6, 8, 10, 12, 14\} \\
C &= \{0.5 \lt c \leq 8.5\} &F &= \{-5.0 \lt f \leq 12.0\}
\end{align*}
Set $A$ is tabularly specified, countable, and finite.  
Set $B$ is also tabularly specified and countable, but is infinite.  
Set $C$ is rule-specified, uncountable, and infinite, since it contains _all_ numbers greater than $0.5$ but not exceeding 8.5.  
Set $D$ and $E$ are countably finite, while set $F$ is uncountably infinite. Set $D$ is _not_ the null set; it has one element, the number zero.  
Set $F$ is uncountably infinite.

Set $A$ is contained in sets $B$, $C$, and $F$. SImilarly, $C \subset F$, $D \subset F$ and $E \subset B$. Sets $B$ and $F$ are not subsets of any of the other sets or of each other. Sets $A$, $D$, and $E$ are mutually exclusive of each other.

The largest or all-encompassing set of objects under discussion in a given situation is called the _universal set_, denoted $S$. All sets (of the situation considered) are subsets of the universal set. An example will help clarify the concept of a universal set.

__EXAMPLE 1.1-2__. Suppose we consider the problem of rolling a die. We are interested in the numbers that show on the upper face. Here the universal set is $S = \{1, 2, 3, 4, 5, 6\}$. In a gambling game, suppose a person wins if the number comes up odd. This person wins for any number in the set $A=\{1, 3, 5\}$. Another person might win if the number shows four or less; that is, for any number in the set $B = \{1, 2, 3, 4\}$.  
Observe that both $A$ and $B$ are subsets of $S$. (The reader should check this for a few values of $N$.) For the present example, $N=6$ and check this for a few values of $N$.) For the present example, $N=6$ and $2^N=64$, so that there are $64$ ways one can define "winning" with one die.

It should be noted that winning or losing in the above gambling game is related to a set. The game itself is partially specified by its universal set (other games typically have a different universal set). These facts are not just coincidence, and we shall shortly find that sets form the basis on which our study of probability is constructed.
