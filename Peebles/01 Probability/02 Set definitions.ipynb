# Set definitions

A _set_ is a collection of objects. If $a$ is an element of set $A$, then we write
\begin{align*}\tag{1.1-1}
a \in A
\end{align*}
If $a$ is not an element of $A$, we write
\begin{align*}\tag{1.1-2}
a \notin A
\end{align*}

A set is specified by the content of two braces: $\{\cdot\}$.

Two methods for specifying content in a set:
1. _tabular method_ - elements are enumerated explicitly. I.e. $\{1, 2, 3, 4\}$.
2. _rule_method_ - elements are specified by some rule. I.e. $\{\text{integers between 5 and 10}\}$.

A set is _countable_ if its elements can be put in one-to-one correspondence with the natural numbers, which are the integers $1, 2, 3, \ldots$. If a set is not countable, it is called _uncountable_.

A set is empty if it has no elements. The empty set symbol is $\emptyset$.

A set is _finite_ if it is either empty, or has elements tha can be counted, with the counting process terminating. If a set is not finite, then it is called _infinite_. An infinite set having countable elements is called _countably infinite_.

If every element of a set $A$ is also an elment in antoher set $B$, $A$ is said to be contained in $B$. $A$ is known as a _subset_ of $B$ and we write
\begin{align*}\tag{1.1-3}
A \subseteq B
\end{align*}
If at least one element exists in $B$ which is not in $A$, then $A$ is a _proper subset_ of $B$, denoted by
\begin{align*}\tag{1.1-4}
A \subset B
\end{align*}

The null set $\{\emptyset\}$ is clearly a subset of all other sets.

Two sets $A$ and $B$, are called _disjoint_ or _mutually exclusive_ if they have no common elements.

__EXAMPLE 1.1-1__. To illustrate the topics discussed above, we identify the sets listed below,
\begin{align*}
A &= \{1, 3, 5, 7\}         &D &= {0.0} \\
B &= \{1, 2, 3, \ldots\}    &E &= \{2, 4, 6, 8, 10, 12, 14\} \\
C &= \{0.5 \lt c \leq 8.5\} &F &= \{-5.0 \lt f \leq 12.0\}
\end{align*}
Set $A$ is tabularly specified, countable, and finite.  
Set $B$ is also tabularly specified and countable, but is infinite.  
Set $C$ is rule-specified, uncountable, and infinite, since it contains _all_ numbers greater than $0.5$ but not exceeding 8.5.  
Set $D$ and $E$ are countably finite, while set $F$ is uncountably infinite. Set $D$ is _not_ the null set; it has one element, the number zero.  
Set $F$ is uncountably infinite.

Set $A$ is contained in sets $B$, $C$, and $F$. SImilarly, $C \subset F$, $D \subset F$ and $E \subset B$. Sets $B$ and $F$ are not subsets of any of the other sets or of each other. Sets $A$, $D$, and $E$ are mutually exclusive of each other.

The largest or all-encompassing set of objects under discussion in a given situation is called the _universal set_, denoted $S$. All sets (of the situation considered) are subsets of the universal set. An example will help clarify the concept of a universal set.

__EXAMPLE 1.1-2__. Suppose we consider the problem of rolling a die. We are interested in the numbers that show on the upper face. Here the universal set is $S = \{1, 2, 3, 4, 5, 6\}$. In a gambling game, suppose a person wins if the number comes up odd. This person wins for any number in the set $A=\{1, 3, 5\}$. Another person might win if the number shows four or less; that is, for any number in the set $B = \{1, 2, 3, 4\}$.  
Observe that both $A$ and $B$ are subsets of $S$. (The reader should check this for a few values of $N$.) For the present example, $N=6$ and check this for a few values of $N$.) For the present example, $N=6$ and $2^N=64$, so that there are $64$ ways one can define "winning" with one die.

It should be noted that winning or losing in the above gambling game is related to a set. The game itself is partially specified by its universal set (other games typically have a different universal set). These facts are not just coincidence, and we shall shortly find that sets form the basis on which our study of probability is constructed.

## Problems

1. Specify the following sets by the rule method.
\begin{align*}
A = \{ 1, 2, 3 \},\ \ B = \{ 8, 10, 12, 14 \},\ \ C = \{ 1, 3, 5, 7, \ldots \}
\end{align*}

_Solution._
\begin{align*}
A = \{1 \le a \le 3 \},\ \ B = \left\{ \sum^7_{b=4}{2b} \right\},\ \ C = \left\{ \sum^{\infty}_{c=1}{2n - 1} \right\}
\end{align*}

2. Use the tabular method to specify a class of sets for the sets of Problem 1.1-1.

3. State whether the following sets are countable or uncountable, or finite or infinite.  
$A = \{1\}$,  
$B = \{x = 1\}$,  
$C = \{0 < \text{integers}\}$,  
$D = \{\text{children in public school No. 5 at 3:00 A.M.}\}$,  
$E = \{\text{girls in public school No. 5}\}$,  
$F = \{\text{girls in class in public school No. 5 at 3:00 A.M.}\}$,  
$G = \{\text{all lengths not exceeding one meter}\}$,  
$H = \{-25 \le x \le -3\}$,  
$I = \{ -2, -1, 1 \le x \le 2\}$.

_Solution._
A is countable and finite.  
B is countable and finite.  
C is countable and infinite.  
D is countable and finite.  
E is countable and finite.  
F is countable and finite.  
G is countable and infinite.  
H is countable and finite.  
I is countable and finite.

4. For each set of Problem 3, determine if it is equal to, or a subset of, any of the other sets.

_Solution._
A and B are equal.  
F is a subset of D.  
F is a subset of E.  
A is a subset of I.  
B is a subset of I.

6. A thermometer measures temperatures from $-40$ to $130\textdegree$F ($-40 to 54.4\textdegree$C).
* (a) State a universal set to describe temperature measurements. Specify subsets for:
* (b) Temperature measurements not exceeding water's freezing point, and
* (c) Measurements exceeding the freezing point but not exceeding $100\textdegree$F ($37.8\textdegree$C).

_Solution._
* (a) $S = \{ -40\textdegree \le f \le 130\textdegree \}$
* (b) $S_0 = \{ -40\textdegree \le f \le 0\textdegree \}$
* (c) $S_1 = \{ 0\textdegree < f < 100\textdegree \}$

7. Prove that a set with $N$ elements has $2^N$ subsets.

8. A random noise voltage at a given time may have any value from $-10$ to $10$V.
* (a) What is the universal set describing noise voltage?
* (b) Find a set to describe the voltages available from a half-wave rectifier for positive voltages with a linear output-input voltage characteristic.
* (c) Repeat parts (a) and (b) if a dc voltage of $-3$V is added to the random noise.

_Solution._  
* (a) $S = \{ -10 \le v \le 10 \}$
* (b) $S_{v^+} = \{ 0 < v \le 10 \}$
* (c) $S_{-3} = \{ -13 \le v \le 7 \}$, $S_{-3v^+} = \{ 0 < v \le 7 \}$

9. Use the tabular method to define a set $A$ that contains all integers with magnitudes not exceeding 7. Define a second set $B$ having odd integers larger than $-2$ and not larger than $5$. Determine if $A \subset B$ and $B \subset A$.

_Solution._  
$A = \{\ldots, -1, 0, 1, 2, 3, 4, 5, 6 \}$  
$B = \{-1, 1, 3, 5 \}$

10. A set $A$ has three elements, $a_1$, $a_2$, and $a_3$. Determine all possible subsets of $A$.

_Solution._
\begin{align*}
\{\emptyset\}, \{a_1\}, \{a_2\}, \{a_3\}, \{a_1, a_2\}, \{a_1, a_3\}, \{a_2, a_3\}, \{a_1, a_2, a_3\}
\end{align*}

11. Specify, by both the tabular and rule methods, each of the following sets: (a) all integers between 1 and 9, (b) all integers from 1 to 9, (c) the five values of equivalent resistance for $n$ identical 10$\ohm$ resistors in parallel where $n = 1, 2, \ldots, 5$, and (d) the six values of equivalent resistance for $n$ identical 2.2$\ohm$ resistors in series where $n = 1, 2, \ldots, 6$.

* (c) $I_{\text{rule}} = \left\{ \sum^5_{n=1}{\frac{10n}{10 + n}} \right\}$, $I_{\text{tabular}} = \left\{1, \frac{20}{12}, \frac{30}{13}, \frac{40}{14}, \frac{50}{15}\right\}$

12. A box contains 100 capacitors (universal set), of which 40 are $0.01\micro$F with a 100$V$ voltage rating, 35 are $0.1\micro$F at a rating of 50$V$, and 25 are $1.0\micro$F and have a 10$V$ rating. Determine the number of elements in the following sets:
* (a) $A = \{$capacitors with capacitance $\ge 0.1\micro\text{F}\}$
* (b) $B = \{$capacitors with voltage rating $> 5\text{V}\}$
* (c) $C = \{$capacitors with both capacitance $\ge 0.1\micro\text{F}$ _and_ voltage rating $\ge 50\text{V}\}$

_Solution._
* (a) A = 75
* (b) B = 100
* (c) C = 35
