## Experiments and Sample Spaces
Right now we will not define, mathematically, an _experiement_. What we will do is use examples. An experiment might consist of rolling a single die and obserbing the number that shows up. Another experiment might be the draw of a card from a deck of 52 playing cards.

The _sample space_ ($S$), is the set of all possible outcomes.

## Discrete and Continuous Sample Spaces
A sample space can be discrete and finite. For example, the set of natural numbers from $S = \{1 \le s \le 10\}$ is discrete and finite. A set can be discrete and infinite, such as the set of all natural numbers $S = \{1, 2, 3, 4, 5, \ldots \}$. A continuous set is an uncountably infinite sample space. An example is all real numbers from $S = \{0 \le s \le 10\}$.

Sample spaces can be defined such as "a deck of 52 playing cards," or "two fair dice".

## Events
An _event_ is defined as a subset of the sample space. Because an event is a set, all the earlier definitions and operations applicable to sets will apply to events. For example, if two events have no common outcomes they are _mutually exclusive_.

An event could be "draw a spade," from the sample space: deck of 52 playing cards. Another event can be a head on the toss of a coin.

## Probability definition and axioms
To each event defined on a sample space $S$, we shall assign a nonnegative number called _probability_. Probability is therefore a function; it is a function of the events defined.

The assigned probabilities are chosen so as to satisfy three _axioms_. Let $A$ be any event defined on a sample space $S$. Then the first two axioms are

axiom 1: $$P(A) \ge 0\tag{1.3-1a}$$

axiom 2: $$P(S) = 1\tag{1.3-1b}$$

The null set $\emptyset$ is an event with no elements; it is known as the _impossible event_ and its probability is 0.

The third axiom applies to $N$ events $A_n, n = 1, 2, \ldots, N$, where $N$ may possibly be infinite, defined on a sample sapce $S$, and having the property $A_m \cap A_n = \emptyset$ for all $m \ne n$. It is  
axiom 3:  
\begin{align*}\tag{1.3-1c}
\qquad \qquad P\left(\bigcup^N_{n=1}A_n\right) = \sum^N_{n=1}P(A_n) \qquad if \qquad A_m \cap A_n = \emptyset
\end{align*}
for all $m \ne n = 1, 2, \ldots, N$, with $N$ possibly infinite. The axiom states that the probability of the event equal to the union of any number of mutually exclusive events is equal to the sum of the individual event probabilities.

__EXAMPLE 1.3-1__. Let an experiment consist of obtaining a number $x$ by spinning  the pointer on a "fair" wheel of chance that is labeled from 0 to 100 points. The sample space is $S = \{0 \lt x \le 100\}$. We reason that probability of the pointer falling between any two numbers $x_2 \ge x_1$ should be $(x_2 - x_1)/100$ since the wheel is fair. As a check on this assignment, we see that the event $A = \{x_1 \gt x \ge x_2\}$ satisfies axiom 1 for all $x_1$ and $x_2$, and axiom 2 when $x_2 = 100$ and $x_1 = 0$.

Now suppose we break the wheel's periphery into $N$ contiguous segments $A_n = \{x_{n-1} \gt x \ge x_n \}$, $x_n = (n)100/N$, $n = 1, 2, \ldots, N$, with $x_0 = 0$. Then $P(A_n) = 1/N$, and, for any $N$,
\begin{align*}
P\left(\bigcup^N_{n=1}A_n\right) = \sum^N_{n=1}P(A_n) = \sum^N_{n=1}\frac{1}{N} = 1 = P(S)
\end{align*}
from axiom 3.

## Mathematical model of experiments
The three axioms given above complete our mathematical model of an experiment. Given some real physical expermiment having a set of particular outcomes possible, we first define a _sample space_ to mathematically represent the physical outcomes. Second, it was recognized that certain characteristics of the outcomes in the real experiment were of interest, as opposed to the outcomes themselves; _events_ were defined to mathematically represent these characteristics. Finally, _probabilities_ were assigned to the defined events to mathematically account for the random nature of the experiment.

Thus, a real experiment is defined by three things:
1. assignment of a sample space;
2. defintion of events of interest; and
3. making probability assignments to the events such that the axioms are statisfied.

Establishing the correct model for an experiment is probably the single most difficult step in solving probability problems.

__EXAMPLE 1.3-2__. An experiment consists of observing the sum of the numbers showing up when two dice are thrown. We develop a model for this experiment.

The sample space consists of $6^2 = 36$ points as shown in Figure 1.3-1. Each possible outcome corresponds to a sum having values from 2 to 12.

Suppose we are mainly interested in three events defined by $A=\{\text{sum} = 7\}$, $B = \{8 \lt \text{sum} \le 11\}$, and $C = \{10 \lt \text{sum}\}$. In assigning probabilities to these events, it is first convenient to define 36 _elementary events_ $A_{ij} = \{\text{sum for outcome} (i,j) = i + j\}$, where $i$ represents the row and $j$ represents the column locating a particular possible outcome in Figure 1.3-1. An elementary event has only one element.

For probability assignments, intuition indicates that each possible outcome has the same likelihood of occurrence if the dice are fair, so $P(A_{ij})=\frac{1}{36}$. Now because the events $A_{ij}, i$ and $j = 1, 2, \ldots, N = 6$, are mutually exclusive, they must satisfy axiom 3. But since the events $A$, $B$, and $C$ are simply the unions of appropriate elementary events, their probabilities are derived from axiom 3. From Figure 1.3-1 we easily find
\begin{align*}
P(A) &= P\left(\bigcup^6_{i=1}A_{i,7-i}\right) = \sum^6_{i=1}P(A_{i,7-i}) = 6\left(\frac{1}{36}\right) = \frac{1}{6} \\
P(B) &= 9\left(\frac{1}{36}\right) = \frac{1}{4} \\
P(C) &= 3\left(\frac{1}{36}\right) = \frac{1}{12}
\end{align*}
As a matter of interest, we also observe the probabilities of the events $B \cap C$ and $B \cup C$ to be $P(B \cap C) = 2(\frac{1}{36})=\frac{1}{18}$ and $P(B \cup C) = 10(\frac{1}{36}) = \frac{5}{18}$.

## Probability as a relative frequency
The use of common sense and engineering and scientific observations leads to a definition of probability as a _relative frequency_ of occurrence of some event. For example, everyone can surmise that if a fair coin is flipped several times, the side that shows up will be "heads" about half the time with good "regularity." What reason is saying here is that, if the coin is flipped many times (say $n$) and heads shows up $n_H$ times out of the $n$ flips, then
\begin{align*}\tag{1.3-2}
\lim_{n \rightarrow \infty}(n_H/n) = P(H)
\end{align*}
where $P(H)$ is interpreted as the probability of the event "heads." The ratio $n_H/n$ is the _relative frequency_ (or average number of successes) for this event. The idea of _statistical regularity_ is used to account for the fact that relative frequencies approach a fixed value (a probability) as $n$ becomes large.

__EXAMPLE 1.3-3__. In a box, there are 80 resistors, each having the same size and shape. Of the 80 resistors 18 are 10$\Omega$, 12 are 22$\Omega$, 33 are 27$\Omega$ and 17 are 47$\Omega$. Suppose the experiment is to randomly draw out one resistor from the box, each being "equally likely" to be drawn. In that case, relative frequency suggests that the following probabilities may reasonably be assumed:
\begin{align*}\tag{1}
P(\text{draw} 10\ohm) = 18/80 \qquad \qquad P(\text{draw} 22\ohm) = 12/80 \\
P(\text{draw} 27\ohm) = 33/80 \qquad \qquad P(\text{draw}47\ohm) = 17/80
\end{align*}
Since all resistors are distinct, mutually exclusive, individually and by type, and have only four types, and some resistor _must_ be chosen, the sum of the probabilities of all four events must equal 1.

Next, suppose a 22-$\Omega$ resistor is drawn from the box and _not replaced_. A second resistor is then drawn from the box. What are now the probabilities of drawing a resistor of any of the four values? Since the "population" in the box is now 79 resistors, we have, for the second drawing,
\begin{align*}\tag{2}
P(\text{draw} 10\ohm|22\ohm,) = 18/79 \qquad \qquad P(\text{draw} 22\ohm|22\ohm) = 11/79 \\
P(\text{draw} 27\ohm|22\ohm) = 33/79 \qquad \qquad P(\text{draw} 47\ohm|22\ohm) = 17/79
\end{align*}
We use the notation $P(\cdot|22\Omega)$ to note that probabilities on the second drawing are now _conditional_ on the outcome of the first drawing. This reasoning is readily extended to other drawings in sequence without replacement.

2. In a game of dice, a "shooter" can win outright if the sum of the two numbers showing up is either 7 or 11 when two dice are thrown. What is his probability of winning outright?

_Solution._

\begin{align*}
P(\{\text{ winning outright }\}) = P(E_7 \cup E_{11}) = P(E_7) + P(E_{11}) = 6/36 + 2/36 = 8/36
\end{align*}
