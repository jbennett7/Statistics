# Quartiles, deciles, and percentiles

## Problem 3-44
Find (a) the quaritles $Q_1$, $Q_2$, and $Q_3$ and (b) the deciles $D_1$, $D_2$,$\ldots$,$D_9$ for the wages of the $65$ employees at the P&R Company.

### Solution 3-44
(a) The first quartile $Q_1$ is that wage obtained by counting $\frac{N}{4} = \frac{65}{4} = 16.25$ of the cases, beginning with the first (lowest) class. Since the first class contains $8$ cases, we must take $8.25(16.25 - 8)$ of the $10$ cases from the second class. Using the method of linear interpolation, we have

\begin{align*}
Q_1 = \$259.995 + \frac{8.25}{10}(\$10.00) = \$268.25
\end{align*}

The second quartile $Q_2$ is obtained by counting the first $\frac{2N}{4} = \frac{N}{2} = \frac{65}{2} = 32.5$ of the cases. Since the first two classes comprise $18$ cases, we must take $32.5 - 18 = 14.5$ of the $16$ cases from the third class; thus

\begin{align*}
Q_2 = \$269.995 + \frac{14.5}{16}(\$10.00) = \$279.06
\end{align*}

Note that $Q_2$ is actually the median.
The third quartile $Q_3$ is obtained by counting the first $\frac{3N}{4} = \frac{3}{4}(65) = 48.75$ of the cases. Since the first four classes comprise 48 cases, we must take $48.75 - 48 = 0.75$ of the $10$ cases from the fifth class; thus

\begin{align*}
Q_3 = \$289.995 + \frac{0.75}{10}(\$10.00) = \$290.75
\end{align*}

Hence $25\%$ of the employees earn $\$268.25$ or less, $50\%$ earn $\$279.06$ or less, and $75\%$ earn $\$290.75$ or less.

(b) The first, second,$\ldots$,ninth deciles are obtained by counting $N/10, 2N/10,\ldots,9N/10$ of the cases, beginning with the first (lowest class). Thus

\begin{align}
D_1 & = \$249.995+\frac{6.5}{8}(\$10.00)=\$258.12 && D_6=\$279.995+\frac{5}{14}(\$10.00)=\$283.57 \\
D_2 & = \$259.995+\frac{5}{10}(\$10.00)=\$265.00 && D_7=\$279.995+\frac{11.5}{14}(\$10.00)=\$288.21 \\
D_3 & = \$269.995+\frac{1.5}{16}(\$10.00)=\$270.94 && D_8=\$289.995+\frac{4}{10}(\$10.00)=\$294.00 \\
D_4 & = \$269.995+\frac{8}{16}(\$10.00)=\$275.00 && D_9=\$299.995+\frac{0.5}{5}(\$10.00)=\$301.00 \\
D_5 & = \$269.995+\frac{14.5}{16}(\$10.00)=\$279.06
\end{align}

Hence $10\%$ of the employees earn $\$258.12$ or less, $20\%$ earn $\$265.00$ or less,$\ldots, 90\%$ earn $\$301.00$ or less.

Note that the fifth decile is the median. The second, fourth, sizth, and eight deciles, which divide the distribution into five equal parts and which are called _quintiles_, are sometimes used in practice.

## Problem 3-45
Determine (a) the 35th and (b) the 60th percentiles for the distribution in Problem [3.44](#Problem-3-44).

### Solution 3-45
(a) The 35th percentile, denoted by $P_{35}$, is obtained by counting the first $35N/100=35(65)/100=22.75$ of the cases, beginning with the first (lowest) class. Then, as in Problem [3.44](#Problem-3-44),
\begin{align}
P_{35}=\$269.995+\frac{4.75}{16}(\$10.00)=\$272.97$
\end{align}

This means that $35\%$ of the employees earn $\$272.97$ or less.

(b) The 60th percentile is $P_{60}=\$279.995+\frac{5}{14}(\$10.00)=\$283.57$. Note that this is the same as the sixth decile or third quintile.
