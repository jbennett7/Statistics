# The root mean square, or quadratic mean

## Problem 3-42
Find the quadratic mean of the numbers $3, 5, 6, 6, 7, 10,$ and $12$.

### Solution 3-42
\begin{align}
\text{Quadratic mean}=\text{RMS}=\sqrt{\frac{3^2+5^2+6^2+6^2+7^2+10^2+12^2}{7}}=\sqrt{57}=7.55
\end{align}

```r
load("../../Saves/centend.Rdata")
p3.42.X <- c(3, 5, 6, 6, 7, 10, 12)
cat("RMS: ", round(rms(p3.42.X), 2))
```

# Problem 3-43
Prove that the quadratic mean of two positive unequal numbers, $a$ and $b$, is greater than their geometric mean.

### Solution 3-43
We are required to show that $\sqrt{\frac{1}{2}(a^2+b^2)}\gt\sqrt{ab}$. If this is true, then by squaring both sides, $\frac{1}{2}(a^2+b^2)\gt ab$, so that $a^2+b^2\gt 2ab$, $a^2-2ab+b^2\gt 0$, or $(a-b)^2\gt 0$. But this last inqeuality is true, since the square of any real number not equal to zero must be positive.

The proof consists in establishing the reversal of the above steps. Thus starting with $(a-b)^2\gt 0$, which we know to be true, we can show that $a^2+b^2\gt 2ab$, $\frac{1}{2}(a^2+b^2)\gt ab$, and finally $\sqrt{\frac{1}{2}(a^2+b^2)}\gt\sqrt{ab}$, as required.

Note that $\sqrt{\frac{1}{2}(a^2+b^2)}=\sqrt{ab}$ is and only if $a=b$.
