# The empirical relation between the mean, median, and mode

# Problem 3.34
(a) Use the empirical formula $\text{mean} - \text{mode} = 3(\text{mean} - \text{median})$ to find the modal wage of the 65 employees at the P&R Company.

(b) Compare your result with the mode obtained in Problem [3.33](06%20The%20mode.ipynb#Problem-3-33).

### Solution 3-34
(a) From Problems [3.23](04%20The%20aritmetic%20mean.ipynb#Problem-3-23) and [3.30](05%20The%20median.ipynb#Problem-3-30) we have $\text{mean} = \$279.77$ and $\text{median} = \$277.64$

(b) From Problem [3.33](06%20The%20mode.ipynb#Problem-3-33) the modal wage is $\$277.50$, so there is good agreement with the empirical result in this case.
