# The arithmetic mean computed from grouped data

## Problem 3-19
Use the method of Problem [3.18(a)](03%20Properties%20of%20the%20arithmetic%20mean.ipynb#Problem-3-18-a) to find the arithmetic mean of the numbers $5, 8, 11, 9, 12, 6, 14,$ and $10$, choosing as the "guessed mean" $A$ the values (a) $9$ and (b) $20$.

### Solution 3-19
(a) The deviations of the given numbers from $9$ are $-4, -1, 2, 0, 3, -3, 5,$ and $1$, and the sum of the deviations is $\sum{d}=-4-1+2+0+3-3+5+1=3$. Thus

\begin{align}
\bar{X}=A+\frac{\sum{d}}{N}=9+\frac{3}{8}=9.375
\end{align}

(b) The deviations of the given numbers from $20$ are $-15, -12, -9, -11, -8, -14, -6,$ and $-10,$ and $\sum{d-85}$. Thus

\begin{align}
\bar{X}=A+\frac{\sum{d}}{N}=20+\frac{(-85)}{8}=9.375
\end{align}

## Problem 3-21
Let $d_j=X_j-A$ denote the deviations of any class mark $X_j$ in a frequency distribution from a given class mark $A$. Show that if all class intervals have equal size $c$, then (a) the deviations are all multiples of $c$ (i.e., $d_j=cu_j$, where $u_j=0,\pm1, \pm2,\ldots$) and (b) the arithmetic mean can be computed from the formula

\begin{align}
\bar{X}=A+\left(\frac{\sum{fu}}{N}\right)c
\end{align}

### Solution 3-21
<a id='Solution-3-21-a'></a>
(a)  
The result is illustrated in Table [3.2](#Table-3-2) of Problem [3.20](#Problem-3-20), where it is observed that the deviations in c0lumn 2 are all multiples of the class-interval size $c=3$ in.

To see that the result is true in general, note that if $X_1, X_2, X_3,\ldots$ are successive class markes, their common difference will for this case be equal to $c$, so that $X_2=X_1+c$, $X_3=X_1+2c$, and in general $X_j=X_1+(j-1)c$. Then any two class markes $X_p$ and $X_q$, for example, will differ by

\begin{align}
X_p-X_q=[X_1+(p-1)c]-[X_1+(q-1)c]=(p-q)c
\end{align}

which is a multiple of $c$.

<a id='Solution-3-21-b'></a>
(b)  
By part (a), the deviations of all the class marks from any given one are multiples of $c$ (i.e., $d_j=cu_j$). Then, using Problem [3.18(b)](03%20Properties%20of%20the%20arithmetic%20mean.ipynb#Problem-3-18-b), we have

\begin{align}
\bar{X}=A\frac{\sum{f_jd_j}}{N}=A+\frac{\sum{f_j(cu_j)}}{N}=A+c\frac{\sum{f_ju_j}}{N}=A+\left(\frac{\sum{fu}}{N}\right)c
\end{align}

Note that this is equivalent to the result $\bar{X}=A+c\bar{u}$, which can be obtained from $\bar{X}=A+\bar{d}$ by placing $d=cu$ and observing that $\bar{d}=c\bar{u}$ (see Problem [3.18](03%20Properties%20of%20the%20arithmetic%20mean.ipynb#Problem-3-18).
