# Properties of the arithmetic mean

## Problem 3-16
Prove that the sum of the deviations of $X_1, X_2,\ldots, X_N$ from their mean $\bar{X}$ is equal to zero.

### Solution 3-16
Let $d_1=X_1-\bar{X}, d_2=X_2-\bar{X},\ldots, d_N=X_N-\bar{X}$ be the deviations of $X_1, X_2,\ldots, X_N$ from their mean $\bar{X}$. Then

\begin{align}
\text{Sum of deviations}=\sum{d_j}=\sum{(X_j-\bar{X})}=\sum{X_j}-N\bar{X} \\
=\sum{X_j}-N\left(\frac{\sum{X_j}}{N}\right)=\sum{X_j}-\sum{X_j}=0
\end{align}

where we have used $\sum$ in place of $\sum^N_{j=1}$. We could, if desired, have omitted the subscript $j$ in $X_j$, provided that it is _understood_.

## Problem 3-17
If $Z_1=X_1+Y_1, Z_2=X_2+Y_2,\ldots,Z_N=X_N+Y_N,$ prove that $\bar{Z}=\bar{X}+\bar{Y}$.

### Solution 3-17
By definition,
\begin{align}
\bar{X}=\frac{\sum{X}}{N}\ \ \ \ \ \ \bar{Y}=\frac{\sum{Y}}{N}\ \ \ \ \ \ \bar{Z}=\frac{\sum{Z}}{N}
\end{align}
Thus
\begin{align}
\bar{Z}=\frac{\sum{Z}}{N}=\frac{\sum{(X+Y)}}{N}=\frac{\sum{X}+\sum{Y}}{N}=\frac{\sum{X}}{N}+\frac{\sum{Y}}{N}=\bar{X}+\bar{Y}
\end{align}

where the subscripts $j$ on $X, Y,$ and $Z$ have been omitted and where $\sum$ means $\sum^N_{j=1}$.

## Problem 3-18
(a) If $N$ numbers $X_1, X_2,...,X_N$ have deviations from any number $A$ given by $d_1 = X_1 - A$, 
$d_2 = X_2 - A,...,d_N = X_N - A$, respectively, prove that:

$$
\begin{align*}
\bar{X} = A + \frac{\sum_{j=1}^N{d_j}}{N} = A + \frac{\sum{d}}{N}
\end{align*}
$$


(b) In case $X_1, X_2,...,X_K$ have respective frequencies $f_1, f_2,...,f_K$ and 
$d_1 = X_1 - A,...,d_K = X_K - A$, show that the result in part (a) is replaced with

$$
\begin{align*}
\bar{X} = A + \frac{\sum_{j=1}^K{f_jd_j}}{\sum_{j=1}^K{f_j}} = A + \frac{\sum{fd}}{N}\text{   where   }
\sum_{j=1}^K{f_j} = \sum{f} = N
\end{align*}
$$



### Solution 3-18
<a id='Problem-3-18-a'></a>
(a) __First method__

\begin{align}
\bar{X} = A + \frac{\sum{X_j}}{N} = \frac{\sum(A+d_j)}{N} = \frac{\sum{A}+\sum{d_j}}{N} = \frac{NA+\sum{d_j}}{N}=
A+\frac{\sum{d_j}}{N}
\end{align}

where we have used $\sum$ in place of $\sum_{j=1}^N$ for brevity.

__Second method__  
We have $d=X-A$, or $X=A+d$, omitting the subscripts on $d$ and $X$. Thus, by Problem 3.17,

\begin{align}
\bar{X} = \bar{A}+\bar{d} = A + \frac{\sum{d}}{N}
\end{align}

since the mean of a number of constants all equal to $A$ is $A$.

<a id='Problem-3-18-b'></a>
(b)

\begin{align}
\bar{X}
= \frac{\sum_{j=1}^K{f_jX_j}}{\sum_{j=1}^K{f_j}}
= \frac{\sum{f_jX_j}}{N}
= \frac{\sum{f_j(A+d_j)}}{N}
= \frac{\sum{Af_j} + \sum{f_jd_j}}{N}
= \frac{A\sum{f_j} + \sum{f_jd_j}}{N}
= \frac{AN + \sum{f_jd_j}}{N}
= A + \frac{\sum{f_jd_j}}{N}
= A + \frac{\sum{fd}}{N}
\end{align}

Note that _formally_ the result is obtained from part (a) by replacing $d_j$ with $f_jd_j$ and summing from
$j = 1$ to $K$ instead of from $j = 1$ to $N$. The result is equialent to $\bar{X} = A + \bar{d}$, where
$\bar{d} = \left(\sum{fd}\right)/N$
