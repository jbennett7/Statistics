# Summation Notation

## Problem 3-1
Write out the terms in each of the following indicated sums:

\begin{align}
\sum^6_{j=1}{X_j}\label{p3-1a}\tag{a} \\
\sum^4_{j=1}{(Y_j-3)^2}\label{p3-1b}\tag{b} \\
\sum^N_{j=1}{a}\label{p3-1c}\tag{c} \\
\sum^5_{k=1}{f_kX_k}\label{p3-1d}\tag{d} \\
\sum^3_{j=1}{(X_j-a)}\label{p3-1e}\tag{e}
\end{align}

### Solution 3-1
(a) $X_1+X_2+X_3+X_4+X_5+X_6$  
(b) $(Y_1-3)^2+(Y_2-3)^2+(Y_3-3)^2+(Y_4-3)^2$  
(c) $a+a+a+a+\ldots+a=Na$  
(d) $f_1X_1+f_2X_2+f_3X_3+f_4X_4+f_5X_5$  
(e) $(X_1-a)+(X_2-a)+(X_3-a)=X_1+X_2+X_3-3a$

## Problem 3-2
Express each of the following by using the summation notation:  
(a) $X^2_1+X^2_2+X^2_3+\ldots+X^2_{10}$  
(b) $(X_1+Y_1)+(X_2+Y_2)+\ldots+(X_8Y_8)$  
(c) $f_1X^3_1+f_2X^3_2+\ldots+f_{20}X^3_{20}$  
(d) $a_1b_1+a_2b_2+a_3b_3+\ldots+a_Nb_N$  
(e) $f_1X_1Y_1+f_2X_2Y_2+f_3X_3Y_3+f_4X_4Y_4$

### Solution 3-2
(a) $\sum^{10}_{j=1}{X^2_j}$  

(b) $\sum^8_{j=1}{(X_j+Y_j)}$  

(c) $\sum^{20}_{j=1}{f_jX^3_j}$  

(d) $\sum^N_{j=1}{a_jb_j}$  

(e) $\sum^4_{j=1}{f_jX_jY_j}$

## Problem 3-3
Prove that $\sum^N_{j=1}{(aX_j+bY_j-cZ_j)}=a\sum^N_{j=1}{X_j}+b\sum^N_{j=1}{Y_j}-c\sum^N_{j=1}{Z_j}, where $a, b,$ and $c$ are any constants.

### Solution 3-3
\begin{align}
\sum^N_{j=1}{(aX_j+bY_j-cZ_j)}=(aX_1+bY_1-cZ_1)+(aX_2+bY_2-cZ_2)+\ldots+(aX_N+bY_N-cZ_N) \\
=(aX_1+aX_2+\ldots+aX_N)+(bY_1+bY_2+\ldots+bY_N)-(cZ_1+cZ_2+\ldots+cZ_N) \\
=a(X_1+X_2+\ldots+X_N)+b(Y_1+Y_2+\ldots+Y_N)-c(Z_1+Z_2+\ldots+Z_N) \\
=a\sum^N_{j=1}{X_j}+b\sum^N_{j=1}{Y_j}-c\sum^N_{j=1}{Z_j}
\end{align}

or brefly, $\sum{(aX+bY-cZ)}=a\sum{X}+b\sum{Y}-c\sum{Z}$.

## Problem 3-4
Two variables, $X$ and $Y$, assume the values $X_1=2, X_2=-5, X_3=4, X_4=-8$ and $Y_1=-3, Y_2=-8, Y_3=10, Y_4=6$, respectively. Calculate (a) $\sum{X}$, (b) $\sum{Y}$, (c) $\sum{XY}$, (d) $\sum{X^2}$, (e) $\sum{Y^2}$, (f) $(\sum{X})(\sum{Y})$, (g) $\sum{XY^2}$, and (h) $\sum{(X+Y)(X-Y)}$.

### Solution 3-4
Note that in each case the subscript $j$ on $X$ and $Y$ has been omitted and $\sum$ is understood as $\sum^4_{j=1}$. Thus $\sum{X}$, for example, is short for $\sum^4_{j=1}{X_j}$.

(a) $\sum{X}=(2)+(-5)+(4)+(-8)=2-5+4-8=-7$  

(b) $\sum{Y}=(-3)+(-8)+(10)+(6)=-3-8+10+6=5$  

(c) $\sum{XY}=(2)(-3)+(-5)(-8)+(4)(10)+(-8)(6)=-6+40+40-48=26$  

(d) $\sum{X^2}=(2)^2+(-5)^2+(4)^2+(-8)^2=4+25+16+64=109$  

(e) $\sum{Y^2}=(-3)^2+(-8)^2+(10)^2+(6)^2=9+64+100+36=209$  

(f) $(\sum{X})(\sum{Y})=(-7)(5)=-35$, using parts (a) and (b). Note that $(\sum{X})(\sum{Y})\ne\sum{XY}$.  

(g) $\sum{XY^2}=(2)(-3)^2+(-5)(-8)^2+(4)(10)^2+(-8)(6)^2=-190$  

(h) $\sum{(X+Y)(X-Y)}=\sum{(X^2-Y^2)}=\sum{X^2}-\sum{Y^2}=109-209=-100$, using parts (d) and (e).
