# The Mode
`R` does not have a function to calculate the mode. Here is a version, from the [text](../01%20Text/09%20The%20mode.ipynb#fun1) that can be used.

```r
mode <- function(x) {
  unique_x <- unique(x)
  tabulate_x <- tabulate(match(x, unique_x))
  ret_val <- unique_x[tabulate_x == max(tabulate_x)]
  if(length(ret_val) == length(x)){
      return(NA)
  } else {
      return(ret_val)
  }
}
```

Find the mean, median, and mode for each set of numbers: (a) $7$, $4$, $10$, $9$, $15$, $12$, $7$, $9$, $7$ and (b) $8$, $11$, $4$, $3$, $2$, $5$, $10$, $6$, $4$, $1$, $10$, $8$, $12$, $6$, $5$, $7$.

## Problem-3-80
Find the mode of the maximum loads of cables of Problem [(3.59)](02%20The%20arithmetic%20mean.ipynb#Problem-3-59).

### Solution-3-80
From a frequency distribution or histogram the mode can be obtained from the formula [(9)](../01%20Text/09%20The%20mode.ipynb#eq9)
\begin{align}
\text{Mode} = L_1 + \left(\frac{\Delta_1}{\Delta_1+\Delta_2}\right)c
\end{align}  
\begin{align}
\text{where } L1 &= \text{lower class boundary of the modal class (i.e., the class containing the mode)} \\
\Delta_1 &= \text{excess of modal frequency over frequency of next-lower class} \\
\Delta_2 &= \text{excess of modal frequency over frequency of next-higher class} \\
c &= \text{sie of the modal class interval}
\end{align}

```r
load("../../Saves/Table_3.8.Rdata")
p3.80 <- new.env()
p3.80$X.lower <- Table.3.8$Max.Load.low
p3.80$X.upper <- Table.3.8$Max.Load.high
p3.80$f <- Table.3.8$Number.of.Cables
p3.80$N <- sum(p3.80$f)
p3.80$c <- p3.80$X.upper[1]-p3.80$X.lower[1]
p3.80$idx <- which.max(p3.80$f)
p3.80$L1 <- p3.80$X.lower[p3.80$idx]
p3.80$D1 <- abs(p3.80$f[p3.80$idx]-p3.80$f[(p3.80$idx-1)])
p3.80$D2 <- abs(p3.80$f[p3.80$idx]-p3.80$f[(p3.80$idx+1)])
p3.80$mode <- p3.80$L1+p3.80$D1/(p3.80$D1+p3.80$D2)*p3.80$c
cat("The mode of the maximum loads of cables:", p3.80$mode)
rm(p3.80)
rm(Table.3.8)
```
