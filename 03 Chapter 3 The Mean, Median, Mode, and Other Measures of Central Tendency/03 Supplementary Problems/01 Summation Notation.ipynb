## Problem 3-47
Write the terms in each of the following indicated sums:

(a) $\sum^4_{j=1}{(X_j+2)}$&ensp;&ensp; (c) $\sum^3_{j=1}{U_j(U_j+6)}$&ensp;&ensp; (e) $\sum^4_{j=1}{4X_jY_j}$  

(b) $\sum^5_{j=1}{f_jX^2_j}$&ensp;&ensp;&ensp;&ensp;&ensp; (d) $\sum^N_{k=1}{(Y^2_k-4)}$

### Solution 3-47
(a) $\sum^4_{j=1}{(X_j+2)}=(X_1+2)+(X_2+2)+(X_3+2)+(X_4+2)$

(c) $\sum^3_{j=1}{U_j(U_j+6)}=(U_1(U_1+6))+(U_2(U_2+6))+(U_3(U_3+6))$  

(d) $\sum^N_{j=1}(Y^2_k-4)=(Y^2_1-4)+(Y^2_2-4)+\ldots+(Y^2_N-4)$  

(e) $\sum^4_{j=1}(4X_jY_j)=(4X_1Y_1)+(4X_2Y_2)+(4X_3Y_3)+(4X_4Y_4)$

## Problem 3-48
Express each of the following by using the summation notation.  

(a) $(X_1+3)^3+(X_2+3)^3+(X_3+3)^3$  

(b) $f_1(Y_1-a)^2+f_2(Y_2-a)^2+\ldots+f_{15}(Y_{15}-a)^2$

(d) $(\frac{X_1}{Y_1}-1)^2+(\frac{X_2}{Y_2}-1)^2+\ldots+(\frac{X_8}{Y_8}-1)^2$  

(e) $\frac{f_1a^2_1+f_2a^2_2+\ldots+f_{12}a^2_{12}}{f_1+f_2+\ldots+f_{12}}$

(a) $\sum^3_{j=1}{(X_j+3)^3}$

(c) $\sum^N_{j=1}{(2X_j-3Y_j)}$  

(d) $\sum^8_{j=1}{(\frac{X_j}{Y_j}-1)^2}$  

(e) $\frac{\sum^{12}_{j=1}{f_ja^2_j}}{\sum^{12}_{j=1}{f_j}}$

## Problem 3-49
Prove that $\sum^N_{j=1}{(X_j-1)^2}=\sum^N_{j=1}{X^2_j}-2\sum^N_{j=1}{X_j+N}$.

### Solution 3-49

\begin{align}
\sum^N_{j=1}{(X_j-1)^2} &=
\sum^N_{j=1}{(X_j-1)(X_j-1)} =
\sum^N_{j=1}{(X^2_j-2X_j+1)}=
\sum^N_{j=1}{X^2_j}-2\sum^N_{j=1}{X_j}+\sum^N_{j=1}{1} \\
\\
\\
&= \sum^N_{j=1}{X^2_j}-2\sum^N_{j=1}{X_j}+N
\end{align}

## Problem 3-50
Prove that $\sum{(X+a)(Y+b)}=\sum{XY}+a\sum{Y}+b\sum{X}+Nab$.

### Solution 3-50

## Problem 3-51
The variables, $U$ and $V$, assume the values $U_1=3$, $U_2=-2$, $U_3=5$, and $V_1=-4$, $V_2=-1$, $V_3=6$, respectively. Calculate
(a) $\sum{UV}$, (b) $\sum(U+3)(V-4)$, (c) $\sum{V^2}$, (d) $(\sum{U})(\sum{V})^2$, (e) $\sum{UV^2}$, (f) $\sum{(U^2-2V^2+2)}$, and (g) $\sum{(\frac{U}{V})}$.

```r
U <- c(3,-2,5)
V <- c(-4,-1,6)
a <- 0; b <- 0; c <- 0; d <- 0; e <- 0; f <- 0; g <- 0
for(i in 1:3){
    a <- a + U[i]*V[i]
    b <- b + (U[i]+3)*(V[i]-4)
    c <- c + V[i]^2
    d <- sum(U)+sum(V)^2
    e <- e + U[i]*V[i]^2
    f <- f + (U[i]^2-2*V[i]^2+2)
    g <- g + U[i]/V[i]
}
cat ("(a) ", a, "\n(b) ", b, "\n(c) ", c, "\n(d)", d, "\n(e) ", e, "\n(f) ", f, "\n(g) ", g)
```

## Problem 3-52
Given $\sum^4_{j=1}{X_j}=7$, $\sum^4_{j=1}{Y_j}=-3$, and $\sum^4_{j=1}{X_iY_j}=5$, find (a) $\sum^4_{j=1}{(2X_j+5Y_j)}$ and (b) $\sum^4_{j=1}(X_j-3)(2Y_j+1)$.

### Solution 3-52
(a)
\begin{align}
\sum^4_{j=1}{(2X_j+5Y_j)} = 2\sum{X} + 5\sum{Y} = (2)(7) + (5)(-3) = 14-15 = -1
\end{align}
(b)
\begin{align}
\sum^4_{j=1}{(X_j-3)(2Y_j+1)} = \sum{(2XY - 6Y + X - 3)} = 2\sum{XY} - 6\sum{Y} + \sum{X} - (3)(4) = (2)(5) - (6)(-3) + 7 - 12 =
10 + 18 + 7 - 12 = 23
\end{align}
