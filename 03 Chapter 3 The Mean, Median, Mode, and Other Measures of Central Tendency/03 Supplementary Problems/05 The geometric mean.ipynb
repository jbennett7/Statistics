# The geometric mean

```r
geometric.mean <- function(x){
    exp(mean(log(x[x>0])))
}
```

## Problem-3-88
Find the geomteric mean of the numbers (a) $4.2$ and $16.8$ and (b) $3.00$ and $6.00$.

### Solution-3-88

```r
p3.88 <- new.env()
p3.88$X.a <- c(4.2,16.8)
p3.88$X.b <- c(3.00, 6.00)
cat("(a) the geometric mean for (", paste(p3.88$X.a, collapse=", "), ") is ",
    (p3.88$G.a <- geometric.mean(p3.88$X.a)), "\n",
    "(b) the geometric mean for (", paste(p3.88$X.b, collapse=", "), ") is ",
    round(p3.88$G.b <- geometric.mean(p3.88$X.b), 2), sep="")
rm(p3.88)
```

## Problem-3-89
Find (a) the geometric mean $G$ and (b) the arithmetic mean $\bar{X}$ of the set $2$, $4$, $8$, $16$, $32$.

```r
p3.89 <- new.env()
p3.89$X <- c(2,4,8,16,32)
p3.89$G <- geometric.mean(p3.89$X)
p3.89$X.bar <- mean(p3.89$X)
cat(" (a) the geometric mean for", p3.89$x, "is: ", p3.89$G, "\n",
    "(b) the arithmetic mean for the same set is", p3.89$X.bar, "\n")
rm(p3.89)
```

## Problem-3-90
Find the geometric mean of the sets (a) $3$, $5$, $8$, $3$, $7$, $2$ and (b) $28.5$, $73.6$, $47.2$, $31.5$, $64.8$.

```r
p3.90 <- new.env()
p3.90$X.a <- c(3, 5, 8, 3, 7, 2)
p3.90$X.b <- c(28.5, 73.6, 47.2, 31.5, 64.8)
p3.90$G.a <- geometric.mean(p3.90$X.a)
p3.90$G.b <- geometric.mean(p3.90$X.b)
cat("(a) the geometric mean for ", paste(p3.90$X.a, collapse=", "), " is ",
    round(p3.90$G.a,1), "\n",
    "(b) the geometric mean for ", paste(p3.90$X.b, collapse=", "), " is ",
    round(p3.90$G.b,2), sep="")
```
