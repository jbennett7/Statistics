# The geometric mean

```r
geometric.mean <- function(x){
    exp(mean(log(x[x>0])))
}
weighted.geometric.mean <- function(x,w){
    exp(weighted.mean(log(x[x>0]),w))
}
```

## Problem-3-88
Find the geomteric mean of the numbers (a) $4.2$ and $16.8$ and (b) $3.00$ and $6.00$.

### Solution-3-88-a

```r
p3.88a <- new.env()
p3.88a$X <- c(4.2,16.8)
p3.88a$G <- geometric.mean(p3.88a$X)
options(repr.plot.width=16, repr.plot.height=6)
plot(data.frame(c(p3.88a$X),1),type='o',pch='|',ylab='',xlab='',cex=2)
points(data.frame(p3.88a$G,1),type='o',pch='x',cex=2)
rm(p3.88a)
```

### Solution 3-88-b

```r
p3.88b <- new.env()
p3.88b$X <- c(3.00, 6.00)
p3.88b$G <- round(geometric.mean(p3.88b$X),2)
plot(data.frame(c(p3.88b$X),1),type='o',pch='|',ylab='',xlab='',cex=2)
points(data.frame(p3.88b$G,1),type='o',pch='x',cex=2)
cat("(b) The geometric mean of ", paste(p3.88b$X,collapse=", "), " is ", p3.88b$G, sep='')
rm(p3.88b)
```

## Problem-3-92

\begin{align}
A=P(1+r)^n
\end{align}

\begin{align}
2 = 1(1+r)^4
\end{align}

\begin{align}
\sqrt[4]{2} = 1 + r
\end{align}

\begin{align}
1.189 = 1 + r
\end{align}
\begin{align}
