# The root mean square, or quadratic mean

```r
rms <- function(x) {
    return(sqrt(mean(x^2)))
}
```

## Problem 3-104
Find the RMS (or quadratic mean) of the numbers (a) 11, 23, and 35 and (b) 2.7, 3.8, 3.2, and 4.3.

### Solution

```r
p3.104 <- new.env()
p3.104$a <- c(11,23,35)
p3.104$b <- c(2.7,3.8,3.2,4.3)
cat("(a): ", rms(p3.104$a), "\n")
cat("(b): ", rms(p3.104$b), "\n")
rm(p3.104)
```
