# The harmonic mean

```r
harmonic.mean <- function(x) {
    return(length(x)/sum(x^(-1)))
}
```

```r
geometric.mean <- function(x) {
    return(exp(mean(log(x))))
}
```

## Problem-3-97
Find the harmonic mean of the numbers (a) $2, 3,$ and $6$ (b) $3.2, 5.2, 4.8, 6.1,$ and $4.2$.

### Solution-3-97-a

```r
p3.97a <- new.env()
p3.97a$X <- c(2,3,6)
p3.97a$H <- harmonic.mean(p3.97a$X)
cat("(a) The harmonic mean of ", paste(p3.97a$X,collapse=", "), " is: ", p3.97a$H, sep="")
options(repr.plot.width=16, repr.plot.height=6)
plot(data.frame(c(p3.97a$X),1),type='o',pch='|',ylab='',xlab='',cex=2)
points(data.frame(p3.97a$H,1),type='o',pch='x', col="red", cex=3)
rm(p3.97a)
```

### Solution-3-97-b

```r
p3.97b <- new.env()
p3.97b$X <- c(3.2, 5.2, 4.8, 6.1, 4.2)
p3.97b$H <- round(harmonic.mean(p3.97b$X),2)
cat("(b) The harmonic mean of ", paste(p3.97b$X,collapse=", "), " is: ", p3.97b$H, sep="")
options(repr.plot.widt=16, repr.plot.height=6)
plot(data.frame(c(p3.97b$X),1),type='o',pch='|',ylab='',xlab='',cex=2)
points(data.frame(p3.97b$H,1),type='o',pch='x',col='red',cex=3)
rm(p3.97b)
```

## Problem-3-98
Find the (a) arithmetic mean, (b) geometric mean, and (c) harmonic mean of the numbers $0, 2, 4,$ and $6$.

### Solution-3-98

```r
p3.98 <- new.env()
p3.98$X <- c(0,2,4,6)
p3.98$X.bar <- mean(p3.98$X)
p3.98$G <- geometric.mean(p3.98$X)
p3.98$H <- harmonic.mean(p3.98$X)
cat("(a) The arithmetic mean of ", paste(p3.98$X,collapse=", "), " is: ", p3.98$X.bar,
    ",\n(b) the geometric mean is: ", p3.98$G,
    ", and\n(c) the harmonic mean is: ", p3.98$H, sep="")
```

## Problem-3-99
If $X_1, X_2, X_3,\ldots$ represent the class marks in a frequency distribution with corresponding class frequencies $f_1, f_2, f_3,\ldots$, respectively, prove that the harmonic mean $H$ of the distribution is given by

\begin{align}
\frac{1}{H} = \frac{1}{N}\left(\frac{f_1}{X_1}+\frac{f_2}{X_2}+\frac{f_3}{X_3}+\ldots\right) = 
\frac{1}{N}\sum{\frac{f}{X}}
\end{align}

where $N = f_1+f_2+\ldots=\sum{f}$.

### Solution-3-99
The harmonic mean is given in equation 13 as

\begin{align}
\frac{1}{H}=\frac{\sum{\frac{1}{X}}}{N}=\frac{1}{N}\sum{\frac{1}{X}}
\end{align}

If the numbers $X_1, X_2, X_3,\dots$ occur $f_1, f_2, f_3,\ldots$ times, respectively, the harmonic mean is then
given by the above equation.
