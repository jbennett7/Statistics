## Summary Statistics
R includes a variety of functions for calculating summary statistics.

```r
X <- c(1, 2, 3, 4, 5)
mean(X)
median(X)
min(X)
max(X)
```

Each of the above functions has an argument `na.rm` for dealing with `NA` values. The default behavior is to return `NA` if any values are `NA`. If `na.r=TRUE`, then `NA` values are ignored.

```r
X <- c(1, 2, 3, 4, 5, NA)
mean(X)
mean(X, na.rm=TRUE)
```

Outliers can be trimed from the dataset as well using the `trim` argument.

```r
mean(c(-1, 0:100, 2000))
mean(c(-1, 0:100, 2000), trim=0.1)
```

## The mode in R
There is no mode method in R. A reasonable version is shown below.

```r
mode <- function(x) {
  unique_x <- unique(x)
  tabulate_x <- tabulate(match(x, unique_x))
  ret_val <- unique_x[tabulate_x == max(tabulate_x)]
  if(length(ret_val) == length(x)){
      return(NA)
  } else {
      return(ret_val)
  }
}
```

```r
X <- c(2, 2, 5, 7, 9, 9, 9, 10, 10, 11, 12, 18)
mode(X)
```

```r
X <- c(3, 5, 8, 10, 12, 15, 16)
mode(X)
```

```r
X <- c(2, 3, 4, 4, 4, 5, 5, 7, 7, 7, 9)
mode(X)
```

## The arithmetic mean computed from grouped data
As stated the _coding method_ should always be used.

__EXAMPLE__

Using Table [2.1](../../02%20Chapter%202%20Frequency%20Distributions/01%20Text/03%20Frequency%20Distributions.ipynb#Table-2-1).

```r
load('../../Saves/Table_2.1.Rdata')
X <- c(61, 64, 67, 70, 73)
Table.2.1.mark <- matrix(c(X, Table.2.1), nrow=nrow(Table.2.1))
rownames(Table.2.1.mark)<-rownames(Table.2.1)
colnames(Table.2.1.mark)<-c('X', 'f')
Table.2.1.mark
```

### The coding method

```r
c <- 3 # class interval size
t.length <- length(Table.2.1.mark[,1])
A.idx <- 3 # guessed mean index
A <- Table.2.1.mark[A.idx,1]
u <- seq(1, t.length, by=1)-A.idx
f <- Table.2.1.mark[,2] # frequency
N <- sum(f) # summation of frequencies
fu <- f*u
fu.sum <- sum(fu)
(X.bar <- A + (fu.sum/N)*c)
```

```r
grouped.mean <- function(X, f) {
    if(var(diff(X)) == 0) { # coding method
        idx <- ceiling(length(X)/2)
        A <- X[idx]
        N <- sum(f)
        u <- seq(1, length(X), by=1)-idx
        fu.sum <- sum(f*u)
        return(as.numeric(A+(fu.sum/N)*c))
    } else { # Long method
        fx.sum <- sum(f*X)
        N <- sum(f)
        return(as.numeric(fx.sum/N))
    }
}
```

```r
grouped.mean(Table.2.1.mark[,1], Table.2.1.mark[,2])
```

## Geometric mean in R

```r
geometric.mean <- function(x) {
    return(exp(mean(log(x))))
}
```

```r
X <- 2^(1:301)
geometric.mean(X)
```

The geometric mean shown above does not work with really large number sets.

```r
geometric.mean <- function(x) {
    return(prod(x)^(1/length(x)))
}
X <- 2^(1:301)
geometric.mean(X)
```

## Harmonic mean in R

```r
harmonic.mean <- function(x) {
    return(length(x)/sum(x^(-1)))
}
```

```r
X <- c(2, 4, 8)
round(harmonic.mean(X), 2)
```

## Root mean square in R

```r
rms <- function(x) {
    return(sqrt(mean(x^2)))
}
```

```r
X <- c(1, 3, 4, 5, 7)
round(rms(X), 2)
```
