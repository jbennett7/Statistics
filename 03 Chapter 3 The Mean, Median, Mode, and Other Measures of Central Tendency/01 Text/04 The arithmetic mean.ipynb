# The arithmetic mean

The _arithmetic mean_, or briefly the _mean_, of a set of $N$ numbers $X_1, X_2, X_3,\ldots, X_N$ is deonted by $\bar{X}$ (read "$X$ bar") and is defoned as

<a id=eq1></a>
\begin{align}\label{eq1}\tag{1}
\bar{X}=\frac{X_1+X_2+X_3+\ldots+X_N}{N}=\frac{\sum^N_{j=1}{X_j}}{N}=\frac{\sum{X}}{N}
\end{align}

__EXAMPLE 4.__ The arithmetic mean of the numbers $8, 3, 5, 12,$ and $10$ is

\begin{align}
\bar{X}=\frac{8+3+5+12+10}{5}=\frac{38}{5}=7.6
\end{align}

If the numbers $X_1, X_2,\ldots, X_K$ occur $f_1, f_2,\ldots,f_K$ times, respectively (i.e., occur with frequencies $f_1, f_2,\ldots, f_K$), the arithmetic mean is

<a id=eq2></a>
\begin{align}\label{eq2}\tag{2}
\bar{X}=\frac{f_1X_1+f_2X_2+\ldots+f_KX_K}{f_1+f_2+\ldots+f_K}=\frac{\sum^K_{j=1}{f_jX_j}}{\sum^K_{j=1}{f_j}}=\frac{\sum{fX}}{\sum{f}}=\frac{\sum{f_X}}{N}
\end{align}

where $N=\sum{f}$ is the _total frequency (i.e., the total number of cases).

__EXAMPLE 5.__ If $5, 8, 6, $ and $2$ occur with frequencies $3, 2, 4, $ and $1$, respectively, the arithmetic mean is


\begin{align}
\bar{X}=\frac{(3)(5)+(2)(8)+(4)(6)+(1)(2)}{3+2+4+1}=\frac{15+16+24+2}{10}=5.7
\end{align}
