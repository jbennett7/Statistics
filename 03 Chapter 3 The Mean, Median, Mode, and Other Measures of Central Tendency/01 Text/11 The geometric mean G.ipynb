# The geometric mean G
The geometric mean $G$ of a set of $N$ positive numbers $X_1, X_2, X_3,\ldots, X_N$ is the $N$th root of the product of the numbers:

<a id=eq11></a>
\begin{align}\label{eq11}\tag{11}
G=\sqrt[N]{X_1X_2X_3\ldots X_N}
\end{align}

__EXAMPLE 13.__ The geometric mean of the numbers $2, 4,$ and $8$ is $G=\sqrt[3]{(2)(4)(8)}=\sqrt[3]{64}=4$.

We can compute $G$ by logarithms (see Problem [3.35](../02%20Solved%20Problems/08%20%20The%20geometric%20mean.ibynb#Problem-3-35) or by using a calculator. The geometric mean can also be expressed as the exponential mean of logarithms. This is the method I use for creating the geometric mean for R

## Geometric mean in R

```r
geometric.mean <- function(x) {
    return(exp(mean(log(x))))
}
```

```r
X <- 2^(1:301)
geometric.mean(X)
```

The geometric mean shown above does not work with really large number sets.

```r
geometric.mean <- function(x) {
    return(prod(x)^(1/length(x)))
}
X <- 2^(1:301)
geometric.mean(X)
```
