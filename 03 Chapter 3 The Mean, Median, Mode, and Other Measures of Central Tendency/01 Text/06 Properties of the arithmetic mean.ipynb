# Properties of the arithmetic mean
1. The algebraic sum of the deviations of a set of numbers from their arithmetic mean is zero.

__EXAMPLE 7.__ The deviations of the numbers $8, 3, 5, 12$ and $10$ from their arithmetic mean $7.6$ are $8-7.6, 3-7.6, 5-7.6, 12-7.6,$ and $10-7.6$, or $0.4, -4.6, -2.6, 4.4,$ and $2.4$, with algebraic sum $0.4-4.6-2.6+4.4+2.4=0$.

2. The sum of the squares of the deviations of a set of numbers $X_j$ from any number $a$ is a minimum if and only if $a-\bar{X}$ (see [Problem 4.27](../../04%20Chapter%204%20The%20Standard%20Deviation%20and%20Other%20Measures%20of%20Dispersion/02%20Solved%20Problems/10%20Properties%20of%20the%20standard%20deviation.ipynb#Problem-4-27)).
3. If $f_1$ numbers have mean $m_1$, $f_2$ numbers have mean $m_2,\ldots, f_K$ numbers have mean $m_k$, then the mean of all the numbers is

<a id=eq4></a>
\begin{align}\label{eq4}\tag{4}
\bar{X}=\frac{f_1m_1+f_2m_2+\ldots+f_Km_k}{f_1+f_2+\ldots+f_K}
\end{align}

4. If $A$ is any _guessed_ or _assumed arithmetic mean_ (which may be any number) and if $d_j=X_j-A$ are the deviations of $X_j$ from $A$, then equations ([<u>1</u>](04%20The%20arithmetic%20mean.ipynb#eq1)) and ([<u>2</u>](04%20The%20arithmetic%20mean.ipynb#eq2)) become, respectively,

<a id=eq5></a>
\begin{align}\label{eq5}\tag{5}
\bar{X}=A+\frac{\sum^N_{j=1}{d_j}}{N}=A+\frac{\sum{d}}{N}
\end{align}

<a id=eq6></a>
\begin{align}\label{eq6}\tag{6}
\bar{X}=A+\frac{\sum^K_{j=1}{f_jd_j}}{\sum^K_{j=1}{f_j}}=A+\frac{\sum{fd}}{N}
\end{align}

where $N=\sum^K_{j=1}{f_j}=\sum{f}$. Note that formulas $\eqref{eq5}$ and $\eqref{eq6}$ are summarized in the equation $\bar{X}=A+\bar{d}$ (see [Problem 3.18](../02%20Solved%20Problems/03%20Properties%20of%20the%20arithmetic%20mean.ipynb#Problem-3-18)).
