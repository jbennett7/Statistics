# Index, or subscript, notation
Let the symbol $X_j$ (read "$X$ sub $j$") denote any of the $N$ _values_ $X_1, X_2, X_3,\ldots,X_N$ assumed by a variable $X$. The letter $j$ in $X_j$, which can stand for any of the numbers $1, 2, 3,\ldots,N$ is called a _subscript_, or _index_. Clearly any letter other than $j$, such as $i, k, p, q,$ or $s$, could have been used as well.
