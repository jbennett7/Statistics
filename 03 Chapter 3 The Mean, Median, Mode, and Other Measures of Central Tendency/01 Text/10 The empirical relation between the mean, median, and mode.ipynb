# The empirical relation between the mean, median, and mode

```r
getmode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}
x.left <- round(rbeta(10000, 2, 5), 2)
x.left.mean <- round(mean(x.left), 2)
x.left.median <- round(median(x.left), 2)
x.left.mode <- getmode(x.left)
x.right <- round(rbeta(10000, 5, 2), 2)
x.right.mean <- round(mean(x.right), 2)
x.right.median <- round(median(x.right), 2)
x.right.mode <- getmode(x.right)
x.lim <- c(0, 1)
```

For unimodal frequency curves that are moderately skewed (asymmetrical), we have the empirical relation

<a id=eq10></a>
\begin{align}\label{eq10}\tag{10}
\text{Mean}-\text{mode}=3(\text{mean}-\text{median})
\end{align}

Figures [3-1](#Figure-3-1) and [3-2](#Figure-3-2) show the relative positions of the mean, median, and mode for frequency curves skewed to the right and left, respectively. For symetrical curves, the mean, mode, and median all coincide.

### Figure 3-1

```r
options(repr.plot.width=18, repr.plot.height=6)
plot(density(x.left, bw=.1), xlim=x.lim, main="Fig. 3-1: Relative positions of mode(green), median(blue), mean(red) for a right-skewed frequency curve")
abline(v=x.left.mean, col='red')
abline(v=x.left.median, col='blue')
abline(v=x.left.mode, col='green')
```

### Figure 3-2

```r
options(repr.plot.width=18, repr.plot.height=6)
plot(density(x.right,bw=.1), xlim=x.lim, main="Fig. 3-2: Relative positions of mode(green), median(blue), and mean(red), for a left-skewed frequency curve")
abline(v=x.right.mean, col='red')
abline(v=x.right.median, col='blue')
abline(v=x.right.mode, col='green')
```
