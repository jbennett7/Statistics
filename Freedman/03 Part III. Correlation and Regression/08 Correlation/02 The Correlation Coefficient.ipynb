# The Correlation Coefficient

```r
correlation.coeff <- function(n, mean, sd, c) {
    a <- c
    b <- 1-c
    x <- rnorm(n, mean=mean, sd=sd)
    cor.l <- c()
    for (i in 1:length(a)){
        y <- a[i]*x + b[i]*rnorm(n, mean=b[i]*mean, sd=sd)
        cor.l <- c(cor.l, round(cor(x, y), 2))
    }
    data.frame(c=c, cor=cor.l)
}
df <- correlation.coeff(50, 3, 1, seq(0, 1, .01))
x <- df$c
y <- df$cor
plot(x, y, pch=19, cex=.4)
```

```r
scatter.plot <- function(n, mean, sd, c) {
    a <- c
    b <- 1-c
    x <- rnorm(n, mean=mean, sd=sd)
    y <- a*x + b*rnorm(n, mean=b*mean, sd=sd)
    m <- round(cor(x, y), 2)
    limits <- c(mean-3*sd, mean+3*sd)
    plot(x, y,  main=paste("Corr Coeff =", m), pch=19, cex=.4, xlab="", ylab="", xlim=limits, ylim=limits)
}
```

```r
set.seed(1)
obs <- 50
limits <- c(-2, 8)
a <- 1
b <- 0
x    <- rnorm(obs, mean = 3, sd = 1)
y.0  <- rnorm(obs, mean = 3, sd = 1)
y.4  <- a*x + b*rnorm(obs, sd=.5)
y.6  <- x + rnorm(obs, sd=.83)
y.8  <- x + rnorm(obs, sd=.58)
y.9  <- x + rnorm(obs, sd=.4)
y.95 <- x + rnorm(obs, sd=.32)
m.0  <- round(cor(x, y.0), 2)
m.4  <- round(cor(x, y.4), 2)
m.6  <- round(cor(x, y.6), 2)
m.8  <- round(cor(x, y.8), 2)
m.9  <- round(cor(x, y.9), 2)
m.95 <- round(cor(x, y.95), 2)
```

The correlation coefficient is a measure of linear association, or clustering around a line. The relationship between two variables can be summarized by:
* the average of the $x$-values, the SD of the $x$-values,
* the average of the $y$-values, the SD of the $y$-values,
* the correlation coefficient $r$.

```r
par(mfrow=c(3, 2))
options(repr.plot.width=8, repr.plot.height=12)
plot(x, y.0,  main=paste("Corr Coeff =", m.0), pch=19, cex=.4, xlab="", ylab="", xlim=limits, ylim=limits)
plot(x, y.4,  main=paste("Corr Coeff =", m.4), pch=19, cex=.4, xlab="", ylab="", xlim=limits, ylim=limits)
plot(x, y.6,  main=paste("Corr Coeff =", m.6), pch=19, cex=.4, xlab="", ylab="", xlim=limits, ylim=limits)
plot(x, y.8,  main=paste("Corr Coeff =", m.8), pch=19, cex=.4, xlab="", ylab="", xlim=limits, ylim=limits)
plot(x, y.9,  main=paste("Corr Coeff =", m.9), pch=19, cex=.4, xlab="", ylab="", xlim=limits, ylim=limits)
plot(x, y.95, main=paste("Corr Coeff =", m.95), pch=19, cex=.4, xlab="", ylab="", xlim=limits, ylim=limits)
```

The correlation coefficient&ndash;six positive values. The diagrams are scaled so that the average equals $3$ and the SD equals $1$, horizontally and vertically; there are $100$ points in each diagram. Clustering is measured by the correlation coefficient.
