# The Scatter Diagram

We will be looking at Galton's data set of Father-Son heights.

```r
pearson <- read.csv('../../09 Data Sets/pearsonheightdata.csv')
```

```r
old.repr.plot.width=getOption("repr.plot.width")
old.repr.plot.height=getOption("repr.plot.height")
options(repr.plot.width=12, repr.plot.height=12)
plot(pearson$father, pearson$son, pch=20, cex=.6, xlim=c(58, 80), ylim=c(58, 80))
abline(a=0, b=1, lty="dashed")
abline(v=72.5, lty="dotted")
abline(v=71.5, lty="dotted")
```

Scatter diagram for heights of $1,078$ fathers and sons. Shows positive association between son's height and father's height. Families where the height of the son equals the height of the father are plotted along the 45-degree line $y=x$. Families where the father is $72$ inches tall (to the nearest inch) are plotted in the vertical strip.
