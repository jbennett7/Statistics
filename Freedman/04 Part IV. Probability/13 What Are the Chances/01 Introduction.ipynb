# Introduction

Defining what is meant by the word _chance_ is very hard in mathematical terms. There are some carefully thought-out theories, but the theories only cover a small set of what ordinary people call chance. Freedman <sup>[\[1\]](../../../README.md#ref4)</sup> presents _frequency theory_ version. This works best for processes that can be repeated under the same conditions.

A simple game of chance is the coin toss. A coin is tossed repeatedly, and the result (either _heads_ or _tails_) is recorded. The chance of getting either _heads_ or _tails_ is $50\%$.  

Another game of chance is rolling dice. A die is rolled repeatedly. There are six possible results, all equal (1 in 6) or $16 \frac{2}{3}\%$ chance.

__DEFINITION__  
The chance of something gives the percentage of time it is expected to happen when the basic process is done repeatedly, independently, and under the same conditions. If something is impossible, it has a $0\%$ chance of happening. If something is inevitable, it has $100\%$ of happening. This means that chances can only lie between $0\%$ and $100\%$.

The complement of the chance of an event happening is mathematically represented as $100\%$ minus the chance of the event happening.

_Example 1_. A box contains red marbles and blue marbles. One marble is drawn at random from the box (each marble has an equal chance to be drawn). If it is red, you win \$1. If it is blue, you win nothing. You can choose between two boxes:
* box A contains 3 red marbles and 2 blue ones;
* box B contains 30 red marbles and 20 blue ones.

Which box offers a better chance of winning, or are they the same?

_Solution_. The two boxes offer the same chance of winning, 3 in 5.
\begin{align*}
\text{Box A} &= \frac{3\text{ red marbles}}{5\text{ total marbles}} \\
\\
\text{Box B} &= \frac{30\text{ red marbles}}{50\text{ total marbles}} = \frac{3}{5}
\end{align*}
The ratios are the same.

As we move forward, there will be some distinctions made about the game of chance that will need to be noted. The first distinction is the difference between drawing an item with or without replacement.

## Exercise Set A

1. A computer is programmed to compute various chances. Match the numerical answers with the verbal descriptions (which may be used more than once).
|Numerical Answer|Verbal description|
|---|---|
|(a) -50%|(i) This is as likely to happen as not.|
|(b) 0%|(ii) This is very likely to happen, but it's not certain.|
|(c) 10%|(iii) This won't happen.|
|(d) 50%|(iv) This may happen, but it's not likely.|
|(e) 90%|(v) This will happen, for sure.|
|(f) 100%|(vi) There's a bug in the program.|
|(g) 200%||

_solution_.  
a -> vi  
b -> iii  
c -> iv  
d -> i  
e -> ii  
f -> v  
g -> vi

2. A coin will be tossed $1,000$ times. About how many heads are expected?

_solution_.
$500$ heads

3. A die will be rolled $6,000$ times. About how many aces are expected?

_solution_.
$1,000$ aces

4. In five-card draw poker, the chance of being dealt a full house (one pair and three of a kind) is $0.14$ of $1\%$. If $10,000$ hands are dealt, about how many will be a full house?

_solution_.
\begin{align*}
10,000 \times .0014 = 14
\end{align*}
