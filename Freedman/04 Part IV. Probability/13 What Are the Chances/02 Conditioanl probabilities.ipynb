# Conditional probabilities

To introduce conditional probabilities, we will be looking at a deck of cards with the following characteristics.
* has four suits: clubs, diamonds, hearts, spades.
* 13 cards in each suit: 2&ndash;10, jack, queen, king, ace.
* There are 52 cards in the deck.

<a id=example2></a>
_Example 2_.&emsp;A deck of cards is shuffled, and the top two cards are put on a table, face down. You with $\$1$ if the second card is the queen of hearts.  

&emsp;(a)&ensp;What is your chance of winning the dollar?  
&emsp;(b)&ensp;You turn over the first card. It is the seven of clubs. Now, what is your chance of winning?

_Solution.&ensp;Part(a)_.&emsp;The bet is about the second card, not the first. The chance of winning is $1/52$.  
_Solution.&ensp;Part(b)_.&emsp;After the seven of clubs is drawn, there are $51$ cards left. Therefore, the chances of drawing a queen of hearts is $1/51$.

## Exercise Set B

1. Two tickets are drawn at random without replacement form the box `| 1, 2, 3, 4 |`.

&emsp;&emsp;(a)&ensp;What is the chance that the second ticket is `4`?  
&emsp;&emsp;(b)&ensp;What is the chance that the second ticket is `4`, given the first is `2`?

_Solution_.&emsp;(a) The chance of a `4` is $1/4$ or $25\%$ chance. (b) If a `2` is drawn first, then the chances of drawing a `4` increases to $1/3$ or $33\%$.

2. Repeat exercise 1, if the draws are made with replacement.

_Solution_. If the draws are made with replacement, then (a) stays the same, but (b) reduces back to a $1/4$ or $50\%$ chance.

3. A penny is tossed 5 times.

&emsp;&emsp;(a)&ensp;Find the chance that the 5th toss is a head.  
&emsp;&emsp;(b)&ensp;Find the chance that the 5th toss is a head, given the first 4 are tails.  

_Solution_. For both (a) and (b), the chances for the 5th coin toss to be heads are the same: $50\%$ because they do not depend on the previous coin tosses.

4. Five cards are dealt off the top of a well-shuffled deck.

&emsp;&emsp;(a)&ensp;Find the chance that the 5th card is the queen of spades.  
&emsp;&emsp;(b)&ensp;Find the chance that the 5th card is the queen of spades, given that the first 4 cards are hearts.

_Solution_. For (a), the chance is $1/52$. For (b) the 5th card does depend on the outcomes of the previous four cards so the chance is $1/48$.
