# The addition rule

Two things are _mutually exclusive_ when the occurrence of one prevents the occurrence of the other: one excludes the other.

_Example 3._  A card is dealt off the top of a well-shuffled deck. The card might be a heart. Or, it might be a spade. Are these two possibilities mutually exclusive?  
_Solution._ If the card is a heart, it can't be a spade. These two possibilities are mutually exclusive.

_Addition RUle._ To find the chance that at least one of two things will happen, check to see if they are mutually exclusive. If the are, add the chances.

_Example 4._ A card is dealt off the top of a well-shuffled deck. There is 1 chance in 4 for it to be a heart. There is 1 chance in 4 for it to be a spade. What is the chance of it being in a major suit (hearts or spades)?  
_Solution._ The question asks for the chance that one of the following two things will happen:
* the card is a heart;
* the card is a spade.
As in example 3, if the card is a heart, it can't be a spade: these are mutually exclusive events. So it is legitimate to add the chances. The chance of getting a card in a major suit is $1/4 + 1/4 = 1/2$. (A check on the reasoning: there are 13 hearts and 13 spades, so $26/52 = 1/2$ of the cards in the deck are in a major suit).

_Example 5._ Someone throws a pair of dice. True or false: the chance of getting at least one ace is $1/6 + 1/6 = 1/3$.  
_Solution._ This is false. Image one of the dice is white, the other is black. The question asks for the chance that one of the following two things will happen:
* the white die lands ace;
* the black die lands ace.
A white ace does not prevent a black ace. These two events are not mutually exclusive, so the addition rule does not apply. Adding the chances gives the wrong answer.

Looking at figure 1, there are 6 ways for the white die to show ace. There are 6 ways for the black die to shoe ace. But the number of ways to get at least one ace is not $6 + 6$. Addition double counts the outcome of (ace, ace) at the top left corner. The chance of getting at least one ace is

\begin{align*}
(6 + 6 - 1)/36 = 11/36\text{, not }(6 + 6)/36 = 12/36 = 1/3$.

If you want to find the chance that at least one event occurs, and the events are not mutually exclusive, do not add the chances: the sum will be too big.

## Exercise Set B

1. Fifty children attended a party where cookies and ice cream were served: 12 children took cookies, and 17 took ice cream. True or false: 29 children must have had cookies or ice cream. Explain briefly.

_Solution_: False. If the two were mutually exclusive, 29 children would have had cookies or ice cream. However, they might not be mutually exclusive.

2. There are 20 dots in the diagram below, and 3 circles. The circles are labeled A, B, and C. One of the dots will be chosen at random.
   * What is the probability that the dot falls inside circle A?
   * What is the probability that the dot falls inside circle B?
   * What is the probability that the dot falls inside circle C?
   * What is the probability that the dot falls inside at least one of the circles?
![circles](./circles.jpg)

_Solution_
* The probability that the dot falls inside circle A is $\frac{4}{20} = \frac{1}{5}$.
* The probability that the dot falls inside circle B is $\frac{8}{20} = \frac{2}{5}$.
* The probability that the dot falls inside circle C is $\frac{12}{20} = \frac{3}{5}$.
* The probability that the dot falls inside at least one of the circles is $\frac{14}{20} = \frac{7}{10}$.

3. Two cards are dealt off the top of a well-shuffled dec. You have a choice:
   * to win $\$1$ if the first card is an ace or the second card is an ace;
   * to win $\$1$ if at least one of the two cards is an ace.
Which option is better? or are they the same? Explain briefly.

_Solution._ The are the same.

4. Two dice will be rolled. The chance that the first one lands an ace is $1/6$, and the chance that the second one lands a duce is $1/6$. True or false, the chance that the first one lands an ace or the second one lands duce equals $1/6 + 1/6$. Explain briefly.

_Solution._ False. The two events are not mutually exclusive since getting an ace on the fir e roll does not prevent getting a duce on the second roll.

5. A box contains 10 tickets numbered 1 through 10. Five draws will be made at random with replacement from this box. True or False: there are 5 chances in 10 of getting `|7|` at least once. Explain briefly.

_Solution._  False. Because the draws are made with replacement, they are not mutually exclusive.

6. A number is drawn at random from a box. There is a $20\%$ chance for it to be 10 or less. There is a $10\%$ chance for it to be 50 or more. True or false: the chance of getting a number between 10 and 50 (exclusive) is $70\%$. Explain briefly.

_Solutuion._  True. The chances should add to $100\%$.
