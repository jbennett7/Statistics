# The multiplication rule

The chance that two things will both happen equals the chance that the first will happen, multiplied by the chance that the second will happen given the first has happened.

<a id=example3></a>
_Example 3_.&emsp;A box has three tickets, colored red, white, and blue.  

&emsp;&emsp;&emsp;&emsp;`|R, W, B|`  

Two tickets will be drawn at random without replacement. What is the chance of drawing the red ticket and then the white?  

_Solution_. Imagine a large group of people. Each of these people holds the exact same box and draws two tickets at random without replacement. About one third of the people get `R` on the first draw. and are left with `|  , W, B|`. On the second draw, about half of these people will get `W`. The fraction who draw `R` `W` is therefore
\begin{align*}
\frac{1}{2}\text{ of }\frac{1}{3}=\frac{1}{2}\times\frac{1}{3}=\frac{1}{6}\text{.}
\end{align*}
The chance is 1 in 5, or $16\frac{2}{3}\%$.
Statisticians usually multiply the chances in reverse order:
\begin{align*}
\frac{1}{3}\times\frac{1}{2}=\frac{1}{6}\text{.}
\end{align*}
The reason: $1/3$ refers to the first draw, and $1/2$ to the second.

_Example 4_.&emsp;Two cards will be dealt off the top of a well-shuffled deck. What is the chance that the first card will be the seven of clubs and the second card will be the queen of hearts?

_Solution_.&emsp;This is like example 3, with a much bigger box. The chance that the first card will be the seven of clubs is $1/52$. Given that the first card was the seven of clubs, the chance that the second card will be the queen of hearts is $1/51$. The chance of getting both cards is
\begin{align*}
\frac{1}{52}\times\frac{1}{51}=\frac{1}{2,652}\text{.}
\end{align*}
This is a small chance: about 4 in $10,000$, or $0.04$ of $1\%$.

_Example 5_.&emsp;A deck of cards is shuffled, and two cards are dealt. What is the chance that both are aces?

_Solution_.&emsp;The chance that the first card is an ace equals $4/52$. Given that the first card is an ace, there are 3 aces among the 51 remaining cards. So the chance of a second ace equals $3/51$. The chance that both cards are aces equals
\begin{align*}
\frac{4}{52}\times\frac{3}{51}=\frac{12}{2,652}\text{.}
\end{align*}
This is about 1 in 200, or $1/2$ of $1\%$.

_Example 6_.&emsp;A coin is tossed twice. What is the chance of a head followed by a tail?

_Solution_.&emsp;The chance of a head on the first toss equals $1/2$. No matter how the first toss turns out, the chance of tials on the second toss equals $1/2$. So the chance of heads followed by tails equals
\begin{align*}
\frac{1}{2}\times\frac{1}{2}=\frac{1}{4}\text{.}
\end{align*}

## Exercise Set C

1. A deck is shuffled and two cards are dealt.

&emsp;&emsp;(a)&ensp;Find the chance that the second card is a heart given the first card is a heart.  
&emsp;&emsp;(b)&ensp;Find the chance that the first cards is a heart and the second card is a heart.

_Solution_.&emsp;For (a) the chance of getting a heart is $13/52$. If a heart is drawn on the first then the chance of getting a heart on the second draw, given a heart was drawn on the first, is $12/51$.
\begin{align*}
\frac{13}{52}\times\frac{12}{51}=\frac{156}{2,652}=\frac{1}{17}=6\%
\end{align*}

2. A die is rolled three times.

&emsp;&emsp;(a)&ensp;Find the chance that the first roll is an ace.  
&emsp;&emsp;(b)&ensp;Find the chance that the first role is an ace, the second role is a deuce, and the third roll is a trey.

_Solution_.&emsp;For (a), the chance that the first role is an ace is $1/6$ or $16\frac{2}{3}\%$. For (b), the chances of each are independent so their combination is:
\begin{align*}
\frac{1}{6}\times\frac{1}{6}\times\frac{1}{6}=\frac{1}{216}
\end{align*}

3. A deck is shuffled and three cards are dealt.

&emsp;&emsp;(a)&ensp;Find the chance that the first card is a king.  
&emsp;&emsp;(b)&ensp;Find the chance that the first card is a king, the second is a queen, and the third is a jack.

_Solution_.&emsp;For (a), the chance is $1/52$. For (b), the chances are dependent on the previous draw:
\begin{align*}
\frac{4}{52}\times\frac{4}{51}\times\frac{4}{50}=\frac{64}{132600}=.048\%
\end{align*}

4. A die will be rolled six times. You have a choice&ndash;

&emsp;&emsp;(i)&ensp;to win $\$1$ if at least one ace shows  
&emsp;&emsp;(ii)&ensp;to win $\$1$ if an ace shows on all the rolls  
Which option offers the better chance of winning? Or are they the same? Explain.

_Solution_.&ensp;Option (i) is the better choice. For (i) you have a $1/6$ chance of winning. For option (ii) you have a chance of:
\begin{align*}
\frac{1}{6}\times\frac{1}{6}\times\frac{1}{6}\times\frac{1}{6}\times\frac{1}{6}\times\frac{1}{6}=\frac{1}{46,656}
\end{align*}

6. "A cat-o'nine-___ can be used to punish ___ of state, but this is seldom done." A coin is tossed twice to fill in the blanks. What is the chance of the coin getting it right?

_Solution_. The answer to fill in the blanks are `tails` and `heads`. The chances that a coin tossed twice would get a `tails` followed by a `heads` are: $\frac{1}{2}\times\frac{1}{2}=\frac{1}{4}$

7. A cion is tossed $3$ times.

&emsp;&emsp;(a)&ensp;What is the chance of getting 3 heads?  
&emsp;&emsp;(b)&ensp;What is the chance of not getting 3 heads?  
&emsp;&emsp;(c)&ensp;What is the chance of getting at least 1 tail?  
&emsp;&emsp;(d)&ensp;What is the chance of getting at least 1 head?

_Solution_.&ensp;For (a) the chances are $\frac{1}{2}\times\frac{1}{2}\times\frac{1}{2}=\frac{1}{8}$. For (b) the chance is the complement of (a) $1 - \frac{1}{8} = \frac{7}{8}$. For (c) the chance of getting at least 1 tail is the same as the chance of not getting 3 heads: $\frac{7}{8}$. For (d) the chance of getting at least one  head is the same as the chance of not getting 3 tails: $\frac{7}{8}$.
