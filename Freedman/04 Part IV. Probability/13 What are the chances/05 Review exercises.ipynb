# Review Exercises

1. True or false, and explain:

&emsp;&emsp;(a)&ensp;If something has probability $1,000\%$, it is sure to happen.  
&emsp;&emsp;(b)&ensp;If something has probability $90\%$, it can be expected to happen about nine times as often as its opposite.

_Solution_.&emsp;For (a), the answer is False. $1,000\%$ is not mathematically possible. For (b), the answer is True. Although it is not guaranteed.

2. Two cards will be dealt off the top of a well-shuffled deck. You have a choice:

&emsp;&emsp;(i)&ensp;To win $\$1$ if the first is a king.  
&emsp;&emsp;(ii)&ensp;To win $\$1$ if the first is a king and the second is a queen.  

Which option is better? Or are they equivalent? Explain briefly.

_Solution_.&emsp;The answer is (i). For (i), there is a $4/52$ chance of getting a king. For (ii), the chance of getting a queen, given the first card is a king, is $\frac{4}{52}\times\frac{4}{51}=\frac{16}{2,656}$.

3. Four cards will be dealt off the top of a well-shuffled deck. There are two options:

&emsp;&emsp;(a)&ensp;To win $\$1$ if the first card is a club and the second is a diamond and the third is a heart and the fourth is a spade.  
&emsp;&emsp;(b)&ensp;To win $\$1$ if the four cards are of four different suits.

&emsp;&emsp;Which option is better? Or are they the same? Explain.

_Solution_.&emsp;The option b is better. This is because (a) is a particular sequence and (b) is a set of sequences. There are $4\times3\times2\times1=4!=24$ sequences.
The chances of (a) is
\begin{align*}
\frac{13\times13\times13\times13}{52\times51\times50\times49}=\frac{28,561}{6,497,400}\approx0.00439\text{,}
\end{align*}
or $0.439\%$. The chance for (b) is $24\times0.00439=0.10536$.

4. A poker hand is dealt. Find the chance that the first four cards are aces and the fifth is a king.

_Solution_.&emsp;
\begin{align*}
\frac{4}{52}\times\frac{3}{51}\times\frac{2}{50}\times\frac{1}{49}\times\frac{4}{48}=\frac{96}{311,875,200}

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;| 1, 1, 8, `1`, `1`, `8` |

_Solution_.&emsp;The color and number are independent. There is a 1 in 3 chance of getting an "8" from both colors. There is also a 2 in 3 chance of getting a "1" from both colors.

6. A deck of cards is shuffled and the top two cards are placed face down on a table. True or false, and explain:

&emsp;&emsp;(a)&ensp;There is 1 chance in 52 for the first card to be the ace of clubs.  
&emsp;&emsp;(b)&ensp;There is 1 chance in 52 for the second card to be the ace of diamonds.  
&emsp;&emsp;(c)&ensp;The chance of getting the ace of clubs and then the ace of diamonds is $1/52\times1/52$.

_Solution_.&emsp;For (a) and (b), its true. For (c), false because the second event depends on the first so the chance of getting the ace of clubs and then the ace of diamonds is $1/52\times1/51$.

7. A coin is tossed six times. Two possible sequences of results are

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;
(i)&emsp;H&emsp;T&emsp;T&emsp;H&emsp;T&emsp;&emsp;&emsp;&emsp;
(ii)&emsp;H&emsp;H&emsp;H&emsp;H&emsp;H&emsp;H  

(The coin must land H or T in the order given; H = heads, T = tails.) Which of the following is correct? Explain.

&emsp;&emsp;(a)&ensp;Sequence (i) is more likely.  
&emsp;&emsp;(b)&ensp;Sequence (ii) is more likely.  
&emsp;&emsp;(c)&ensp;Both sequences are equally likely.

_Solution_.&emsp;Each coin toss is independent and has a $50\%$ chance of either heads or tails. Therefore, they are both equally likely. The chance is $\frac{1}{2}\times\frac{1}{2}\times\frac{1}{2}\times\frac{1}{2}\times\frac{1}{2}\times\frac{1}{2}=\frac{1}{64}$

8. A die is rolled four times. What is the chance that&mdash;

&emsp;&emsp;(a)&ensp;all the rolls show 3 or more spots?  
&emsp;&emsp;(b)&ensp;none of the rolls show 3 or more spots?
&emsp;&emsp;(c)&ensp;not all the rolls show 3 or more spots?

_Solution_.&emsp;For (a), the chance is $\frac{4}{6}\times\frac{4}{6}\times\frac{4}{6}\times\frac{4}{6}=\frac{256}{1,296}=0.2$. For (b), the chance is $\frac{2}{6}\times\frac{2}{6}\times\frac{2}{6}\times\frac{2}{6}=\frac{16}{1,296}=0.012$. For (c), the chance is the complement of the first $1-0.2=0.8$.

9. A die is rolled $10$ times. Find the chance of&mdash;

&emsp;&emsp;(a)&ensp;getting $10$ sixes.  
&emsp;&emsp;(b)&ensp;not getting $10$ sixes.  
&emsp;&emsp;(c)&ensp;all the rolls showing $5$ spots or less.

_Solution_.&emsp;For (a) the chance is $(\frac{1}{6})^{10}=1.65\times10^{-8}$, which is nearly impossible. For (b), the chances are near certainty. The chance for (c) is $(\frac{5}{6})^{10}=0.16$.

10. Which of the two options is better, or are they the same. Explain briefly.

&emsp;&emsp;(i)&ensp;You toss a coin $100$ times. On each toss, if the coin lands heads, you win $\$1$. If it lands tails, you lose $\$1$.  
&emsp;&emsp;(ii)&ensp;You draw $100$ times at random with replacement from |1, 0|. On each draw, you are paid (in dollars) the number on the ticket.

_Solutuion_.&emsp;For both options, each event is independent. Option (i) is a wash, you have a $50\%$ chance of winning or loosing so you win no money. Option (ii) you gain $\$50$, so option (ii) is better.

12. You are thinking about playing a lottery. The rules: you buy a ticket, choose 3 different numbers from $1$ to $100$, and write them on the ticket. The lottery has a box with 100 balls numbered from 1 through 100. Three balls are drawn at random without replacement. If the numbers on these balls are the same as the numbers on your ticket, you win. (Order doesn't matter.) If you decide to play, what is your chance of winning?

_Solution_.
\begin{align*}
\frac{1}{100}\times\frac{1}{99}\times\frac{1}{98}=\frac{1}{970,200}=1\times10^{-6}
\end{align*}
