Independence

Two things are _independent_ if the chances for the second given the first are the same, no matter how the first one turns out. Otherwise the two are _dependent_.

If two things are independent, the chance that both will happen equals the product of their unconditional probabilities. This is a special case of the multiplication rule.

_Example 7_.&emsp;Someone is going to toss a coin twice. If the coin lands heads on the second toss, you win a dollar.

&emsp;&emsp;(a)&ensp;If the first toss is heads, what is your chance of winning the dollar?  
&emsp;&emsp;(b)&ensp;If the first toss is tails, what is your chance of winning the dollar?  
&emsp;&emsp;(c)&ensp;Are the tosses independent?

_Solution_.&emsp;No matter what the first toss is, the chances of getting heads on the second toss is the same $\frac{1}{2}$. This is independence.

_Example 8_.&emsp;Two draws will be made at random with replacement from

&emsp;&emsp;&emsp;&emsp;`| 1, 1, 2, 2, 3 |`

&emsp;&emsp;(a)&ensp;Suppose the first draw is `1`. What is the chance of getting a `2` on the second draw?  
&emsp;&emsp;(b)&ensp;Suppose the first draw is a `2`. What is the chance of getting `2` on the second draw?  
&emsp;&emsp;(c)&ensp;Are the draws independent?

_Solution_.&emsp;Whether the first draw is `1` or `2` or anything else, the chance of getting `2` on the second draw stays the same&mdash;two in five, or $40\%$. Since after each draw, the item drawn is replaced, the draws are independent.

_Example 9_.&emsp;As in example 8, but the draws are made without replacement.

_Solution_.&emsp;Since the draws are not replaced, each draw is no longer independent. For (a), the chance of drawing a `2` given a `1` is drawn previously is $\frac{2}{4}=\frac{1}{2}$ or $50\%$. For (b), the chance of drawing a `2` if a `2` was drawn previously is $\frac{1}{4}$ or $25\%$.

_Example 10_.&emsp;A box has three ticket, colored red, white, and blue.

&emsp;&emsp;&emsp;&emsp;`| R, W, B|`  

Two tickets will be drawn at random with replacement. What is the chance of drawing the red ticket and then the white?

_Solution_.&emsp;The draws are independent, so the chance is
\begin{align*}
\frac{1}{3}\times\frac{1}{3}=\frac{1}{9}\text{.}
\end{align*}
Compare this with [example 3](03%20The%20multiplicaiton%20rule.ipynb#example3). The answers are different. Independence matters. And it's easier this time: you don't need to work out conditional probabilities.

1. For each of the following boxes, say whether color and number are dependent or independent.

&emsp;&emsp;(a)&ensp;| 1, 2, 2, `1`, `2`, `2` |

&emsp;&emsp;(b)&ensp;| 1, 2, '1', '2', `1`, `2` |  

&emsp;&emsp;(c)&ensp;| 1, 2, 3, `1`, `1`, `3` |

_Solution_.&emsp;For (a), color and number are independent, since there is an equal chance of getting a 1 or 2 for both colors. For (b), color and number are independent as well. For (c), they are dependent. With white tickets there is only 1 chance in 3 to get "1" and with black tickets, there are 2 chances in 3.

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;| <1|2>, <1|3>, <4|2>, <4|3> |

&emsp;&emsp;&emsp;&emsp;(For instance, the first number on <4|2> is 4 and the second is 2.) A ticket is drawn at random. Are the two numbers dependent or independent?

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;| <1|2>, <1|3>, <1|3>, <4|2>, <4|3>, <4|3> |

3. Every week you buy a ticket in a lottery that offers one chance in a million of winning. What is the chance that you never win, even if you keep this up for ten years?

_Solution_. Ten years is $520$ weeks. The chance of not winning is $(999,999/1,000,000)^{520}\approx0.9995$.

4. Two draws are made at random without replacement from the box `|1, 2, 3, 4|`. The first ticket is lost, and nobody knows what was written on it. True or false, and explain: the two draws are independent.

_Solution_.&emsp;False. To figure out whether two things are independent or not, you pretend to know how the first one turned out, and then see if the chances for the second change.

5. Suppose that in a certain class there are
- $80\%$ men and $20\%$ women;
- $15\%$ freshmen and $85\%$ sophomores

&emsp;&emsp;(a)&ensp;The percentage of sophomore women in the class can be as small as __.  
&emsp;&emsp;(b)&ensp;This percentage can be as large as __.

_Solution_.&emsp;(a) $5\%$, (b) $20\%$  

To figure (a) out, suppose you have $80$ men and $20$ women in the class. You also have $15$ cards marked "freshman" and $85$ cards marked "sophomore." You want to give out a card to each student, so that as few women as possible get "sophomore." The strategy is to give a sophomore card out each man: you are left with $5$, which have to go to $5$ women. The $15$ freshman cards go to the other $15$ women.  
_Comment_.&emsp;If year and sex are independent, the percentage of sophomore women would be $85\%$ of $20\%=17\%$, between the two extremes.

8. (Hard) In a certain psychology experiment, each subject is presented with three ordinary playing cards, face down. The subject takes one of these cards.
