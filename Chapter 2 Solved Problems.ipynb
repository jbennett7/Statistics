7. In answering Problem 6(a), a student chose the class intervals $5.10$&mdash;$5.40$, $5.40$&mdash;$5.70$,$\ldots$, $6.90$&mdash;$7.20$, and $7.20$&mdash;$7.50$. Was there anything wrong with this choice?

### Solution
These class intervals overlap at $5.40$, $5.70$,$\ldots$, $7.20$. Thus a measurement recorded as $5.40$, for example, could be placed in either of the first two class intervals. Some statisticians justify this choice by agreeing to place half of such ambiguous cases in one class and half in the other.<p>
    The ambiguity is removed by writing the class intervals as $5.10$ to under $5.40$, $5.40$ to under $5.70$, etc. In this case, the class limits coincide with the class boundaries, and the class marks can coincide with the observed data.<p>
        In general it is desirable to avoid overlapping class intervals whenever possible and to choose them so that the class boundaries are values not coinciding with actual observed data. For example, the class intervals for Problem $6$ could have been chosen as $5.095$&mdash;$5.395$, $5.395$&mdash;$5.695$, etc., without ambiguity. A disadvantage of this particular choice is that the class marks do not coincide with the observed data.

## Cumulative-Frequency Distributions and Ogives

14. For Table 2.6, Construct:  

(a) a cumulative-frequency distribution, (b) a percentage cumulative distribution,

```r
load("Saves/Table_2.6.Rdata")
p14.cf.table <- matrix(c(0, cumsum(Table_2.6)), nrow=length(Table_2.6)+1)
rownames(p14.cf.table) <- c("Less than $250.00", "Less than $260.00", "Less than $270.00", "Less than $280.00", "Less than $290.00",
                            "Less than $300.00", "Less than $310.00", "Less than $320.00")
p14.cf.table <- cbind(p14.cf.table, c(0, round(cumsum(Table_2.6)/sum(Table_2.6)*100,1)))
colnames(p14.cf.table) <- c("Cumulative Frequency", "Percentage Cumulative Distribution")
p14.cf.table
```

(c) an ogive, and (d) a percentage ogive  from the frequency distribution in Table 2.6.

```r
p14.cf.X <- c(250, 260, 270, 280, 290, 300, 310, 320)
p14.cf.f <- as.vector(p14.cf.table[,1])
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8)
plot(p14.cf.X, p14.cf.f, type="l", col="red", main="Cumulative Frequency Wages", xlab="Wages",ylab="Frequency")
points(p14.cf.X, p14.cf.f, pch=20)
plot(p14.cf.X, p14.cf.f, type="l", col="red", main="Percentage Cumulative Frequency Wages", xlab="Wages", ylab="Frequency")
points(p14.cf.X, p14.cf.f, pch=20)
options(repr.plot.width=7, repr.plot.height=7)
```
