```r
# 2.2 The final grades in mathematics of 80 students at State University are recorded in the accompanying table.
grades <- c( 68, 84, 75, 82, 68, 90, 62, 88, 76, 93,
             73, 79, 88, 73, 60, 93, 71, 59, 85, 75,
             61, 65, 75, 87, 74, 62, 95, 78, 63, 72,
             66, 78, 82, 75, 94, 77, 69, 74, 68, 60,
             96, 78, 89, 61, 75, 95, 60, 79, 83, 71,
             79, 62, 67, 97, 78, 85, 76, 65, 71, 75,
             65, 80, 73, 57, 88, 78, 62, 76, 53, 74,
             86, 67, 73, 81, 72, 63, 76, 75, 85, 77)

# With reference to this table, find:
# (a) The highest grade.
print(paste("highest grade: ", max(grades), sep=""))

# (b) The lowest grade.
print(paste("lowest grad: ", min(grades), sep=""))

# (c) The range.
print(paste("range of grades: ", diff(range(grades)), sep=""))

# (d) The grades of the five highest-ranking students.
print("five highest scores: ")
tail(sort(grades))

# (e) The grades of the five lowest-ranking students.
print("five lowest scores: ")
head(sort(grades))

# (f) The grade of the student ranking tenth highest.
print(paste("tenth highest score: ", sort(grades)[length(grades)-9], sep=""))

# (g) The number of students who received grades of 75 or higher.
print(paste("number of students >= 75: ", sum(grades >= 75), sep=""))

# (h) The number of students who received grades below 85.
print(paste("number of students < 85: ", sum(grades < 85), sep=""))

# (i) The percentage of students who received grades higher than 65 but not higher than 85.
print(paste("percentage of students > 65 <= 85: ", (length(grades) - sum(grades <= 65) - sum(grades > 85))/length(grades) * 100, "%", sep=""))

# (j) The grades that did not appear at all.
print("grades not in the set: ")
setdiff(0:100, grades)

# Creating Distribution tables
bins <- seq(50,100,by=5)
scores <- cut(grades, bins)
table(scores)
#transform(table(scores),Rel_Freq=prop.table(Freq),Cum_Freq=cumsum(Freq))
```

### Frequency distributions, histograms, and frequency polygons

```r
# 2.4 If the class marks in a frequency distribution of the weights of students are
# 128, 137, 146, 155, 164, 173, and 182 pounds (lb), find (a) the class-interval size,
# (b) the class boundaries, and (c) the class limits, assuming that the weights were
# measured to the nearest pound.
marks <- c( 128, 137, 146, 155, 164, 173, 182)

# (a)
cis <- class.interval.size <- 137-128
class.interval.size

# (b)
cb <- class.boundaries <- unique(sort(c(marks-cis/2,marks+cis/2)))
cb

# (c)
lower <- ceiling(cb[1])
upper <- floor(tail(cb,1))
ll <- lower.limit <- seq(lower,upper,by=9)
ul <- upper.limit <- seq(lower+8,upper+8,by=9)
cl <- class.limits <- c()
```

```r
# 2.6 The smallest of 150 measurements is 5.18 in, and the largest is 7.44 in. Determine a suitable set of
# (a) class intervals, (b) class boundaries, and (c) class marks that might be used in forming a frequency
# distribution of these measurements.

# (a)
interval <- .2
inc <- interval-.01
l.I <- lower.class.I <- seq(5.10, 7.3, by=interval)
u.I <- upper.class.I <- seq(5.10+inc, 7.3+inc, by=interval)
interval <- .3
inc <- interval-.01
l.II <- lower.class.II <- seq(5.10, 7.3, by=interval)
u.II <- upper.class.II <- seq(5.10+inc, 7.3+inc, by=interval)
interval <- .4
inc <- interval-.01
l.III <- lower.class.III <- seq(5.10, 7.3, by=interval)
u.III <- upper.class.III <- seq(5.10+inc, 7.3+inc, by=interval)

#(b)
boundary.increment <- .005
l.I-boundary.increment
u.I+boundary.increment
l.II-boundary.increment
u.II+boundary.increment
l.III-boundary.increment
u.III+boundary.increment
```

```r
# 2.10 The data is the one-way commuting distances for 50 Metropolitan College students in miles.

# The histogram is displayed. Answer the follwing questions:
# (a) What data values are in the first class?
# (b) What data values are in the second class?
# (c) What data values are in the third class?
# (d) What data values are in the fourth class?
# (e) What data values are in the fifth class?
# (f) What data values are in the sixth class?

data <- c( 4.3,  7.0,  8.0,  3.9,  3.7,  8.4, 2.6,  1.0,  15.7, 3.9,
           6.5,  8.7,  0.9,  0.9,  12.6, 4.0, 10.3, 10.0, 6.2,  1.1,
           7.2,  8.8,  7.8,  4.9,  2.0,  3.0, 4.2,  3.3,  4.8,  4.4,
           7.7,  2.4,  8.0,  8.0,  4.6,  1.4, 2.2,  1.9,  3.2,  4.8,
           5.0,  10.3, 12.3, 3.8,  3.8,  6.6, 2.0,  1.6,  4.4,  4.3)
h <- hist(data, right=FALSE)
h$counts
j <- 1
for(i in h$counts){
    k <- j + i - 1
    class <- sort(data)[j:k]
    if(! is.na(class[1])){
        print(class)
    }
    j <- k + 1
}
#sort(data)[1:7]
#sort(data)[8:20]
#sort(data)[21:31]
#sort(data)[32:38]
#sort(data)[39:44]
#sort(data)[45:47]
#sort(data)[48:49]
#sort(data)[50]
```
