## Generating sample data from a distribution table

Reference table is Table 2.1 from page 36.

List of important functions:
1. sample - used to generate random integers, you can use `runif` to generate random decimal numbers.
2. cut - groups the sample data into intervals.
3. table - creates a contingency table.
4. transform - converts the table to a dataframe.

```r
lower.limits <- c(60, 63, 66, 69, 72)
upper.limits <- c(62, 65, 68, 71, 74)
f <- c(5, 18, 42, 27, 8)
s1 <- c()
for(i in 1:length(f)){
    s1 <- sample(c(s1, sample(lower.limits[i]:upper.limits[i], f[i], replace=TRUE)))
}
boundaries <- seq(59.5,74.5,by=3)
transform(table(cut(s1, breaks=boundaries)))
```
