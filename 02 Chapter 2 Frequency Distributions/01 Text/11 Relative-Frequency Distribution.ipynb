# Relative-frequency distributions
The _relative frequency_ of a class is the frequency of the class divided by the total frequency of all classes and is generally expressed as a percentage. For example, the relative frequency of the class $66$&mdash;$68$ in Table 2.1 is $42/100=42\%$. The sum of the relative frequencies of all classes is clearly $1$, or $100\%$.  
If the frequencies in Table 2.1 are replaced with the corresponding relative frequencies, the resulting table is called a _relative-frequency distributution_, _percentage distribution_, or _relative-frequency table_.  
Graphic representation of relative-frequency distributions can be obtained from the histogram or frequency polygon simply by changing the vertical scale from frequency to relative frequency, keeping exactly the same diagram. The resulting graphs are called _relative-frequency histograms_ (or _percentage histograms_) and _relative-frequency polygons_ (or _percentage polygons_), respectively.

# Relative-frequency distributions in R
The function `prop.table` will generate the relative proportions of the frequencies

```r
load("../../Saves/Problem_2.2.Rdata")
P2.2.data
```

```r
r <- range(P2.2.data)
bins <- seq(r[1], r[2], by=(diff(r)/11))
binned.data <- cut(P2.2.data, bins)
fq.table <- table(binned.data)
transform(fq.table, Rel_Freq=round(prop.table(as.vector(fq.table))*100,2))
```
