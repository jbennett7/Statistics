# Arrays
An _array_ is an arrangement of raw numerical data in ascending or descending order of magnitude. The difference between the largest and smallest numbers is called the _range_ of data. For example, if the largest height of $100$ male students is $74$ inches (in) and the smallest height is $60$in, the range is $74-60=14$in.

To convert an R vector to the definition of an array, sort the vector either ascending or descending. The range in R is a two value vector that contains the smallest and largest value. To get the range defined above use `diff`:

```r
cat("Sorted array:\n", sort(P2.2.data), "\n\n")
cat("Reverse sorted array:\n", rev(sort(P2.2.data)), "\n\n")
cat("Range:", diff(range(P2.2.data)), "\nSmallest value:", range(P2.2.data)[1], "\nLargest value:", range(P2.2.data)[2], "\n")
```
