# The class mark
The _class mark_ is the midpoint of the class interval and is obtained by adding the lower and upper class limits and by dividing by $2$. Thus the class mark of the interval `60-62` is $(60+62)/2=61$. The class mark is also called the _class midpoint_.

For purposes of further mathematical analysis, all observations belonging to a given class interval are assumed to coincide with the class mark. Thus all heights in the class interval `60-62` in are considered to be $61$ in.
