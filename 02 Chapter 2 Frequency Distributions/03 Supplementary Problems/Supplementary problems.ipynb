## Problem 2-21
Construct (a) a histogram and (b) a frequency polygon corresponding to the frequency distribution of Table [2.15](#Table-2-15).

```r
p2.21.breaks <- seq(299.5, 1199.5, by=100)
p2.21.f <- as.vector(Table.2.15[,2])
p2.21.h <- list(breaks=p2.21.breaks, counts=p2.21.f, density=p2.21.f/diff(p2.21.breaks), xname="Minutes", yname="Frequency")
class(p2.21.h) <- "histogram"
p2.21.marks <- seq(349.5, 1149.5, by=100)
```

## Problem 2-23
For the data in Table [2.15](#Table-2-15), construct (a) a cumulative-frequency distribution, (b) a percentage cumulative distribution, (c) an ogive, and (d) a percentage ogive. (Note that unless otherwise specified, a cumulative distribution refers to one made on a "less than" basis.)

```r
# insert 0 at the beginning of vector for values less than first class interval.
p2.23.cumsum <- c( 0, cumsum(Table.2.15[,2]))
p2.23.pct.cumsum <- round(p2.23.cumsum/sum(Table.2.15[,2])*100,2)
p2.23.marks <- seq(349.5, 1149.5, by=100)
p2.23.viewing.time <- as.factor(c("less than 300", "less than 400", "less than 500",
                                  "less than 600", "less than 700", "less than 800",
                                  "less than 900", "less than 1000", "less than 1100",
                                  "less than 1200"))
P2.23.table <- data.frame(Viewing.Time=p2.23.viewing.time, Number.of.Students=p2.23.cumsum)
P2.23.table.2 <- data.frame(Viewing.Time=p2.23.viewing.time, Percent.of.Students=p2.23.pct.cumsum)
```

(a) Cumulative-frequency distribution

```r
P2.23.table
```

(b) Relative cumulative-frequency distribution

```r
P2.23.table.2
```

(c) Ogive, and (d) a percentage ogive

```r
# Remove the inserted 0 when constructing the ogives.
cumsum <- p2.23.cumsum[2:length(p2.23.cumsum)]
pct.cumsum <- p2.23.pct.cumsum[2:length(p2.23.cumsum)]
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8)
plot(p2.23.marks,cumsum,type="l",col="red",main="Cumulative Frequency",xlab="Viewing Time (minutes)",ylab="Frequency")
points(p2.23.marks,cumsum,pch=20)
plot(p2.23.marks,pct.cumsum,type="l",col="red",main="Percentage Cumulative Frequency", xlab="Viewing Time (minutes)",ylab="Frequency")
points(p2.23.marks,pct.cumsum,pch=20)
options(repr.plot.width=7, repr.plot.height=7)
```

## Problem 2-27
The following table shows the diameters in centimeters (cm) of a sample of $60$ ball bearings manufactured by a company. Construct a frequency distribution of the diameters, using appropriate class intervals.

```r
p2.27.data <- c( 1.738, 1.729, 1.743, 1.740, 1.736, 1.741, 1.735, 1.731, 1.726, 1.737,
                 1.728, 1.737, 1.736, 1.735, 1.724, 1.733, 1.742, 1.736, 1.739, 1.735,
                 1.745, 1.736, 1.742, 1.740, 1.728, 1.738, 1.725, 1.733, 1.734, 1.732,
                 1.733, 1.730, 1.732, 1.730, 1.739, 1.734, 1.738, 1.739, 1.727, 1.735,
                 1.735, 1.732, 1.735, 1.727, 1.734, 1.732, 1.736, 1.741, 1.736, 1.744,
                 1.732, 1.737, 1.731, 1.746, 1.735, 1.735, 1.729, 1.734, 1.730, 1.740)
```

### Producing a frequency distribution by hand
Following the [General rules for forming frequency distributions](../01%20Text/08%20General%20Rules%20for%20Forming%20Frequency%20Distributions.ipynb#General-rules-for-forming-frequency-distributions)

```r
cat("The range of the ball bearings diameters:", diff(p2.27.range <- range(p2.27.data)), "\n")
cat("Choose the upper and lower range limits of the data set:\n\t")
cat("Lower:", (p2.27.lower <- floor(p2.27.range[1]*100)/100), "\n\t")
cat("Upper:", (p2.27.upper <- ceiling(p2.27.range[2]*100)/100), "\n")
cat("Choosing a class interval size:", p2.27.ci <- (p2.27.ul.range <- p2.27.upper - p2.27.lower)/6, "\n")
cat("Choosing the class boundaries:", (bins <- seq(p2.27.lower, p2.27.upper, by=p2.27.ci)), "\n")
Diameter <- cut(p2.27.data, bins)
P2.27.table <- transform(table(Diameter))
```

```r
P2.27.table
```

### A shorter method
You can use the `hist` function to automatically set the break points, and the counts.

```r
p2.27.h <- hist(p2.27.data,plot=FALSE)
Diameter <- cut(p2.27.data, p2.27.h$breaks)
P2.27.table.2 <- transform(table(Diameter))
p2.27.marks <- p2.27.h$mids
```

```r
P2.27.table.2
```
