## Problem 2-21
Construct (a) a histogram and (b) a frequency polygon corresponding to the frequency distribution of Table [2.15](#Table-2-15).

```r
p2.21.breaks <- seq(299.5, 1199.5, by=100)
p2.21.f <- as.vector(Table.2.15)
p2.21.h <- list(breaks=p2.21.breaks, counts=p2.21.f, density=p2.21.f/diff(p2.21.breaks), xname="Minutes", yname="Frequency")
class(p2.21.h) <- "histogram"
p2.21.marks <- seq(349.5, 1149.5, by=100)
```

### Solution 2-21
<a id="Figure-p2-21"></a>

```r
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8)
plot(p2.21.h, main="Viewing Time")
plot(p2.21.marks, p2.21.f, type="l", col="red", xlab="Minutes", ylab="Frequency", main="Viewing Time")
points(p2.21.marks, p2.21.f, pch=20)
options(repr.plot.width=7, repr.plot.height=7)
```

# Problem 2.22
For the data in Table [2.15](#Table-2-15) of Problem [2.20](#Problem-2-20), construct (a) a relative-frequency distribution, (b) a relative-frequency histogram, and (c) a relative-frequency polygon.

### Solution 2-22
<a id="Figure-P2-22"></a>

```r
p2.22.breaks <- seq(299.5, 1199.5, by=100)
p2.22.f.relative <- as.vector(Table.2.15)/sum(Table.2.15)*100
table.P2.22 <- matrix(p2.22.f.relative, nrow=length(p2.22.f.relative))
colnames(table.P2.22) <- c("% of Students")
rownames(table.P2.22) <- rownames(Table.2.15)
p2.22.h <- list(breaks=p2.22.breaks, counts=p2.22.f.relative, density=p2.22.f.relative/diff(p2.22.breaks), xname="Minutes", yname="Relative Frequency")
class(p2.22.h) <- "histogram"
p2.22.marks <- seq(349.5, 1149.5, by=100)
```

```r
table.P2.22
```

```r
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8)
plot(p2.22.h, main="Relative")
plot(p2.22.marks, p2.22.f.relative, type="l", col="red", xlab="Minutes", ylab="Frequency", main="Relative")
points(p2.22.marks, p2.22.f.relative, pch=20)
options(repr.plot.width=7, repr.plot.height=7)
```
